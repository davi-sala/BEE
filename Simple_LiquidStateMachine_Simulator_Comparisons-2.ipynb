```python
%%file serial_SNN_simulator_comp.c

#include <stdlib.h>
#include <stdio.h>
#include <math.h>
#include <limits.h> // has the maximum values of types: e.g UCHAR_MAX

#include <string.h> // only for the memcpy inside the main()
#include <time.h> // only for the time comparisons inside the main()

//
// Sets the right function for the timing debug stuff for GCC LINUX(__linux) and OSX (__APPLE_CC__)
// in Linux it is necessary to add gcc (linker) option: -lrt
//
#if defined (__linux__) || defined (linux)
#include <time.h> // only for the time comparisons inside the main()
double clock_my_gettime() {
    //  time_t tv_sec; //seconds 
    //  long tv_nsec;  //NANOseconds    
    struct timespec now;
    clock_gettime(CLOCK_MONOTONIC, &now);
    return ((double)now.tv_sec*1E9 + (double)now.tv_nsec); //nanoseconds
}
#endif

#if defined (__APPLE_CC__)
#include <sys/time.h>
double clock_my_gettime(){
    // time_t       tv_sec;    //seconds since Jan. 1, 1970 
    // suseconds_t  tv_usec;   //and MICROseconds    
    struct timeval now;
    int rv = gettimeofday(&now, NULL);
    if (rv) return rv; //rv=0 means OK!
    return ((double)now.tv_sec*1E9 + (double)now.tv_usec*1E3); //nanoseconds
}
#endif





//
// Functions prototypes
unsigned SpkLiq_time_seed();

double SpkLiq_uniform_deviate(int seed);

void SpikingLiquid_Reset();

int SpikingLiquid_generate_connections();

void SpikingLiquid_init();

double SpkLiq_gaussrand(unsigned int my_seed);

double SpkLiq_euclidean_dist(int i, int j, int xd, int yd);

void SpkLiq_process_exc_spikes(int *spikes, double *weights, int number_of_spikes);

void SpkLiq_process_inh_spikes(int *spikes, double *weights, int number_of_spikes);

void SpikingLiquid_process_internal_spikes();

void SpikingLiquid_update(int *spikes_exc, int *spikes_inh, double *weights_exc, double *weights_inh, int size_exc, int size_inh);


// This is used to define when realloc is going to ask for more memory
// inside the SpikingLiquid_generate_connections function
#define MAX_CONNECTIONS 10



// SpkLiq_liquid_parametres
// #1:  CGupta       # Parameter used at the connection probability
// #2:  UMarkram     # Use (U) - Parameter used at the Dynamic Synapse
// #3:  DMarkram     # Time constant for Depression (tau_rec) - used at the Dynamic Synapse
// #4:  FMarkram     # Time constant for Facilitation (tau_facil) - used at the Dynamic Synapse
// #5:  AMaass       # Connection gain (weight) (in nA)
// #6:  Delay_trans  # Transmission delay
//
// The Dynamic Synapses (STP - Short Term Plasticity) and the Transmission delay are not implemented... yet.
//
double const SpkLiq_liquid_parametres[2][2][6] = {{
                                              { 0.1  ,  0.32 ,  0.144,  0.06 ,  -2.8  ,  0.8  },
                                              { 0.4  ,  0.25 ,  0.7  ,  0.02 ,  -3.0  ,  0.8  }
                                          },
                                          {
                                              { 0.2  ,  0.05 ,  0.125,  1.2  ,  1.6  ,  0.8  },
                                              { 0.3  ,  0.5  ,  1.1  ,  0.05 ,  1.2  ,  1.5  }
                                          }};


//
// Global variables
// These variables are global because they are used by most of the functions and would be a pain to pass as parametres!
// The first version of this algorithm was developed in an object-oriented way (Python), so these global variables were inside the main class.
// Yes, I know about global variables being bad practice, but sometimes it is faster to use them directly instead of moving memory contents, pointers, etc.
//

// User's input variables
int SpkLiq_net_shape[3]; // the sides of the parallelepiped modeling the liquid's shape {x,y,z} (unit less, because it's a relative measurement)
double SpkLiq_lbd_value; // value that controls the probability of a connection inside the liquid (see Maass,Markram 2002)
double SpkLiq_step; // simulation time step size (in seconds)
double SpkLiq_taum; // neuron membrane time constant (in seconds)
double SpkLiq_cm; // neuron membrane capacitance (in Farads)
double SpkLiq_taue; // excitatory neuron decay time constant (in seconds)
double SpkLiq_taui; // inhibitory neuron decay time constant (in seconds)
double SpkLiq_refractory[2]; // refractory periods {inhibitory, excitatory} (in seconds)
unsigned int SpkLiq_user_seeds[5]; //random seeds supplied by the user
double SpkLiq_vresets; // stores the voltage value the membrane must assume after a reset (refractory or spike) (in Volts)
double SpkLiq_vthres; // stores the membrane voltage value used to verify if the neuron is going to spike (in Volts)
double SpkLiq_vrest; // stores the voltage value the membrane must assume after total discharge (in Volts)
double SpkLiq_inhibitory_percentage; // the percentage of inhibitory neurons in the liquid (unit less, %)
double SpkLiq_membrane_rand[2]; //range of the possible values to the initialization of the membrane voltage (in Volts)
double SpkLiq_current_rand[2]; //range of the possible values to the initialization of the constant noisy offset current (in Amperes)
double SpkLiq_noisy_current_rand; //max range of the possible values to the initialization of the variable noisy current (in Amperes)




unsigned int SpkLiq_rand_r_seeds[5]; //stores the random seeds variables used by the rand_r function

// Pointers are used here because the memory size varies according to the user values
double *SpkLiq_neurons_membrane; //the membrane potentials for all neurons
double *SpkLiq_neurons_membrane_init; //the initial membrane potentials for all neurons
double *SpkLiq_neurons_exc_curr; //excitatory currents levels for all neurons
double *SpkLiq_neurons_inh_curr; //inhibitory currents levels for all neurons
double *SpkLiq_refrac_timer; //timers used to control the refractory period
int *SpkLiq_test_vthres; //controls which neuron is above the voltage threshold and should spike
double *SpkLiq_noisy_offset_currents; //values of the neuron model constant noisy offset currents
double *SpkLiq_noisy_currents; //values of the neuron model variable noisy currents
double *SpkLiq_refrac_values; //stores the refractory periods for all neurons in the liquid 
                                     //because the refractory period is different for inhibitory and excitatory neurons

int *SpkLiq_inhibitory_indices; //Indices of the inhibitory neurons inside the liquid
int *SpkLiq_excitatory_indices; //Indices of the excitatory neurons inside the liquid
int *SpkLiq_liquid_indices_shuffled; //used to generate the random SpkLiq_inhibitory_indices/SpkLiq_excitatory_indices


int *SpkLiq_pre_i; //stores the indices of the inhibitory=>? pre-synaptic connections in the liquid
int *SpkLiq_pos_i; //stores the indices of the inhibitory=>? pos-synaptic connections in the liquid
double *SpkLiq_w_i; //stores the weights of the connections above

int *SpkLiq_pre_e; //stores the indices of the inhibitory=>? pre-synaptic connections in the liquid
int *SpkLiq_pos_e; //stores the indices of the inhibitory=>? pos-synaptic connections in the liquid
double *SpkLiq_w_e; //stores the weights of the connections above

// int *SpkLiq_spiked_i; //indices of the inhibitory neurons that spiked in the last simulation SpkLiq_step
// int *SpkLiq_spiked_e; //indices of the inhibitory neurons that spiked in the last simulation SpkLiq_step

int SpkLiq_number_of_neurons; //total number of neurons inside the liquid, automatically calculated based on the liquid's shape
double SpkLiq_current_time; //the current simulation time (in seconds)

int SpkLiq_number_of_exc_neurons; //total number of excitatory neurons in the liquid
int SpkLiq_number_of_inh_neurons; //total number of inhibitory neurons in the liquid

// int SpkLiq_total_number_of_connections; //stores the total number of connections (synapses) generated inside the liquid after initialization
int SpkLiq_inh_connections; //stores how many inhibitory=>? connections were created
int SpkLiq_exc_connections; //stores how many excitatory=>? connections were created


//time_seed and uniform_deviate functions came from:
//http://eternallyconfuzzled.com/arts/jsw_art_rand.aspx
unsigned SpkLiq_time_seed(){
     time_t now = time ( 0 );
     unsigned char *p = (unsigned char *)&now;
     unsigned seed = 0;
     size_t i;

    for ( i = 0; i < sizeof now; i++ )
        seed = seed * ( UCHAR_MAX + 2U ) + p[i];

    return seed;
}


//Tries to minimize the problem of a not very uniform distribution generated using rand()
double SpkLiq_uniform_deviate( int seed ){
    return seed * ( 1.0 / ( RAND_MAX + 1.0 ) );
} 
// int r = SpkLiq_uniform_deviate( rand() ) * N; // => (0,N]
// int r = M + SpkLiq_uniform_deviate( rand() ) * ( N - M ); // => (M,N]


void SpikingLiquid_Reset(){
        //
        // Resets the initial values of the simulation, but keeps the internal structure of the liquid
        // The affected variables are:
        // - unsigned int SpkLiq_rand_r_seeds[5]
        // - double SpkLiq_neurons_membrane[SpkLiq_number_of_neurons]
        // - double SpkLiq_neurons_exc_curr[SpkLiq_number_of_neurons]
        // - double SpkLiq_neurons_inh_curr[SpkLiq_number_of_neurons]
        // - double SpkLiq_refrac_timer[SpkLiq_number_of_neurons]
        // - int SpkLiq_test_vthres[SpkLiq_number_of_neurons]
        // - double SpkLiq_noisy_offset_currents[SpkLiq_number_of_neurons]
        // - double SpkLiq_noisy_currents[SpkLiq_number_of_neurons]

        printf("Reseting the simulation...\n");

        // Resets the current time to zero!
        SpkLiq_current_time = 0;

        // Resets the random state variables to their initial values
        for(int i=0; i<5; i++){
            SpkLiq_rand_r_seeds[i] = SpkLiq_user_seeds[i];
            // RANDOM-0: Membrane initial potentials
            // RANDOM-1: Noisy offset currents
            // RANDOM-2: Selection of the inhibitory and excitatory neurons
            // RANDOM-3: Internal connections of the liquid
            // RANDOM-4: Noisy corrents            
        }

        // Re-initializes the random generator of the gaussian distribution function
        // but now it will be used by the noisy currents during each simulation SpkLiq_step.
        // THIS ONLY WORKS BECAUSE NOBODY (BUT NOISY CURRENTS) IS GOING TO USE THIS FUNCTION AFTER THIS POINT!
        //
        // During the initialization the generating_connections change this, but there is his same instruction
        // before that point to make sure the noisy currents are going to use the correct seed.
        //
        // # RANDOM-4
        // # Noisy corrents
        // double const temp = SpkLiq_gaussrand(SpkLiq_rand_r_seeds[3]);
        SpkLiq_gaussrand(SpkLiq_rand_r_seeds[3]);
        
        for(int i=0; i<SpkLiq_number_of_neurons; i++){
            // # RANDOM-0
            // # Membrane initial potentials
            // # Initializes the membrane potentials for all neurons randomly according to (Maass, Natschlager, and Markram 2002)
            SpkLiq_neurons_membrane[i] = SpkLiq_membrane_rand[0] + SpkLiq_uniform_deviate(rand_r(&SpkLiq_rand_r_seeds[0])) * (SpkLiq_membrane_rand[1] - SpkLiq_membrane_rand[0]); //in mV 
            SpkLiq_neurons_membrane_init[i] = SpkLiq_neurons_membrane[i];

            SpkLiq_neurons_exc_curr[i] = 0.0; //Initializes the excitatory currents levels for all neurons
            
            SpkLiq_neurons_inh_curr[i] = 0.0; //Initializes the inhibitory currents levels for all neurons
            
            SpkLiq_refrac_timer[i] = 0.0; //Initializes the refractory timers
            
            SpkLiq_test_vthres[i]  = 0; // Initializes all positions with False

            // # RANDOM-1
            // # Noisy offset currents
            // # These are the offset noisy currents according to (Maass, Natschlager, and Markram 2002)
            SpkLiq_noisy_offset_currents[i] = SpkLiq_current_rand[0] + SpkLiq_uniform_deviate(rand_r(&SpkLiq_rand_r_seeds[1])) * (SpkLiq_current_rand[1] - SpkLiq_current_rand[0]); //in nA
        }

}
        
void SpikingLiquid_init(){

        printf("Generating the liquid internal structure...\n");

        // Calculates the total number of neurons based on the liquids shape
        SpkLiq_number_of_neurons = SpkLiq_net_shape[0]*SpkLiq_net_shape[1]*SpkLiq_net_shape[2];

        // Calculates the number of inhibitory and excitatory neurons
        SpkLiq_number_of_inh_neurons = (int) (SpkLiq_number_of_neurons*(SpkLiq_inhibitory_percentage/100.0));
        SpkLiq_number_of_exc_neurons = SpkLiq_number_of_neurons-SpkLiq_number_of_inh_neurons;

        // Here I need to allocate memory for all the dynamic variables acessed by SpikingLiquid_Reset.
        SpkLiq_neurons_membrane = malloc(sizeof(double)*SpkLiq_number_of_neurons);
        SpkLiq_neurons_membrane_init = malloc(sizeof(double)*SpkLiq_number_of_neurons);
        SpkLiq_neurons_exc_curr = malloc(sizeof(double)*SpkLiq_number_of_neurons);
        SpkLiq_neurons_inh_curr = malloc(sizeof(double)*SpkLiq_number_of_neurons);
        SpkLiq_refrac_timer = malloc(sizeof(double)*SpkLiq_number_of_neurons);
        SpkLiq_test_vthres = malloc(sizeof(int)*SpkLiq_number_of_neurons);
        SpkLiq_noisy_offset_currents = malloc(sizeof(double)*SpkLiq_number_of_neurons);
        SpkLiq_noisy_currents = malloc(sizeof(double)*SpkLiq_number_of_neurons);

        // And also the variables used in this function, but that are not going to be reset by SpikingLiquid_Reset
        SpkLiq_inhibitory_indices = malloc(sizeof(int)*SpkLiq_number_of_inh_neurons);
        SpkLiq_excitatory_indices = malloc(sizeof(int)*SpkLiq_number_of_exc_neurons);

        SpkLiq_liquid_indices_shuffled = malloc(sizeof(int)*SpkLiq_number_of_neurons);

        SpkLiq_refrac_values = malloc(sizeof(double)*SpkLiq_number_of_neurons);

        // Tests to verify if the malloc was successful with all the variables
        if (SpkLiq_neurons_membrane==NULL || SpkLiq_neurons_membrane_init==NULL || SpkLiq_neurons_exc_curr==NULL || SpkLiq_neurons_inh_curr==NULL || 
            SpkLiq_refrac_timer==NULL || SpkLiq_test_vthres==NULL ||
            SpkLiq_noisy_offset_currents==NULL || SpkLiq_noisy_currents==NULL || SpkLiq_inhibitory_indices==NULL || SpkLiq_excitatory_indices==NULL || 
            SpkLiq_liquid_indices_shuffled==NULL || SpkLiq_refrac_values==NULL){

            fprintf(stderr,"SpikingLiquid_init - Malloc ERROR!");
            exit(EXIT_FAILURE); //I need to check which error I should signal here...
        }

        printf("Calling SpikingLiquid_Reset from the SpikingLiquid_init...\n");
        // Initializes the internal variables to start the simulation
        SpikingLiquid_Reset();

        for(int i=0; i<SpkLiq_number_of_neurons; i++){
            SpkLiq_liquid_indices_shuffled[i] = i; //Fills with integers from 0 to (SpkLiq_number_of_neurons-1) to be shuffled
        }

        //# RANDOM-2
        //# Selection of the inhibitory and excitatory neurons
        // Knuth-Fisher-Yates shuffle algorithm - from http://blog.codinghorror.com/the-danger-of-naivete/
        // Creates a shuffled array with all the liquid's indices
        for(int i = SpkLiq_number_of_neurons - 1; i > 0; i--){
            // Generates a random integer from 0 to i+1
            int n = (int) (SpkLiq_uniform_deviate( rand_r(&SpkLiq_rand_r_seeds[2]) ) * (i+1));

            int temp;

            // Swaps the current position 'i' with the random position 'n'
            temp = SpkLiq_liquid_indices_shuffled[i], 
            SpkLiq_liquid_indices_shuffled[i] = SpkLiq_liquid_indices_shuffled[n];
            SpkLiq_liquid_indices_shuffled[n] = temp;
        }


        
        for(int i=0; i<SpkLiq_number_of_inh_neurons; i++){

            //Populates the SpkLiq_inhibitory_indices with the first N=SpkLiq_number_of_inh_neurons neurons from the shuffled liquid
            SpkLiq_inhibitory_indices[i] = SpkLiq_liquid_indices_shuffled[i];

            //Populates the refractory_vetor with inhibitory values
            SpkLiq_refrac_values[SpkLiq_inhibitory_indices[i]] = SpkLiq_refractory[1]; // Sets the refractory periods according to the type of neuron
        }


        for(int i=SpkLiq_number_of_inh_neurons; i<SpkLiq_number_of_neurons; i++){
            //Populates the SpkLiq_excitatory_indices with the neurons that are not inhibitory
            SpkLiq_excitatory_indices[(i-SpkLiq_number_of_inh_neurons)] = SpkLiq_liquid_indices_shuffled[i];

            //Populates the refractory_vetor with excitatory values
            SpkLiq_refrac_values[SpkLiq_excitatory_indices[(i-SpkLiq_number_of_inh_neurons)]] = SpkLiq_refractory[0]; // Sets the refractory periods according to the type of neuron

        }


        printf("Calling SpikingLiquid_generate_connections from SpikingLiquid_init...\n");
        // # RANDOM-3
        // # Internal connections of the liquid
        int total_connections = SpikingLiquid_generate_connections();

        // Re-initializes the random generator of the gaussian distribution function
        // but now it will be used by the noisy currents during each simulation SpkLiq_step.
        // THIS ONLY WORKS BECAUSE NOBODY (BUT NOISY CURRENTS) IS GOING TO USE THIS FUNCTION AFTER THIS POINT!
        // # RANDOM-4
        // # Noisy corrents
        SpkLiq_gaussrand(SpkLiq_rand_r_seeds[3]);


        printf("Total number of neurons: %d\n", SpkLiq_number_of_neurons);
        printf("Number of connections INH=>: %d\n", SpkLiq_inh_connections);
        printf("Number of connections EXC=>: %d\n", SpkLiq_exc_connections);
        printf("Total number of connections: %d\n", total_connections);
      
}

    
int SpikingLiquid_generate_connections(){

    // # RANDOM-3 - self.seeds[3]
    // # Internal connections of the liquid
    unsigned int *gen_seed = &SpkLiq_rand_r_seeds[3];
    
    int PRE; //flags to indicate the type of connection: inhibitory=>0 and excitatory=>1
    int POS;

    SpkLiq_gaussrand(*gen_seed); //initializes the random generator of the gaussian distribution function

    int n_malloc_i = 1; //Used with realloc to dynamically allocate more memory: inhibitory connections
    int n_malloc_e = 1; //Used with realloc to dynamically allocate more memory: excitatory connections

    int SpkLiq_total_number_of_connections = 0; //counts how many connections are created

//NEW
    SpkLiq_exc_connections = 0;
    SpkLiq_inh_connections = 0;

    //Allocate memory to save the inhibitory=>? connections
    SpkLiq_pre_i = calloc(MAX_CONNECTIONS,sizeof(int));
    SpkLiq_pos_i = calloc(MAX_CONNECTIONS,sizeof(int));
    SpkLiq_w_i = calloc(MAX_CONNECTIONS,sizeof(double));

    //Allocate memory to save the excitatory=>? connections
    SpkLiq_pre_e = calloc(MAX_CONNECTIONS,sizeof(int));
    SpkLiq_pos_e = calloc(MAX_CONNECTIONS,sizeof(int));
    SpkLiq_w_e = calloc(MAX_CONNECTIONS,sizeof(double));

    if (SpkLiq_pre_i==NULL || SpkLiq_pos_i==NULL || SpkLiq_w_i==NULL || SpkLiq_pre_e==NULL || SpkLiq_pos_e==NULL || SpkLiq_w_e==NULL){
        fprintf(stderr,"connections memory malloc ERROR!");
        exit(EXIT_FAILURE); //I need to check which error I should signal here...
    }

    // Goes through all the neurons indices and verify if they are inhibitory (excitatory test is implicit)
    // Instead of using the ordered list, I'm going to use the shuffled one because this way I don't need to check 
    // which neuron is excitatory or inhibitory in a difficult (cpu costly) way.
    for (int i=0;i<SpkLiq_number_of_neurons; i++){ //These are the PRE indices
        PRE = 1; //Because it's only tested against inhibitory indices

        // Verifies if the neuron index 'i' belongs to SpkLiq_inhibitory_indices
        if(i<SpkLiq_number_of_inh_neurons){
            PRE = 0; //Indicates the PRE-synaptic neuron is inhibitory
        }


        for (int j=0;j<SpkLiq_number_of_neurons; j++){ //These are the POS indices
            POS = 1;//Because it's only tested against inhibitory indices
            
            // Verifies if the neurons indices 'i' and 'j' belongs to SpkLiq_inhibitory_indices
            if(j<SpkLiq_number_of_inh_neurons){
                POS = 0; //Indicates the POS-synaptic neuron is inhibitory
            }
            
            //Here the variables about PRE/POS are set

            // Verifies if a connection (synapse) will be created or not depending on the calculated probability.
            if ( SpkLiq_uniform_deviate(rand_r(gen_seed)) <= (SpkLiq_liquid_parametres[PRE][POS][0]*exp(-pow(SpkLiq_euclidean_dist(SpkLiq_liquid_indices_shuffled[i],SpkLiq_liquid_indices_shuffled[j],SpkLiq_net_shape[0],SpkLiq_net_shape[1])/SpkLiq_lbd_value,2))) ){
                //It means we have a synapse here!
                SpkLiq_total_number_of_connections++;                

                if (PRE==0){//Inhibitory=>? connection
                    SpkLiq_inh_connections++;
                    
                    //Verifies if the buffer (used to store the Inhibitory=>? connections) needs more memory
                    if (SpkLiq_inh_connections>(MAX_CONNECTIONS*n_malloc_i)){
                        n_malloc_i++;
                        SpkLiq_pre_i = realloc(SpkLiq_pre_i,n_malloc_i*MAX_CONNECTIONS*sizeof(int));
                        SpkLiq_pos_i = realloc(SpkLiq_pos_i,n_malloc_i*MAX_CONNECTIONS*sizeof(int));
                        SpkLiq_w_i = realloc(SpkLiq_w_i,n_malloc_i*MAX_CONNECTIONS*sizeof(double));
                        if (SpkLiq_pre_i==NULL || SpkLiq_pos_i==NULL || SpkLiq_w_i==NULL){
                            fprintf(stderr,"SpikingLiquid_generate_connections INH - realloc ERROR!");
                            exit(EXIT_FAILURE);
                        }
                    }
                         
                    SpkLiq_pre_i[SpkLiq_inh_connections-1]=SpkLiq_liquid_indices_shuffled[i];
                    SpkLiq_pos_i[SpkLiq_inh_connections-1]=SpkLiq_liquid_indices_shuffled[j];
                    SpkLiq_w_i[SpkLiq_inh_connections-1]=1E-9*SpkLiq_liquid_parametres[PRE][POS][4]*fabs(SpkLiq_gaussrand(0)*0.5+1); //AMaass

                }else{//Excitatory=>? connections
                    SpkLiq_exc_connections++;   
                    
                    //Verifies if the buffer (used to store the Inhibitory=>? connections) needs more memory
                    if (SpkLiq_exc_connections>(MAX_CONNECTIONS*n_malloc_e)){
                        n_malloc_e++;
                        SpkLiq_pre_e = realloc(SpkLiq_pre_e,n_malloc_e*MAX_CONNECTIONS*sizeof(int));
                        SpkLiq_pos_e = realloc(SpkLiq_pos_e,n_malloc_e*MAX_CONNECTIONS*sizeof(int));
                        SpkLiq_w_e = realloc(SpkLiq_w_e,n_malloc_e*MAX_CONNECTIONS*sizeof(double));
                        if (SpkLiq_pre_e==NULL || SpkLiq_pos_e==NULL || SpkLiq_w_e==NULL){
                            fprintf(stderr,"SpikingLiquid_generate_connections EXC - realloc ERROR!");
                            exit(EXIT_FAILURE);
                        }
                    }
                
                    SpkLiq_pre_e[SpkLiq_exc_connections-1]=SpkLiq_liquid_indices_shuffled[i];
                    SpkLiq_pos_e[SpkLiq_exc_connections-1]=SpkLiq_liquid_indices_shuffled[j];
                    SpkLiq_w_e[SpkLiq_exc_connections-1]=1E-9*SpkLiq_liquid_parametres[PRE][POS][4]*fabs(SpkLiq_gaussrand(0)*0.5+1); //AMaass
                }
            }

        }
    }

    return SpkLiq_total_number_of_connections;
}

double SpkLiq_gaussrand(unsigned int my_seed){
    /*
    Gaussian generator algorithm from: 
            http://c-faq.com/lib/gaussian.html
            Abramowitz and Stegun, Handbook of Mathematical Functions
    Expect value: 0
    Standard deviation: 1
    */

    static unsigned int seed;
    static double V1, V2, S;
    static int phase = 0;
    double X;

    //Initializes the gaussian random generator
    if(my_seed!=0){
        seed = my_seed;
        phase = 0;
    }

    if(phase == 0) {
        do {
            double U1 = (double)rand_r(&seed) / RAND_MAX;
            double U2 = (double)rand_r(&seed) / RAND_MAX;

            V1 = 2 * U1 - 1;
            V2 = 2 * U2 - 1;
            S = V1 * V1 + V2 * V2;
            } while(S >= 1 || S == 0);

        X = V1 * sqrt(-2 * log(S) / S);
    } else
        X = V2 * sqrt(-2 * log(S) / S);

    phase = 1 - phase;

    return X;
}


double SpkLiq_euclidean_dist(int i, int j, int xd, int yd){
    // I'm assuming that in the 3D structure the values i and j are equivalent 
    // to the incrementing of x (until complete the line) then y and (after complete xy plane) z values.
    // xd, yd are the x and y dimensions of the liquid (shape)
    double xy = xd*yd;

    double zi = (int) (i/xy); //the integer cast is used here to round down the values. 
    double zj = (int) (j/xy); //the integer cast is used here to round down the values. 

    int yi = (int) (i-zi*xy)/xd; 
    int yj = (int) (j-zj*xy)/xd;

    double xi = (int) (i-yi*zi*xy); //the integer cast is used here to round down the values. 
    double xj = (int) (j-yj*zj*xy); //the integer cast is used here to round down the values. 

    // It's necessary to centralize the axis, otherwise we are going to have a bias
    xi = -xi/2.0;   // Centralize in the x=0 axis
    xj = -xj/2.0;   // Centralize in the x=0 axis 

    zi = -zi/2.0;   // Centralize in the z=0 axis
    zj = -zj/2.0;   // Centralize in the z=0 axis    

    return sqrt(pow((xi-xj),2)+pow((yi-yj),2)+pow((zi-zj),2));
}




void SpkLiq_process_exc_spikes(int *spikes, double *weights, int number_of_spikes){
    // '''
    // Processes the received spikes at the current time
    // spikes: list with the indexes of the neurons who spiked.
    // weights: the weights to the neurons who spiked
    // '''

    for(int i=0; i<number_of_spikes; i++){
        SpkLiq_neurons_exc_curr[spikes[i]] +=  weights[i]; // Adds to the ones that received spikes
        // printf("spikes_e:%d\n", spikes[i]);
    }
}


void SpkLiq_process_inh_spikes(int *spikes, double *weights, int number_of_spikes){
    // '''
    // Processes the received spikes at the current time
    // spikes: list with the indexes of the neurons who spiked.
    // weights: the weights to the neurons who spiked        
    // '''

    for(int i=0; i<number_of_spikes; i++){
        SpkLiq_neurons_inh_curr[spikes[i]] +=  weights[i]; // Adds to the ones that received spikes
        // printf("spikes_i:%d\n", spikes[i]);
    }
}



void SpikingLiquid_process_internal_spikes(){
    // '''
    // Tests if the liquid's neurons spiked
    // This function is called every simulation SpkLiq_step, so it's important to optimize for speed!
    // '''

    // Checks which internal INHIBITORY neurons spiked
    // This is accomplished checking against the SpkLiq_test_vthres array (length: SpkLiq_number_of_neurons)
    // SpkLiq_test_vthres indicates with 1 which neurons inside the liquid spiked in the last simulation SpkLiq_step
    int SpkLiq_spiked_i[SpkLiq_number_of_inh_neurons]; // Stores which inhibitory neuron spiked or not
                                         // It's NOT necessary initialize the array because the SpkLiq_spiked_i_idx
                                         // controls the maximum valid index.
    int SpkLiq_spiked_i_idx = 0; // Counts how many inhibitory neurons spiked
    for(int i=0; i<SpkLiq_number_of_inh_neurons; i++){
        //SpkLiq_test_vthres has the size of SpkLiq_number_of_neurons and has the value one
        //if the neuron spiked in the last SpkLiq_step
        if(SpkLiq_test_vthres[SpkLiq_inhibitory_indices[i]]){
            SpkLiq_spiked_i[SpkLiq_spiked_i_idx] = SpkLiq_inhibitory_indices[i];
            SpkLiq_spiked_i_idx++;
        }
    } // At the end of this loop, SpkLiq_spiked_i has which inhibitory neurons spiked, and the total is SpkLiq_spiked_i_idx
    
    // Now I need to check which neurons are going to receive these INHIBITORY spikes
    // Liquid's inhibitory neurons are connected to the neurons indicated in the array SpkLiq_pos_i
    // The arrays SpkLiq_pre_i, SpkLiq_pos_i and SpkLiq_w_i always have the same lenght: SpkLiq_inh_connections
    int receive_spike_i[SpkLiq_inh_connections];
    double receive_spike_i_w[SpkLiq_inh_connections];
    int receive_spike_i_idx = 0;
    for(int i=0; i<SpkLiq_inh_connections; i++) { //iterates through the SpkLiq_pre_i array
      for (int j=0;j<SpkLiq_spiked_i_idx;j++){ //iterates through the SpkLiq_spiked_i array (can be smaller than SpkLiq_number_of_inh_neurons!)
        if (SpkLiq_pre_i[i]==SpkLiq_spiked_i[j]){ // Checks for INHIBITORY neurons (that are connected like INH=>) who spiked
            receive_spike_i[receive_spike_i_idx] = SpkLiq_pos_i[i];
            receive_spike_i_w[receive_spike_i_idx] = SpkLiq_w_i[i];
            // printf("receive_i:%d - weight:%f\n",receive_spike_i[receive_spike_i_idx],1E9*receive_spike_i_w[receive_spike_i_idx]);            
            receive_spike_i_idx++;
        }
      }
    } // At the end of this loop, receive_spike_i has which neurons should receive a spike from INHIBITORY neurons, 
      // and the total is receive_spike_i_idx.
      // ATTENTION: one neuron can receive more than one spike at each simulation SpkLiq_step.


    // Checks which internal EXCITATORY neurons spiked
    // This is accomplished checking against the SpkLiq_test_vthres array (length: SpkLiq_number_of_neurons)
    // SpkLiq_test_vthres indicates with 1 which neurons inside the liquid spiked in the last simulation SpkLiq_step
    int SpkLiq_spiked_e[SpkLiq_number_of_exc_neurons]; // Stores which excitatory neuron spiked or not    
                                         // It's NOT necessary initialize the array because the SpkLiq_spiked_e_idx
                                         // controls the maximum valid index.    
    int SpkLiq_spiked_e_idx = 0; // Counts how many excitatory neurons spiked
    for(int i=0; i<SpkLiq_number_of_inh_neurons; i++){
        //SpkLiq_test_vthres has the size of SpkLiq_number_of_neurons and has the value one
        //if the neuron spiked in the last SpkLiq_step
        if(SpkLiq_test_vthres[SpkLiq_excitatory_indices[i]]){
            SpkLiq_spiked_e[SpkLiq_spiked_e_idx] = SpkLiq_excitatory_indices[i];
            SpkLiq_spiked_e_idx++;
        }
    } // At the end of this loop, SpkLiq_spiked_i has which excitatory neurons spiked, and the total is SpkLiq_spiked_e_idx
    
    // Now I need to check which neurons are going to receive these EXCITATORY spikes
    // Liquid's excitatory neurons are connected to the neurons indicated in the array SpkLiq_pos_e
    // The arrays SpkLiq_pre_e, SpkLiq_pos_e and SpkLiq_w_e always have the same lenght: SpkLiq_exc_connections
    int receive_spike_e[SpkLiq_exc_connections];
    double receive_spike_e_w[SpkLiq_exc_connections];
    int receive_spike_e_idx = 0;
    for(int i=0; i<SpkLiq_exc_connections; i++) { //iterates through the SpkLiq_pre_e array
      for (int j=0;j<SpkLiq_spiked_e_idx;j++){ //iterates through the SpkLiq_spiked_e array (can be smaller than SpkLiq_number_of_exc_neurons!)
        if (SpkLiq_pre_e[i]==SpkLiq_spiked_e[j]){ // Checks for EXCITATORY neurons (that are connected like INH=>) who spiked
            receive_spike_e[receive_spike_e_idx] = SpkLiq_pos_e[i];
            receive_spike_e_w[receive_spike_e_idx] = SpkLiq_w_e[i];
            // printf("receive_e:%d - weight:%f\n",receive_spike_e[receive_spike_e_idx],1E9*receive_spike_e_w[receive_spike_e_idx]);
            receive_spike_e_idx++;
        }
      }
    } // At the end of this loop, receive_spike_e has which neurons should receive a spike from EXCITATORY neurons, 
      // and the total is receive_spike_e_idx.
      // ATTENTION: one neuron can receive more than one spike at each simulation SpkLiq_step.

    // Now, finally, the internal spikes are going to be processed!
    // Maybe if I group this with the external spikes processing I could avoid doing two big for loops
    SpkLiq_process_inh_spikes(receive_spike_i, receive_spike_i_w, receive_spike_i_idx);
    SpkLiq_process_exc_spikes(receive_spike_e, receive_spike_e_w, receive_spike_e_idx);
}



void SpikingLiquid_update(int *spikes_exc, int *spikes_inh, double *weights_exc, double *weights_inh, int size_exc, int size_inh){
    // '''
    // Updates the simulation, running one time SpkLiq_step, and processes the received excitatory and inhibitory spikes.
    // SpikingLiquid_update(spikes_exc, spikes_inh, weights_exc, weights_inh):
    // spikes_exc: list/array with the index of the neurons receiving excitatory spikes
    // spikes_inh: list/array with the index of the neurons receiving inhibitory spikes
    // weights_exc: list/array with the weights in the same order as spikes_exc
    // weights_inh: list/array with the weights in the same order as spikes_inh        
    
    // The first update is time=0s
    // '''

    //
    // I could use the Trapezoidal integration rule for the membrane and currents diff eq solver
    // instead of the simple one I'm using:
    // http://en.wikipedia.org/wiki/Trapezoidal_rule
    // But I don't know if it worths the extra cpu cycles...
    //


    // These updates must occur BEFORE the spikes, otherwise the spikes are going to be reduced just like when one time SpkLiq_step had passed
    for(int i=0; i<SpkLiq_number_of_neurons; i++){
        SpkLiq_neurons_inh_curr[i] += (-SpkLiq_neurons_inh_curr[i]/SpkLiq_taui)*SpkLiq_step; // Updates the INHIBITORY current values
                                                                                             // integrating the function dii/dt = -ii/taui
        SpkLiq_neurons_exc_curr[i] += (-SpkLiq_neurons_exc_curr[i]/SpkLiq_taue)*SpkLiq_step; // Updates the EXCITATORY current values
                                                                                             // integrating the function die/dt = -ie/taue
    }

    // I need to think more about WHEN the spikes should be processed:
    // - BEFORE the neurons updates
    // or
    // - AFTER the neurons updates
    //
    // Seems that if they are processed AFTER, makes easier to incorporate the 
    // exponential current integration into the main for loop and this could speed up things or
    // make it easier to do the CUDA-fu in the near future...

    // Processes the internal spikes (inhibitory and excitatory)
    SpikingLiquid_process_internal_spikes();

    // Processes the external spikes (inhibitory and excitatory)
    if(size_inh!=0)
        SpkLiq_process_inh_spikes(spikes_inh, weights_inh, size_inh);

    if(size_exc!=0)
        SpkLiq_process_exc_spikes(spikes_exc, weights_exc, size_exc);


    for(int i=0; i<SpkLiq_number_of_neurons; i++){

        // Generates the new values to the noisy currents
        // # RANDOM-4
        // # Noisy corrents
//>>>COMPARISON CHANGE
        SpkLiq_noisy_currents[i] = 0;//SpkLiq_gaussrand(0)*SpkLiq_noisy_current_rand; // Variable noisy currents


        // # After the currents, it is necessary to update the membrane voltage
        // # Integrating the function:
        // dv/dt  = (ie + ii + i_offset + i_inj)/c_m + (v_rest-v)/tau_m
        if(SpkLiq_refrac_timer[i]>0){
            SpkLiq_neurons_membrane[i] = SpkLiq_vresets; // Resets the membrane of the refractory ones     
        }else{ //If the membrane did not reset, then it is necessary to integrate
            SpkLiq_neurons_membrane[i] += ((SpkLiq_neurons_exc_curr[i] + SpkLiq_neurons_inh_curr[i] + SpkLiq_noisy_offset_currents[i] + SpkLiq_noisy_currents[i])/SpkLiq_cm + (SpkLiq_vrest-SpkLiq_neurons_membrane[i])/SpkLiq_taum)*SpkLiq_step;
        }


// dv(v,t,ie,ii,i_offset,i_noise,vrest):
// (ie + ii + i_offset + i_noise)/self.cm + (vrest-v)/self.taum
// self.neurons_membrane+=self.dv(neurons_membrane,time,neurons_exc_curr,neurons_inh_curr,noisy_offset_currents,noisy_currents,vrest)*step

        // After the integration, check if the neuron spiked by comparing its membrane voltage with the threshold
        if(SpkLiq_neurons_membrane[i]>SpkLiq_vthres){
            SpkLiq_test_vthres[i] = 1; //# Verifies who should spike and creates a boolean vector with this information
            SpkLiq_neurons_membrane[i] = SpkLiq_vresets; //# Resets the membrane of the spiked ones
            SpkLiq_refrac_timer[i] = SpkLiq_refrac_values[i]; //# Resets the refractory timer to the ones who spiked            
        }else{
//
//NEW
//            
            SpkLiq_test_vthres[i] = 0; //# Verifies who should spike and creates a boolean vector with this information
        }

// NOTE: When should this occur?????
        SpkLiq_refrac_timer[i] -= SpkLiq_step; //# Subtracts the refractory timer to enable the refractory calculation

//
//NEW - I don't know if this is necessary, but if the simulation goes on during a long period of time this could overflow.
//        
        // if (SpkLiq_refrac_timer[i]<0)
        //     SpkLiq_refrac_timer[i]=0;
    }

    //# If I update the time here makes it possible to generate things at time zero!
    SpkLiq_current_time += SpkLiq_step; //# Advances the simulation time one SpkLiq_step
}


// ##########################################################################################
// ##########################################################################################
// ##########################################################################################
// ##########################################################################################
// ##########################################################################################

int user_setup(
                int *_SpkLiq_net_shape,
                double _SpkLiq_lbd_value,
                double _SpkLiq_step,
                double _SpkLiq_taum,
                double _SpkLiq_cm,
                double _SpkLiq_taue,
                double _SpkLiq_taui,
                double *_SpkLiq_membrane_rand,
                double *_SpkLiq_current_rand,
                double _SpkLiq_noisy_current_rand,
                double _SpkLiq_vresets,
                double _SpkLiq_vthres,
                double _SpkLiq_vrest,
                double *_SpkLiq_refractory,
                double _SpkLiq_inhibitory_percentage,
                unsigned int *my_seeds,
                int _SpkLiq_threads_N
                ){
    //
    // USER PARAMETRES
    //

    // Network user parametres
    SpkLiq_net_shape[0] = _SpkLiq_net_shape[0];//15;
    SpkLiq_net_shape[1] = _SpkLiq_net_shape[1];//30;
    SpkLiq_net_shape[2] = _SpkLiq_net_shape[2];//3;

    SpkLiq_lbd_value = _SpkLiq_lbd_value;//1.2;
    SpkLiq_step = _SpkLiq_step;//0.2E-3;
    SpkLiq_taum = _SpkLiq_taum;//30E-3;
    SpkLiq_cm = _SpkLiq_cm;//30E-9;
    SpkLiq_taue = _SpkLiq_taue;//3E-3;
    SpkLiq_taui = _SpkLiq_taui;//6E-3;

    SpkLiq_membrane_rand[0] = _SpkLiq_membrane_rand[0];//13.5E-3;
    SpkLiq_membrane_rand[1] = _SpkLiq_membrane_rand[1];//15E-3; //range of the possible values to the initialization of the membrane voltage
    

    SpkLiq_current_rand[0] = _SpkLiq_current_rand[0];//14.975E-9;
    SpkLiq_current_rand[1] = _SpkLiq_current_rand[1];//15.025E-9; //range of the possible values to the initialization of the constant noisy offset current

    SpkLiq_noisy_current_rand = _SpkLiq_noisy_current_rand;//0.2E-9; //max range of the possible values to the initialization of the variable noisy current    

    SpkLiq_vresets = _SpkLiq_vresets;//13.5E-3;
    SpkLiq_vthres = _SpkLiq_vthres;//15E-3;
    SpkLiq_vrest = _SpkLiq_vrest;//0.0;

    SpkLiq_refractory[0] = _SpkLiq_refractory[0];//3E-3;
    SpkLiq_refractory[1] = _SpkLiq_refractory[1];//2E-3;
    SpkLiq_inhibitory_percentage = _SpkLiq_inhibitory_percentage;//20; //it means the liquid is going to have approximately 20% of inhibitory neurons


    // RANDOM-0: Membrane initial potentials
    // RANDOM-1: Noisy offset currents
    // RANDOM-2: Selection of the inhibitory and excitatory neurons
    // RANDOM-3: Internal connections of the liquid
    // RANDOM-4: Noisy corrents
    // User random seeds
    for(int i=0;i<5;i++)
        SpkLiq_user_seeds[i] = my_seeds[i];

    // SpkLiq_threads_N = _SpkLiq_threads_N;

    return EXIT_SUCCESS;
}


// Exposes some of the variables
void reads_neurons(int *output){
    memcpy(output,SpkLiq_test_vthres,sizeof(int)*SpkLiq_number_of_neurons); // saves the spikes output from the liquid
}

void reads_membranes(double *output){
    memcpy(output,SpkLiq_neurons_membrane,sizeof(double)*SpkLiq_number_of_neurons); // saves the membranes voltages
}

void reads_membranes_init(double *output){
    memcpy(output,SpkLiq_neurons_membrane_init,sizeof(double)*SpkLiq_number_of_neurons); // saves the initial membranes voltages
}

void reads_exc_synapses(double *output){
    memcpy(output,SpkLiq_neurons_exc_curr,sizeof(double)*SpkLiq_number_of_neurons); // saves the excitatory currents
}

void reads_inh_synapses(double *output){
    memcpy(output,SpkLiq_neurons_inh_curr,sizeof(double)*SpkLiq_number_of_neurons); // saves the inhibitory currents
}

void reads_pre_i_connections(int *output){
    memcpy(output,SpkLiq_pre_i,sizeof(int)*SpkLiq_inh_connections); 
}

void reads_pos_i_connections(int *output){
    memcpy(output,SpkLiq_pos_i,sizeof(int)*SpkLiq_inh_connections); 
}

void reads_pre_i_weights(double *output){
    memcpy(output,SpkLiq_w_i,sizeof(double)*SpkLiq_inh_connections);
}

void reads_pre_e_connections(int *output){
    memcpy(output,SpkLiq_pre_e,sizeof(int)*SpkLiq_exc_connections); 
}

void reads_pos_e_connections(int *output){
    memcpy(output,SpkLiq_pos_e,sizeof(int)*SpkLiq_exc_connections);
}

void reads_pre_e_weights(double *output){
    memcpy(output,SpkLiq_w_e,sizeof(double)*SpkLiq_exc_connections); // saves the inhibitory currents
}

void reads_inhibitory_indices(int *output){
    memcpy(output,SpkLiq_inhibitory_indices,sizeof(int)*SpkLiq_number_of_inh_neurons); // saves the inhibitory currents
}

void reads_excitatory_indices(int *output){
    memcpy(output,SpkLiq_excitatory_indices,sizeof(int)*SpkLiq_number_of_exc_neurons); // saves the inhibitory currents
}

void reads_refrac_values(double *output){
    memcpy(output,SpkLiq_refrac_values,sizeof(double)*SpkLiq_number_of_neurons); // saves the inhibitory currents
}

void reads_noisy_offset_currents(double *output){
    memcpy(output,SpkLiq_noisy_offset_currents,sizeof(double)*SpkLiq_number_of_neurons); // saves the inhibitory currents
}

void reads_noisy_currents(double *output){
    memcpy(output,SpkLiq_noisy_currents,sizeof(double)*SpkLiq_number_of_neurons); // saves the inhibitory currents
}


void stats_liquid(int *output){
    output[0]=SpkLiq_number_of_inh_neurons;
    output[1]=SpkLiq_number_of_exc_neurons;
    output[2]=SpkLiq_inh_connections;
    output[3]=SpkLiq_exc_connections;
}


void free_all(){

    // Pointers are used here because the memory size varies according to the user values
    free(SpkLiq_neurons_membrane); //the membrane potentials for all neurons
    free(SpkLiq_neurons_exc_curr); //excitatory currents levels for all neurons
    free(SpkLiq_neurons_inh_curr); //inhibitory currents levels for all neurons
    free(SpkLiq_refrac_timer); //timers used to control the refractory period
    free(SpkLiq_test_vthres); //controls which neuron is above the voltage threshold and should spike
    free(SpkLiq_noisy_offset_currents); //values of the neuron model constant noisy offset currents
    free(SpkLiq_noisy_currents); //values of the neuron model variable noisy currents
    free(SpkLiq_refrac_values); //stores the refractory periods for all neurons in the liquid 
                                         //because the refractory period is different for inhibitory and excitatory neurons

    free(SpkLiq_inhibitory_indices); //Indices of the inhibitory neurons inside the liquid
    free(SpkLiq_excitatory_indices); //Indices of the excitatory neurons inside the liquid
    free(SpkLiq_liquid_indices_shuffled); //used to generate the random SpkLiq_inhibitory_indices/SpkLiq_excitatory_indices


    free(SpkLiq_pre_i); //stores the indices of the inhibitory=>? pre-synaptic connections in the liquid
    free(SpkLiq_pos_i); //stores the indices of the inhibitory=>? pos-synaptic connections in the liquid
    free(SpkLiq_w_i); //stores the weights of the connections above

    free(SpkLiq_pre_e); //stores the indices of the inhibitory=>? pre-synaptic connections in the liquid
    free(SpkLiq_pos_e); //stores the indices of the inhibitory=>? pos-synaptic connections in the liquid
    free(SpkLiq_w_e); //stores the weights of the connections above

}

// Here I'm testing the simulator. In the future I want to call this simulator from Python passing all the parametres
// and also the data from where the input spikes are going to be read.
int main(){

    double time_begin, time_end;
    
    int number_of_iterations = 500;

    printf("Initializing the input variables...\n");

    // Network user parametres
    SpkLiq_net_shape[0] = 15;
    SpkLiq_net_shape[1] = 3;
    SpkLiq_net_shape[2] = 3;

    SpkLiq_lbd_value = 1.2;
    SpkLiq_step = 0.2E-3;
    SpkLiq_taum = 30E-3;
    SpkLiq_cm = 30E-9;
    SpkLiq_taue = 3E-3;
    SpkLiq_taui = 6E-3;

    SpkLiq_membrane_rand[0] = 13.5E-3;
    SpkLiq_membrane_rand[1] = 15E-3; //range of the possible values to the initialization of the membrane voltage

    SpkLiq_current_rand[0] = 14.975E-9;
    SpkLiq_current_rand[1] = 15.025E-9; //range of the possible values to the initialization of the constant noisy offset current

    SpkLiq_noisy_current_rand = 0.2E-9; //max range of the possible values to the initialization of the variable noisy current    

    SpkLiq_vresets = 13.5E-3;
    SpkLiq_vthres = 15E-3;
    SpkLiq_vrest = 0.0;

    SpkLiq_refractory[0] = 3E-3;
    SpkLiq_refractory[1] = 2E-3;

    SpkLiq_inhibitory_percentage = 20; //it means the liquid is going to have approximately 20% of inhibitory neurons

    // RANDOM-0: Membrane initial potentials
    // RANDOM-1: Noisy offset currents
    // RANDOM-2: Selection of the inhibitory and excitatory neurons
    // RANDOM-3: Internal connections of the liquid
    // RANDOM-4: Noisy corrents


    // User random seeds
    unsigned int my_seeds[5] = {1234,1234,1234,1234,1234};
    for(int i=0;i<5;i++)
        SpkLiq_user_seeds[i] = my_seeds[i];



    for(int k=0; k<1; k++){
        printf("Calling SpikingLiquid_init...\n");

        time_begin=clock_my_gettime();

        SpikingLiquid_init();
        
        time_end=clock_my_gettime();
        printf("Total init time (ms):%f\n", (time_end-time_begin)/1E6);    


        // Creates the input spikes
        int input_spikes_exc[50];
        double input_spikes_exc_w[50];
        for(int i=0; i<50; i++){
          input_spikes_exc[i]=i;
          input_spikes_exc_w[i]=1000E-9;
        }


        // Array to save the output spikes
        // int *output_spikes_mem = malloc(number_of_iterations*SpkLiq_number_of_neurons*sizeof(int));
        int **output_spikes_mem = malloc(number_of_iterations*sizeof(int*));

        for (int i=0; i<number_of_iterations; i++){
            output_spikes_mem[i] = malloc(SpkLiq_number_of_neurons*sizeof(int));
            
            if(output_spikes_mem[i]==NULL){
                printf("Error!!!\n");
                exit(1);
            }

        }

        printf("Allocated memory output(%dpositions):%luKbytes\n",number_of_iterations*SpkLiq_number_of_neurons ,(unsigned long int) ((number_of_iterations*SpkLiq_number_of_neurons*sizeof(int))/1000.0));

        // Gets the initial time in nanoseconds
        time_begin=clock_my_gettime();

        for(int i=0; i<50; i++){
            SpikingLiquid_update(NULL, NULL, NULL, NULL, 0, 0);
            reads_neurons(output_spikes_mem[i]);
            // reads_neurons((output_spikes_mem+i*SpkLiq_number_of_neurons*sizeof(int)));
            // memcpy((output_spikes_mem+i*sizeof(int*)),SpkLiq_test_vthres,sizeof(int)*SpkLiq_number_of_neurons); // saves the spikes output from the liquid
        } // i=0 to 49

        //Sends 50 spikes (0 to 49) @ i=50
        SpikingLiquid_update(input_spikes_exc, NULL, input_spikes_exc_w, NULL, 50, 0);

        reads_neurons(output_spikes_mem[50]);
        // reads_neurons((output_spikes_mem+50*SpkLiq_number_of_neurons*sizeof(int)));
        // memcpy((output_spikes_mem+50*sizeof(int*)),SpkLiq_test_vthres,sizeof(int)*SpkLiq_number_of_neurons); // saves the spikes output from the liquid


        for(int i=51; i<500; i++){
            SpikingLiquid_update(NULL, NULL, NULL, NULL, 0, 0);
            reads_neurons(output_spikes_mem[i]);
            // reads_neurons((output_spikes_mem+i*SpkLiq_number_of_neurons*sizeof(int)));
            // memcpy((output_spikes_mem+i*sizeof(int*)),SpkLiq_test_vthres,sizeof(int)*SpkLiq_number_of_neurons); // saves the spikes output from the liquid
        } // i=51 to 499

        // Gets the final time in nanoseconds
        time_end=clock_my_gettime();
        printf("Total update time (ms):%f\n", (time_end-time_begin)/1E6);
        printf("Average update time (ms):%f\n", ((time_end-time_begin)/1E6)/number_of_iterations);

        int c=0;
        for(int i=0; i<number_of_iterations; i++){
          printf("(%4.1f):",1E3*i*SpkLiq_step);
          for(int j=0;j<SpkLiq_number_of_neurons;j++){
              printf("%d",output_spikes_mem[i][j]);
              if(output_spikes_mem[i][j]){
                c++;
              }
              // printf("%d",*(output_spikes_mem+(i+j)*sizeof(int)));
              // if(*(output_spikes_mem+i*j*sizeof(int))){
              //   c++;
              // }
          }
          printf("\n");
        }

        printf("Total Spikes:%d\n",c);

        //free_all();

    }
    return 0;


}
```

```python
# Reads the output spikes
def output_sim(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.int32)
    liquid_output(output)
    return numpy.array(output>0)

def output_sim2(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.int32)
    liquid_output(output)
    return numpy.array(output)

# Reads the membrane voltages
def output_voltages(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.double)
    liquid_voltages(output)
    return numpy.array(output)

# Reads the initial membrane voltages
def output_initial_voltages(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.double)
    liquid_initial_voltages(output)
    return numpy.array(output)

# Reads the excitatory currents
def output_exc_currents(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.double)
    liquid_exc_currents(output)
    return numpy.array(output)

# Reads the inhibitory currents
def output_inh_currents(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.double)
    liquid_inh_currents(output)
    return numpy.array(output)

# Reads the pre_i connections
def output_pre_i_connections(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.int32)
    liquid_pre_i(output)
    return numpy.array(output)

# Reads the pos_e connections
def output_pos_i_connections(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.int32)
    liquid_pos_i(output)
    return numpy.array(output)

# Reads the pre_i weights
def output_pre_i_weights(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.double)
    liquid_w_i(output)
    return numpy.array(output)

# Reads the pre_e connections
def output_pre_e_connections(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.int32)
    liquid_pre_e(output)
    return numpy.array(output)

# Reads the pos_e connections
def output_pos_e_connections(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.int32)
    liquid_pos_e(output)
    return numpy.array(output)


# Reads the pre_e weights
def output_pre_e_weights(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.double)
    liquid_w_e(output)
    return numpy.array(output)

# Reads the excitatory indices
def output_exc_indices(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.int32)
    liquid_exc_indices(output)
    return numpy.array(output)

# Reads the inhibitory indices
def output_inh_indices(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.int32)
    liquid_inh_indices(output)
    return numpy.array(output)

# Reads the noisy currents
def output_noisy_currents(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.double)
    liquid_noisy_currents(output)
    return numpy.array(output)

# Reads the noisy offsets
def output_noisy_offsets(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.double)
    liquid_noisy_offset(output)
    return numpy.array(output)

# Reads the noisy offsets
def output_refrac_values(number_of_neurons):
    output = numpy.empty(number_of_neurons,dtype=numpy.double)
    liquid_refrac_values(output)
    return numpy.array(output)



# Liquid's stats
def output_stats():
    output = numpy.empty(4,dtype=numpy.int32)
    liquid_stats(output)
    print "Total number of neurons:", NofN    
    print "Number of inhibitory neurons:",output[0]
    print "Number of excitatory neurons:",output[1]
    print "Number of inhibitory connections:",output[2]
    print "Number of excitatory connections:",output[3]
    return output
```

```python

voltages_brian = []

import brian

try:
    brian.reinit() # Necessary when using a python console
except:
    print "Brian REINIT error..."

import numpy # I could use "brian.", because Brian imports numpy, but I prefer not.
import time


Net_shape=(15,3,3)
Number_of_neurons_lsm=135

my_STEP = 0.2

AINP = 100 # Input gains in nA

# These make easier to use the Brian objects without the "brian." at the beginning
ms = brian.ms
mV = brian.mV
nA = brian.nA
nF = brian.nF
NeuronGroup = brian.NeuronGroup
SpikeGeneratorGroup = brian.SpikeGeneratorGroup
Synapses = brian.Synapses
SpikeMonitor = brian.SpikeMonitor
network_operation = brian.network_operation 
defaultclock = brian.defaultclock

defaultclock.dt = my_STEP*ms



initial_time = time.time()

print "Initial time (in seconds):",initial_time


#
# Number of Inhibitory neurons - LIQUID - 20% of the total neurons
lsm_indices = numpy.arange(Number_of_neurons_lsm) # Generates a list of indices

#### INPUT FROM MY SIMULATOR!!!
inhibitory_index_L = output_inh_indices(NoINHN)
excitatory_index = output_exc_indices(NoEXCN)


# Generate the connections matrix inside the Liquid (Liquid->Liquid) - according to Maass2002
#
print "Liquid->Liquid connections..."


#
# These are the cell (neuron) parameters according to Maass 2002
#
cell_params_lsm = {'cm'          : 30*nF,    # Capacitance of the membrane 
                                             # =>>>> MAASS PAPER DOESN'T MENTION THIS PARAMETER DIRECTLY
                                             #       but the paper mentions a INPUT RESISTANCE OF 1MEGA Ohms and tau_m=RC=30ms, so cm=30nF
                   'i_offset'    : 0.0*nA,   # Offset current - random for each neuron from [14.975nA to 15.025nA] => Masss2002 - see code below
                   'tau_m'       : 30.0*ms,  # Membrane time constant => Maass2002
                   'tau_refrac_E': 3.0*ms,   # Duration of refractory period - 3mS for EXCITATORY => Maass2002
                   'tau_refrac_I': 2.0*ms,   # Duration of refractory period - 2mS for INHIBITORY => Maass2002
                   'tau_syn_E'   : 3.0*ms,   # Decay time of excitatory synaptic current => Maass2002
                   'tau_syn_I'   : 6.0*ms,   # Decay time of inhibitory synaptic current => Maass2002
                   'v_reset'     : 13.5*mV,  # Reset potential after a spike => Maass2002
                   'v_rest'      : 0.0*mV,   # Resting membrane potential => Maass2002
                   'v_thresh'    : 15.0*mV,  # Spike threshold => Maass2002
                   'i_noise'     : 0.2*nA    # Used in Maass2002: mean 0 and SD=0.2nA
                }

# IF_curr_exp - MODEL EXPLAINED
# Leaky integrate and fire model with fixed threshold and
# decaying-exponential post-synaptic current. 
# (Separate synaptic currents for excitatory and inhibitory synapses)
lsm_neuron_eqs='''
  dv/dt  = (ie + ii + i_offset + i_noise)/c_m + (v_rest-v)/tau_m : mV
  die/dt = -ie/tau_syn_E                : nA
  dii/dt = -ii/tau_syn_I                : nA
  tau_syn_E                             : ms
  tau_syn_I                             : ms
  tau_m                                 : ms
  c_m                                   : nF
  v_rest                                : mV
  i_offset                              : nA
  i_noise                               : nA
  '''
# lsm_neuron_eqs='''
#   dv/dt  = (ie + ii + i_offset + i_noise)/c_m + (v_rest-v)/tau_m : mV
#   ie                                    : nA
#   ii                                    : nA
#   tau_m                                 : ms
#   c_m                                   : nF
#   v_rest                                : mV
#   i_offset                              : nA
#   i_noise                               : nA
#   '''



########################################################################################################################
#
# LIQUID - Setup
#
print "LIQUID - Setup..."

#### INPUT FROM MY SIMULATOR!!!
refractory_vector=output_refrac_values(NofN)


# This is the population (neurons) used exclusively to the Liquid (pop_lsm).
# All the neurons receive the same threshold and reset voltages.
pop_lsm = NeuronGroup(Number_of_neurons_lsm, model=lsm_neuron_eqs, 
                                             threshold=cell_params_lsm['v_thresh'], 
                                             reset=cell_params_lsm['v_reset'], 
                                             refractory=refractory_vector, 
                                             max_refractory=max(cell_params_lsm['tau_refrac_E'], 
                                                                cell_params_lsm['tau_refrac_I']))


# Here I'm mixing numpy.fill with the access of the state variable "c_m" in Brian (because Brian is using a numpy.array)
# Sets the value of the capacitance according to the cell_params_lsm (same value to all the neurons)
pop_lsm.c_m.fill(cell_params_lsm['cm'])


# Sets the value of the time constant RC (or membrane constant) according to the cell_params_lsm (same value to all the neurons)
pop_lsm.tau_m.fill(cell_params_lsm['tau_m'])

# Sets the i_offset according to Maass2002
# The i_offset current is random, but never changes during the simulation.
# this current should be drawn from a uniform distr [14.975,15.025]
# Joshi2005 does [13.5,14.5] ???? Maybe is to avoid spikes without inputs...
#
#### INPUT FROM MY SIMULATOR!!!
pop_lsm.i_offset=output_noisy_offsets(NofN)*1E9*nA #1E9 is used to convert from A to nA



pop_lsm.tau_syn_E.fill(cell_params_lsm['tau_syn_E']) # (same value to all the neurons)
pop_lsm.tau_syn_I.fill(cell_params_lsm['tau_syn_I']) # (same value to all the neurons)


# All neurons receive the same value to the resting voltage.
pop_lsm.v_rest.fill(cell_params_lsm['v_rest']) # (same value to all the neurons)


# Sets the initial membrane voltage according to Maass2002. Doesn't change during the simulation.
# this current should be drawn from a uniform distr [13.5mV,15.0mV]
#
#### INPUT FROM MY SIMULATOR!!!
pop_lsm.v=output_initial_voltages(NofN)*1E3*mV #1E3 is used to convert from V to mV


#
# Loading or creating the Synapses objects used within the Liquid
print "Liquid->Liquid connections..."

####THE WEIGHTS MUST BE IN nA, not in A!!!!!!
def create_synapses(population, synapse_type, indices_pre, indices_pos, weights):
    #
    # Creates the connections (Synapse object) among the neurons in the liquid
    #
    # synapse_type = 'exc'
    # EXCITATORY (PRE) TO ANYTHING (POST) 
    # synapse_type = 'inh'
    # INHIBITORY (PRE) TO ANYTHING (POST)
    if synapse_type=='exc':
            model_eq='''w : 1'''
            pre_eq='''ie+=w'''

    elif synapse_type=='inh':
            model_eq='''w : 1'''
            pre_eq='''ii+=w'''

    syn_lsm=brian.Synapses(population, population, model=model_eq, pre=pre_eq)
    # Caution about creation of synapses:
    # 1) there is no deletion
    # 2) synapses are added, not replaced (e.g. S[1,2]=True;S[1,2]=True creates 2 synapses)
    for ipre,ipos in zip(indices_pre,indices_pos):        
        syn_lsm[ipre,ipos] = True # here is where the synapse is really created

####THE WEIGHTS MUST BE IN nA, not in A!!!!!!
    syn_lsm.w=numpy.array(weights)*nA

    syn_lsm.delay=0*ms
    
    return syn_lsm


#### INPUT FROM MY SIMULATOR!!!
# Generates the Liquid->Liquid - EXCITATORY synapses
syn_lsm_exc = create_synapses(pop_lsm,'exc', output_pre_e_connections(NoEXCC), output_pos_e_connections(NoEXCC), output_pre_e_weights(NoEXCC)*1E9)
# the multiplication by *1E9 changes  the weights from A to nA

#### INPUT FROM MY SIMULATOR!!!
# Generates the Liquid->Liquid - INHIBITORY synapses
syn_lsm_inh = create_synapses(pop_lsm,'inh', output_pre_i_connections(NoINHC), output_pos_i_connections(NoINHC), output_pre_i_weights(NoINHC)*1E9)
# the multiplication by *1E9 changes  the weights from A to nA

print "Liquid->Liquid connections...Done!"

total_number_of_connections_liquid = len(syn_lsm_exc) + len(syn_lsm_inh)

print "Number of inhibitory synapses in the Liquid: " + str(len(syn_lsm_inh)) # DEBUG to verify if it is working
print "Number of excitatory synapses in the Liquid: " + str(len(syn_lsm_exc)) # DEBUG to verify if it is working


# To understand what is being returned:
# pop_lsm: it is necessary to connect the neuron network with the rest of the world
# [syn_lsm_obj, syn_lsm_exc, syn_lsm_inh]: to include these objects at the simulation (net=Net(...); net.run(total_sim_time*ms)); 
# It is a list because is easy to concatenate lists :D

print "LIQUID - Setup...Done!"

#
# End of the LIQUID - Setup
########################################################################################################################


########################################################################################################################
#
# INPUT - Setup
#
print "INPUT - Setup..."

tspk = defaultclock.dt*50 # The neurons spike after 50 time steps!
number_of_spikes = 50

spiketimes = [(i,tspk) for i in range(number_of_spikes)] 
                # The spikes are going to be received during the simulation, 
                # so this is always an empty list when using the step_by_step_brian_sim!


# I'm using only one big input layer because Brian docs say it is better for the performance
SpikeInputs = SpikeGeneratorGroup(number_of_spikes, spiketimes)


#
#
# Here the synapses are created. The synapses created are ALWAYS excitatory because it is 
# connecting through 'ie' in the neuron model!

syn_world_Input = Synapses(SpikeInputs, pop_lsm,
                                     model='''w : 1''',
                                     pre='''ie+=w''')

for i in range(len(spiketimes)):
    syn_world_Input[i,i] = True

weights_input_liquid = numpy.array([AINP]*number_of_spikes)*nA # Creates a numpy.array with number_of_spikes itens

syn_world_Input.w = weights_input_liquid
syn_world_Input.delay=0*ms

print "INPUT - Setup...Done!"

#
# End of the INPUT - Setup (creation of the connections between the Poisson input and the Liquid!)
#
########################################################################################################################


# DON'T FORGET TO INSERT THE FUNCTION "generate_i_noise" INTO THE MONITORS_OBJECT LIST!!!!
# Generates the noisy current at each time step (as seen in Joshi2005)
@network_operation(clock=defaultclock)
def generate_i_noise():
    global voltages_brian
    voltages_brian.append(pop_lsm.v.mean()) #Saves the mean value of the membrane voltages
    # These are the noise currents inside each liquid's neuron
    pop_lsm.i_noise=0 #rndst_noisy.normal(loc=0, scale=cell_params_lsm['i_noise'],size=Number_of_neurons_lsm)*nA
    # Here I'm using a RandomState that was initialized with the SAME seed as my simulator noisy generation!!!

pop_objects = [pop_lsm, SpikeInputs]

syn_objects = [syn_lsm_exc, syn_lsm_inh, syn_world_Input]

monitors_objects = [generate_i_noise]

OutputMonitor=brian.SpikeMonitor(pop_lsm, record=True)
VMonitor=brian.StateMonitor(pop_lsm,'v', record=True)
IEMonitor=brian.StateMonitor(pop_lsm,'ie', record=True)

net = brian.Network(pop_objects + syn_objects + monitors_objects + [OutputMonitor,VMonitor,IEMonitor])

print "Setup time:", time.time()-initial_time
initial_time = time.time()

net.run(500*my_STEP*ms)
print "Simulation time:", time.time()-initial_time
```
