This is a VERY VERY simple Spiking Neural Network simulator that creates a Liquid State Machine

It uses the ideas from:
Maass, Wolfgang, Thomas Natschlager, and Henry Markram. 2002. 
“Real-Time Computing without Stable States: A New Framework for Neural Computation Based on Perturbations.” 
Neural Computation 14 (11): 2531–60. doi:10.1162/089976602760407955.

However, the dynamic synapses (STP) and the time delays are NOT implemented (yet) in an attempt of making the system faster
and also because I'm not sure if they are really necessary.

Probably this code has a lot of bugs because I've stopped developing it without properly testing it. Right now I'm working in a complete C version of a LSM simulator.

```python
%%file multiple_find.c
/* File multiple_find.c */
void multiple_find(int n, int m, int *pre, int *spiked, int *output, int *l) {
   //n=>length of pre
   //m=>length of spiked
  int i;
  int j;
  int k=0;

  for (i=0;i<n;i++) {
      for (j=0;j<m;j++){
        if (pre[i]==spiked[j]){
            output[k] = i;
            k++;
        }
      }
  }
  *l=k;
}
```

```python
%%writefile liquid_structure.c

#include <stdlib.h>
#include <stdio.h>
#include <math.h>


/*
Compile as a shared library:
gcc -shared -Wl,-install_name,liquid_structure.so -o liquid_structure.so -fPIC liquid_structure.c -O2
*/

// This is used to define when realloc is going to ask for more memory
#define MAX_CONNECTIONS 10

// These are the addresses we need to free sometime in the future :)
// They are global variables to make easier to access from two different functions
int *f_ptr_pre_i = NULL; 
int *f_ptr_pos_i = NULL; 
int *f_ptr_pre_e = NULL;
int *f_ptr_pos_e = NULL;
double *f_ptr_w_i = NULL;
double *f_ptr_w_e = NULL;
    
double gaussrand(unsigned int my_seed){
    /*
    Gaussian generator algorithm from: 
            http://c-faq.com/lib/gaussian.html
            Marsaglia and Bray, "A Convenient Method for Generating Normal Variables"
            http://en.wikipedia.org/wiki/Marsaglia_polar_method
            Abramowitz and Stegun, Handbook of Mathematical Functions
    Expect value: 0
    Standard deviation: 1
    */

    static unsigned int seed;
    static double V1, V2, S;
    static int phase = 0;
    double X;

    //Initializes the gaussian random generator
    if(my_seed!=0){
        seed = my_seed;
        phase = 0;
    }

    if(phase == 0) {
        do {
            double U1 = (double)rand_r(&seed) / RAND_MAX;
            double U2 = (double)rand_r(&seed) / RAND_MAX;

            V1 = 2 * U1 - 1;
            V2 = 2 * U2 - 1;
            S = V1 * V1 + V2 * V2;
            } while(S >= 1 || S == 0);

        X = V1 * sqrt(-2 * log(S) / S);
    } else
        X = V2 * sqrt(-2 * log(S) / S);

    phase = 1 - phase;

    return X;
}


double euclidean_dist(int i, int j, int xd, int yd){
    // I'm assuming that in the 3D structure the values i and j are equivalent 
    // to the incrementing of x (until complete the line) then y and (after complete xy plane) z values.
    // xd, yd are the x and y dimensions of the liquid (shape)
    double xy = xd*yd;

    double zi = (int) (i/xy); 
    double zj = (int) (j/xy); 

    int yi = (int) (i-zi*xy)/xd; 
    int yj = (int) (j-zj*xy)/xd;

    double xi = (int) (i-yi*zi*xy); 
    double xj = (int) (j-yj*zj*xy); 

    // It's necessary to centralize the axis, otherwise we are going to have a bias
    xi = -xi/2.0;   // Centralize in the x=0 axis
    xj = -xj/2.0;   // Centralize in the x=0 axis 

    zi = -zi/2.0;   // Centralize in the z=0 axis
    zj = -zj/2.0;   // Centralize in the z=0 axis    

    return sqrt(pow((xi-xj),2)+pow((yi-yj),2)+pow((zi-zj),2));
}


//Frees the memory space allocated to the liquid structure
int free_variables(void){
    static int memory_used = 1;

    if(memory_used!=0){
        //Deallocate the memory
        free(f_ptr_pre_i); 
        free(f_ptr_pos_i); 
        free(f_ptr_pre_e);
        free(f_ptr_pos_e);
        free(f_ptr_w_i);
        free(f_ptr_w_e);
        //Signalize the memory was freed
        memory_used = 0;
        return 0;
    }else{
        return 1;
    }

}

int generate_connections(int x, 
                         int y, 
                         int z, 
                         double lbd, 
                         int *inh_indices, int inh_size, 
                         double parametres[2][2][6], 
                         int *inh_connections,
                         int **ptr_pre_i, int **ptr_pos_i, double **ptr_w_i, 
                         int *exc_connections,                         
                         int **ptr_pre_e, int **ptr_pos_e, double **ptr_w_e, 
                         unsigned int my_seed){

    // inh_indices MUST be a list made of UNIQUE indices (integers)!

    unsigned int c_seed = my_seed;
    int number_of_neurons = x*y*z;
    int total_number_of_connections = 0; //counts how many connecitons are created
    
    int inh_c = 0; //counts how many inhibitory=>? connections
    int exc_c = 0; //counts how many excitatory=>? connections

    int PRE; //flags to indicate the type of connection: inhibitory=>0 and excitatory=>1
    int POS;

    double temp = gaussrand(my_seed); //initializes the random generator

    int n_malloc_i = 1; //Used with realloc to dynamically allocate more memory: inhibitory connections
    int n_malloc_e = 1; //Used with realloc to dynamically allocate more memory: excitatory connections

    int *pre_i = NULL;
    int *pos_i = NULL;
    double *w_i = NULL;

    int *pre_e = NULL;
    int *pos_e = NULL;
    double *w_e = NULL;

    //Allocate memory to save the inhibitory=>? connections
    pre_i = calloc(MAX_CONNECTIONS,sizeof(int));
    pos_i = calloc(MAX_CONNECTIONS,sizeof(int));
    w_i = calloc(MAX_CONNECTIONS,sizeof(double));

    //Allocate memory to save the excitatory=>? connections
    pre_e = calloc(MAX_CONNECTIONS,sizeof(int));
    pos_e = calloc(MAX_CONNECTIONS,sizeof(int));
    w_e = calloc(MAX_CONNECTIONS,sizeof(double));

    if (pre_i==NULL || pos_i==NULL || w_i==NULL || pre_e==NULL || pos_e==NULL || w_e==NULL){
        fprintf(stderr,"connections memory malloc ERROR!");
        exit(1); //I need to check which error I should signal here...
    }


    // Goes through all the neurons indices and verify if they are inhbitory (excitatory test is implicit)
    for (int i=0;i<number_of_neurons; i++){ //These are the PRE indices
        PRE = 1; //Because it's only tested against inhbitory indices

        // Verifies if the neuron index 'i' belongs to inh_indices
        for (int k=0; k<inh_size; k++){
            if(inh_indices[k]==i){// Tests if the connection is I?
                PRE = 0; //Indicates the PRE-synaptic neuron is inhbitory
                break; //Saves CPU cycles...
            }
        }


        for (int j=0;j<number_of_neurons; j++){ //These are the POS indices
            POS = 1;//Because it's only tested against inhbitory indices
            
            // Verifies if the neurons indices 'i' and 'j' belongs to inh_indices
            for (int k=0; k<inh_size; k++){
                if(inh_indices[k]==j){// Tests if the connection is I?
                    POS = 0; //Indicates the POS-synaptic neuron is inhbitory
                    break; //Saves CPU cycles...
                }
            }
            
            //Here the variables about PRE/POS are set

            // Verifies if a connection (synapse) will be created or not depending on the calculated probability.
            if ( ((double)rand_r(&c_seed)/(double)RAND_MAX) <= (parametres[PRE][POS][0]*exp(-pow(euclidean_dist(i,j,x,y)/lbd,2))) ){
                //It means we have a synapse here!
                total_number_of_connections++;                

                if (PRE==0){//Inhibitory=>? connection
                    inh_c++;
                    
                    //Verifies if the buffer (used to store the Inhibitory=>? connections) needs more memory
                    if (inh_c>(MAX_CONNECTIONS*n_malloc_i)){
                        n_malloc_i++;
                        pre_i = realloc(pre_i,n_malloc_i*MAX_CONNECTIONS*sizeof(int));
                        pos_i = realloc(pos_i,n_malloc_i*MAX_CONNECTIONS*sizeof(int));
                        w_i = realloc(w_i,n_malloc_i*MAX_CONNECTIONS*sizeof(double));
                        if (pre_i==NULL || pos_i==NULL || w_i==NULL){
                            fprintf(stderr,"connections memory realloc ERROR!");
                            exit(1);
                        }
                    }
                         
                    pre_i[inh_c-1]=i;
                    pos_i[inh_c-1]=j;
                    w_i[inh_c-1]=1E-9*parametres[PRE][POS][4]*fabs(gaussrand(0)*0.5+1); //AMaass

                }else{//Excitatory=>? connections
                    exc_c++;   
                    
                    //Verifies if the buffer (used to store the Inhibitory=>? connections) needs more memory
                    if (exc_c>(MAX_CONNECTIONS*n_malloc_e)){
                        n_malloc_e++;
                        pre_e = realloc(pre_e,n_malloc_e*MAX_CONNECTIONS*sizeof(int));
                        pos_e = realloc(pos_e,n_malloc_e*MAX_CONNECTIONS*sizeof(int));
                        w_e = realloc(w_e,n_malloc_e*MAX_CONNECTIONS*sizeof(double));
                        if (pre_e==NULL || pos_e==NULL || w_e==NULL){
                            fprintf(stderr,"connections memory realloc ERROR!");
                            exit(1);
                        }
                    }
                
                    pre_e[exc_c-1]=i;
                    pos_e[exc_c-1]=j;
                    w_e[exc_c-1]=1E-9*parametres[PRE][POS][4]*fabs(gaussrand(0)*0.5+1); //AMaass
                }
            }

        }
    }


    *ptr_pre_i = pre_i;
    *ptr_pos_i = pos_i;
    *ptr_w_i = w_i;
    *inh_connections = inh_c;

    *ptr_pre_e = pre_e;
    *ptr_pos_e = pos_e;
    *ptr_w_e = w_e;
    *exc_connections = exc_c;

    //Saves the memory addresses to be freed later
    f_ptr_pre_i = pre_i; 
    f_ptr_pos_i = pos_i; 
    f_ptr_pre_e = pre_e;
    f_ptr_pos_e = pos_e;
    f_ptr_w_i = w_i;
    f_ptr_w_e = w_e;

    return total_number_of_connections;
}
```

```python
%%file LSM_simulator.py

'''
USING EXPONENTIAL INPUT CURRENTS
'''
import numpy

import sys




import snnsim_c # C compiled multiple search algorithm inside numpy arrays
reload(sys.modules['snnsim_c']) 

# from scipy.integrate import odeint # I'm not using because I'm doing the numerical integration by hand, but it is there if I need to test

# C library used for the liquid structure generation
import ctypes
liquid_structure=ctypes.CDLL("./liquid_structure.so")


class SpikingLiquid(object):
    
    def reset(self):
        '''
        Resets the initial values of the simulation, but keeps the internal structure of the liquid
        '''
        # Generates individual RandomState objects for each stochastic thing in the code
        self.rndste=[numpy.random.RandomState(self.seeds[i] & 0xFFFFFFFF) if self.seeds[i]!=None else numpy.random.RandomState() for i in range(5)]
        # RANDOM-0: Membrane initial potentials
        # RANDOM-1: Noisy offset currents
        # RANDOM-2: Selection of the inhibitory and excitatory neurons
        # RANDOM-3: Internal connections of the liquid
        # RANDOM-4: Noisy corrents
        #
        
        
        # RANDOM-0
        # Membrane initial potentials
        
        # Initializes the membrane potentials for all neurons randomly according to (Maass, Natschlager, and Markram 2002)
        self.neurons_membrane = self.rndste[0].uniform(13.5E-3,15E-3, self.Number_of_Neurons) #in mV 

        self.neurons_exc_curr = numpy.zeros(self.Number_of_Neurons) # Initializes the excitatory currents levels for all neurons
        self.neurons_inh_curr = numpy.zeros(self.Number_of_Neurons) # Initializes the inhibitory currents levels for all neurons
        
        self.time = 0 # Starts the simulation at time=0
        self.times_exc_curr = numpy.zeros(self.Number_of_Neurons) # Initializes the times for the excitatory currents
        
        self.refrac_timer = numpy.zeros(self.Number_of_Neurons) # Initializes the refractory timers
        
        self.test_vrest  = numpy.zeros(self.Number_of_Neurons)>0 # Initializes all positions with False
        self.test_vthres = numpy.zeros(self.Number_of_Neurons)>0 # Initializes all positions with False
        self.test_refrac = numpy.zeros(self.Number_of_Neurons)>0 # Initializes all positions with False
        

        # RANDOM-1
        # Noisy offset currents
        
        # These are the offset noisy currents according to (Maass, Natschlager, and Markram 2002)
        self.noisy_offset_currents = self.rndste[1].uniform(14.975E-9,15.025E-9, self.Number_of_Neurons) # in nA
    
    
    def __init__(self, Number_of_Neurons=135, Net_shape=(15,3,3), lbd_value=1.2, step=0.2E-3, taum=30E-3, cm=30E-9, vrest=(0.0)*numpy.ones(135), vthres=(15E-3)*numpy.ones(135), vresets=numpy.ones(135)*13.5E-3, taue=3E-3, taui=6E-3, refractory=(3E-3,2E-3), seeds=(None,None,None,None,None)):
        '''
        Creates a liquid state machine.
        
        Number_of_Neurons: total number of neurons to be simulated
        Net_shape: a tuple with the liquid's 3D shape (x,y,z)
        step: simulation time step (in seconds)
        
        lbd_value: coefficient of the liquid's connections (see Maass, Natschlager, and Markram 2002)
        
        taum: membrane time constant (in seconds)
        cm: membrane capacitance (in farads)
        
        vrest: array with the resting membrane potential (in volts)
        vthres: array with the membrane spiking threshold (in volts)
        vresets: array with the reset voltages of all neurons (in volts)        
        
        taue: current decay time constant excitatory (in seconds)
        taui: current decay time constant inhibitory (in seconds)
        
        refractory: tuple where the elements tell the refractory for excitatory and inhibitory neurons (in seconds)
        
        seeds: tuple with five values to be used as random seed to the RandomState objects. If None is passed to one
        of the tuple values that RandomState object will keep as initialized with .RandomState().
            seeds[0]: Membrane initial potentials
            seeds[1]: Noisy offset currents
            seeds[2]: Selection of the inhibitory and excitatory neurons
            seeds[3]: Internal connections of the liquid
            seeds[4]: Noisy corrents
        
        To run the simulation call:
        .update(spikes_exc, spikes_inh, weights_exc, weights_inh):
        spikes_exc: list/array with the index of the neurons receiving excitatory spikes
        spikes_inh: list/array with the index of the neurons receiving inhibitory spikes
        weights_exc: list/array with the weights in the same order as spikes_exc
        weights_inh: list/array with the weights in the same order as spikes_inh
        
        To verify the membrane voltages call:
        .verify_membrane()

        To verify which neuron spiked call:        
        .verify_spikes()
        '''
        print "Initializing the liquid..."
        print "Number of neurons in the liquid: ", Number_of_Neurons
        print "Liquid's shape: ",Net_shape
        print "Simulation time step: ",step
        
        self.seeds = seeds
                
        self.Number_of_Neurons = Number_of_Neurons
        
        self.step = step
        
        self.cm = cm
        self.vrest = numpy.array(vrest) # Guarantees it is a Numpy.Array
        self.vthres = numpy.array(vthres)
        self.vresets = numpy.array(vresets)
                
        self.taum = taum
        self.taue = taue
        self.taui = taui

        self.reset() # Initializes the internal variables to start the simulation
        
        # RANDOM-2
        # Selection of the inhibitory and excitatory neurons

        # Selects some random neurons to be inhibitory/excitatory ones
        lsm_3dGrid_flat = numpy.arange(self.Number_of_Neurons) # Creates a dummy array to be used with the suffle        
        self.rndste[2].shuffle(lsm_3dGrid_flat) # Using the shuffle method I have guarantee there are NO repetitions
        
        self.inhbitory_index = numpy.array(lsm_3dGrid_flat[:int(self.Number_of_Neurons*0.2)], dtype=numpy.int32) # The first (approx.) 20% as inhibitory
        self.excitatory_index = numpy.array(lsm_3dGrid_flat[int(self.Number_of_Neurons*0.2):])# The rest of the neurons as excitatory

        self.refractory_vector = numpy.zeros(self.Number_of_Neurons)
        self.refractory_vector[self.excitatory_index] = refractory[0] # Sets the refractory periods according to the type of neuron
        self.refractory_vector[self.inhbitory_index] = refractory[1]  # using the information from the user input tuple.
        
        # This is the dictionary that has all the connections parameters according to (Maass, Natschlager, and Markram 2002).
        # It is necessary to create the 3D connections and the STP configuration matrixes (STP is not implemented )
        # E=>1 (excitatory) and I=>0 (inhibitory)
        # Ex.: (0,0) => II
        # Dynamical Synapses Parameters (STP/Liquid Structure):
        Connections_Parameters = numpy.array([
             [
                numpy.array(
                                [ # II [0,0]
                                      0.1,       # CGupta=0.1        # Parameter used at the connection probability - from Maass2002 paper
                                      0.32,      # UMarkram=0.32     # Use (U) - Parameter used at the Dynamic Synapse - from Maass2002 paper
                                      0.144,     # DMarkram=0.144    # Time constant for Depression (tau_rec) - used at the Dynamic Synapse - from Maass2002 paper                    
                                      0.06,      # FMarkram=0.06     # Time constant for Facilitation (tau_facil) - used at the Dynamic Synapse - from Maass2002 paper
                                      -2.8,      # AMaass=-2.8        # (nA) In the Maass2002 paper
                                      0.8        # Delay_trans = 0.8 # In Maass paper the transmission delay is 0.8 to II, IE and EI        
                                ],dtype=numpy.double
                            ),
                numpy.array(
                                [ # IE [0,1]
                                      0.4,    # CGupta=0.4
                                      0.25,   # UMarkram=0.25
                                      0.7,    # DMarkram=0.7
                                      0.02,   # FMarkram=0.02
                                      -3.0,   # AMaass=-3.0
                                      0.8     # Delay_trans = 0.8 # in Maass paper the transmission delay is 0.8 to II, IE and EI
                                  ],dtype=numpy.double
                            )
            ],
            [
                numpy.array(
                                [ # EI [1,0]
                                      0.2,    # CGupta=0.2
                                      0.05,   # UMarkram=0.05
                                      0.125,  # DMarkram=0.125
                                      1.2,    # FMarkram=1.2
                                      1.6,    # AMaass=1.6
                                      0.8     # Delay_trans = 0.8 # in Maass paper the transmission delay is 0.8 to II, IE and EI
                                  ],dtype=numpy.double
                            ),
                numpy.array(
                                [ # EE [1,1]
                                      0.3,    # CGupta=0.3 
                                      0.5,    # UMarkram=0.5
                                      1.1,    # DMarkram=1.1
                                      0.05,   # FMarkram=0.05
                                      1.2,    # AMaass=1.2 #scaling parameter or absolute synaptic efficacy or weight - from Maass2002 paper
                                      1.5     # Delay_trans = 1.5 # in Maass paper the transmission delay is 1.5 to EE connection
                                  ],dtype=numpy.double
                            )
            ]
        ])

        # RANDOM-3 - self.seeds[3]
        # Internal connections of the liquid
        ptr2int=ctypes.POINTER(ctypes.c_int32)
        ptr2double=ctypes.POINTER(ctypes.c_double)

        gen_conn=liquid_structure.generate_connections
        gen_conn.restype=ctypes.c_int32
        gen_conn.argtypes = [
                        ctypes.c_int32, 
                        ctypes.c_int32, 
                        ctypes.c_int32, 
                        ctypes.c_double,
                        numpy.ctypeslib.ndpointer(dtype=numpy.int32, flags='ALIGNED,C_CONTIGUOUS'), ctypes.c_int32,
                        numpy.ctypeslib.ndpointer(dtype=numpy.double, flags='ALIGNED,C_CONTIGUOUS'),
                        ctypes.POINTER(ctypes.c_int32),
                        ctypes.POINTER(ptr2int), ctypes.POINTER(ptr2int), ctypes.POINTER(ptr2double),
                        ctypes.POINTER(ctypes.c_int32),
                        ctypes.POINTER(ptr2int), ctypes.POINTER(ptr2int), ctypes.POINTER(ptr2double),
                        ctypes.c_uint32
                       ]
        # int generate_connections(int x, 
        #                          int y, 
        #                          int z, 
        #                          double lbd, 
        #                          long int *inh_indices, int inh_size, 
        #                          double parametres[2][2][6], 
        #                          int *inh_connections,
        #                          int **ptr_pre_i, int **ptr_pos_i, double **ptr_w_i, 
        #                          int *exc_connections,                         
        #                          int **ptr_pre_e, int **ptr_pos_e, double **ptr_w_e, 
        #                          unsigned int my_seed)

        number_of_inh_neurons = ctypes.c_int(len(self.inhbitory_index))

        inh_connections = ctypes.c_int32(0)
        p_pre_i = ptr2int()
        p_pos_i = ptr2int()
        p_w_i = ptr2double()

        exc_connections = ctypes.c_int32(0)
        p_pre_e = ptr2int()
        p_pos_e = ptr2int()
        p_w_e = ptr2double()

        seed = ctypes.c_uint32(123422)

        total_number_of_connections = 0;

        # int generate_connections(int x, 
        #                          int y, 
        #                          int z, 
        #                          double lbd, 
        #                          long int *inh_indices, int inh_size, 
        #                          double parametres[2][2][6], 
        #                          int *inh_connections,
        #                          int **ptr_pre_i, int **ptr_pos_i, double **ptr_w_i, 
        #                          int *exc_connections,                         
        #                          int **ptr_pre_e, int **ptr_pos_e, double **ptr_w_e, 
        #                          unsigned int my_seed)
        total_number_of_connections = gen_conn(ctypes.c_int32(Net_shape[0]),
                                               ctypes.c_int32(Net_shape[1]),
                                               ctypes.c_int32(Net_shape[2]),
                                               ctypes.c_double(lbd_value),
                                               self.inhbitory_index, number_of_inh_neurons,
                                               Connections_Parameters,
                                               ctypes.byref(inh_connections),
                                               ctypes.byref(p_pre_i), ctypes.byref(p_pos_i), ctypes.byref(p_w_i),
                                               ctypes.byref(exc_connections),
                                               ctypes.byref(p_pre_e), ctypes.byref(p_pos_e), ctypes.byref(p_w_e),
                                               ctypes.c_uint(self.seeds[3])
                                              )
        
        # Processes the information about the liquid=>liquid connections generated with the lm.generate_connections
        # The lists comprehensions below are processing the neurons that HAVE a connection. Because it is a probabilist function
        # some neurons will have more than one connection, but some can have none!
        # These connections will be used to verify and inject spikes at each time step during the simulation.
        self.pre_i = numpy.ctypeslib.as_array((ctypes.c_int32 * inh_connections.value).from_address(ctypes.addressof(p_pre_i.contents)))
        self.pos_i = numpy.ctypeslib.as_array((ctypes.c_int32 * inh_connections.value).from_address(ctypes.addressof(p_pos_i.contents)))
        self.w_i = numpy.ctypeslib.as_array((ctypes.c_double * inh_connections.value).from_address(ctypes.addressof(p_w_i.contents)))
        
#         self.pre_i = numpy.array([ind[0][0] for ind in self.output_L_L['inh']],dtype=int) # Extracts the indices of the PRE neurons (inhibitory)
#         self.pos_i = numpy.array([ind[0][1] for ind in self.output_L_L['inh']],dtype=int) # Extracts the indices of the POS neurons (inhibitory)
#         self.w_i   = numpy.array([ind[2][0]*1E-9 for ind in self.output_L_L['inh']],dtype=float) # Extracts the weights connecting PRE to POS (inhibitory)

        self.pre_e = numpy.ctypeslib.as_array((ctypes.c_int32 * exc_connections.value).from_address(ctypes.addressof(p_pre_e.contents)))
        self.pos_e = numpy.ctypeslib.as_array((ctypes.c_int32 * exc_connections.value).from_address(ctypes.addressof(p_pos_e.contents)))
        self.w_e = numpy.ctypeslib.as_array((ctypes.c_double * exc_connections.value).from_address(ctypes.addressof(p_w_e.contents)))

#         self.pre_e = numpy.array([ind[0][0] for ind in self.output_L_L['exc']],dtype=int) # Extracts the indices of the PRE neurons (excitatory)
#         self.pos_e = numpy.array([ind[0][1] for ind in self.output_L_L['exc']],dtype=int) # Extracts the indices of the POS neurons (excitatory)
#         self.w_e   = numpy.array([ind[2][0]*1E-9 for ind in self.output_L_L['exc']],dtype=float) # Extracts the weights connecting PRE to POS (excitatory)

        # pre_i, pos_i and w_i have the same length
        # pre_e, pos_e and w_e have the same length
        
        print "Number of connections INH=>: ", str(len(self.w_i))
        print "Number of connections EXC=>: ", str(len(self.w_e))
        
        print "Initializing the liquid...Done!"
    
    
    # Based on 2.4 Synapses - http://icwww.epfl.ch/~gerstner/SPNM/node16.html, 
    # but using the same model to inhibitory and excitatory (with differents time constants)
    def process_exc_spikes(self, spikes, weights):
        '''
        Processes the received spikes at the current time updating the current values.
        spikes: list with the indexes of the neurons who spiked.
        weights: the weights to the neurons who spiked
        '''
        self.neurons_exc_curr += (-self.neurons_exc_curr/self.taue)*self.step # Calculates the current values
                                                                              # FOR ALL EXCITATORY NEURONS!
        self.neurons_exc_curr[spikes] +=  weights*numpy.ones(len(spikes)) # Adds to the ones that received spikes
        
    def process_inh_spikes(self, spikes, weights):
        '''
        Processes the received spikes at the current time updating the current values.
        spikes: list with the indexes of the neurons who spiked.
        weights: the weights to the neurons who spiked        
        '''
        self.neurons_inh_curr += (-self.neurons_inh_curr/self.taui)*self.step # Calculates the current values
                                                                              # FOR ALL INHIBITORY NEURONS!        
        self.neurons_inh_curr[spikes] += weights*numpy.ones(len(spikes)) # Adds to the ones that received spikes

    def dv(self,v,t,ie,ii,i_offset,i_noise,vrest):
        '''
        Differential equation describing the behaviour of the neuron model
        dv(t)/dt = (ie(t) + ii(t) + i_offset + i_noise(t))/self.cm + (vrest-v(t))/self.taum 
        '''
        return (ie + ii + i_offset + i_noise)/self.cm + (vrest-v)/self.taum

    def check_internal_spikes(self):
        '''
        Tests if the liquid's neurons spiked
        '''
        # Checks which internal inhibitory neurons spiked
        spiked_i = self.inhbitory_index[self.test_vthres[self.inhbitory_index]] # Gives the index of the inh spiked ones
        # The next line (Pure Python Version), probably, is the bottleneck of this method
#         who_receives_spikes_i_idx = [idx for idx in xrange(len(self.pre_i)) if self.pre_i[idx] in spiked_i] # Verifies if the spiked one bellongs to the connected ones
        # Replace by the C compiled version (below):
        who_receives_spikes_i_idx,l = snnsim_c.multiple_find(self.pre_i,spiked_i)
        
        self.who_receives_spikes_i = self.pos_i[who_receives_spikes_i_idx[:l]] # Finds to who it is connected
        self.who_receives_spikes_i_w = self.w_i[who_receives_spikes_i_idx[:l]] # Finds the connection weights

        # Checks which internal excitatory neurons spiked
        spiked_e = self.excitatory_index[self.test_vthres[self.excitatory_index]] # Gives the index of the exc spiked ones
        # The next line (Pure Python Version), probably, is the bottleneck of this method
#         who_receives_spikes_e_idx = [idx for idx in xrange(len(self.pre_e)) if self.pre_e[idx] in spiked_e]
        # Replace by the C compiled version (below):
        who_receives_spikes_e_idx,l = snnsim_c.multiple_find(self.pre_e,spiked_e)

        self.who_receives_spikes_e = self.pos_e[who_receives_spikes_e_idx[:l]]
        self.who_receives_spikes_e_w = self.w_e[who_receives_spikes_e_idx[:l]]
        
        
    def verify_membrane(self):
        '''
        Returns a tuple: time,neurons_membrane
        time: current simulation time (in seconds)
        neurons_membrane: numpy.array with voltage values of the neurons' membrane (in volts)
        '''
        return self.time,self.neurons_membrane

    def verify_spikes(self):
        '''
        Returns a tuple: time, test_vthres 
        time: current simulation time (in seconds)
        test_vthres: numpy.array of booleans indicating which neuron spiked
        Ex:
        indices = numpy.arange(total_number_of_neurons)
        indices[test_vthres] => gives the indices of the neurons who spiked
        '''
        return self.time,self.test_vthres
        
    def update(self, spikes_exc, spikes_inh, weights_exc, weights_inh):
        '''
        Updates the simulation, running one time step, and processes the received excitatory and inhibitory spikes.
        .update(spikes_exc, spikes_inh, weights_exc, weights_inh):
        spikes_exc: list/array with the index of the neurons receiving excitatory spikes
        spikes_inh: list/array with the index of the neurons receiving inhibitory spikes
        weights_exc: list/array with the weights in the same order as spikes_exc
        weights_inh: list/array with the weights in the same order as spikes_inh        
        
        The first update is time=0s
        '''
        self.check_internal_spikes() # runs the C compiled function to check which neuron is going to spike and returns the index of the POS neuron connected to it.
        
        # Here I concatenate the received spikes from outside with the internally generated spikes
        # and update the excitatory and inhibitory currents.
        # I used .astype(int) because numpy generates floats if one list is empty and was cheaper than testing for an empty list
        self.process_inh_spikes(numpy.concatenate((spikes_inh,self.who_receives_spikes_i)).astype(int), numpy.concatenate((weights_inh,self.who_receives_spikes_i_w))) # Processes the inhibitories arriving spikes
        self.process_exc_spikes(numpy.concatenate((spikes_exc,self.who_receives_spikes_e)).astype(int), numpy.concatenate((weights_exc,self.who_receives_spikes_e_w))) # Processes the excitatories arriving spikes

        # RANDOM-4
        # Noisy corrents
        self.noisy_currents = self.rndste[4].normal(loc=0, scale=0.2E-9,size=self.Number_of_Neurons) # Variable noisy currents
        
        # After the currents, it is necessary to update the membrane voltage
        # Integrating the function: dv(self,v,t,ie,ii,i_offset,i_noise,vrest)
        
        # Manually doing the integration:
        self.neurons_membrane+=self.dv(self.neurons_membrane,self.time,self.neurons_exc_curr,self.neurons_inh_curr,self.noisy_offset_currents,self.noisy_currents,self.vrest)*self.step

        
        self.test_refrac = self.refrac_timer>0 # Verifies which ones are still in the refractory period
        self.neurons_membrane[self.test_refrac] = self.vresets[self.test_refrac] # Resets the membrane of the refractory ones
        
        self.test_vthres = self.neurons_membrane>self.vthres # Verifies who should spike and creates a boolean vector with this information


        self.neurons_membrane[self.test_vthres] = self.vresets[self.test_vthres] # Resets the membrane of the spiked ones

        self.refrac_timer[self.test_vthres] = self.refractory_vector[self.test_vthres] # Resets the refractory timer to the ones who spiked

        # If I update the time here makes it possible to generate things at time zero!
        self.time = self.time + self.step # Advances the simulation time one step
        self.refrac_timer -= self.step # Subtracts the refractory timer to enable the refractory calculation

    def expose_exc_currents(self):
        return self.neurons_exc_curr
    
    def expose_inh_currents(self):
        return self.neurons_inh_curr    
```

```python
numpy.random.seed() # Forces the numpy to seed the random generator
my_seed=numpy.random.randint(0,10000,5)

#             my_seed[0]: Membrane initial potentials
#             my_seed[1]: Noisy offset currents
#             my_seed[2]: Selection of the inhibitory and excitatory neurons
#             my_seed[3]: Internal connections of the liquid
#             my_seed[4]: Noisy corrents

import brian

try:
    brian.reinit() # Necessary when using a python console
except:
    print "Brian REINIT error..."

import numpy # I could use "brian.", because Brian imports numpy, but I prefer not.
import time

import lsm_connections_probability as lm # Creates the 3D Grid and the connections according to Maass 2002

import lsm_dynamical_synapses_v1 as ls # Creates the dynamical synapses according to Maass 2002 and using the output 
                                       # from the lsm_connections_probability    


Net_shape=(15,3,3)
Number_of_neurons_lsm=135

my_STEP = 0.2
STP_OFF = True # Tells the system to use or not STP in the liquid
lbd_value =1.2 # lbd controls the connections probabilities

AINP = 100 # Input gains in nA

# These make easier to use the Brian objects without the "brian." at the beginning
ms = brian.ms
mV = brian.mV
nA = brian.nA
nF = brian.nF
NeuronGroup = brian.NeuronGroup
SpikeGeneratorGroup = brian.SpikeGeneratorGroup
Synapses = brian.Synapses
SpikeMonitor = brian.SpikeMonitor
network_operation = brian.network_operation 
defaultclock = brian.defaultclock

defaultclock.dt = my_STEP*ms



initial_time = time.time()

print "Initial time (in seconds):",initial_time

print "#"*78
print "#"*78
print "Liquid State Machine - 2 DoF arm experiments!"
print "#"*78
print "#"*78



lsm_3dGrid_flat = numpy.zeros(Number_of_neurons_lsm) 
# This creates a numpy 1D array with 'Number_of_neurons_lsm' positions
# I'm using a numpy array to be able to use the reshape method to 
# change from 1D (vector) to 3D (matrix)


#
# Number of Inhibitory neurons - LIQUID - 20% of the total neurons
lsm_indices = numpy.arange(Number_of_neurons_lsm) # Generates a list of indices

numpy.random.seed(my_seed[2]) # Forces the numpy to seed the random generator

numpy.random.shuffle(lsm_indices) # Shuffles the list

inhibitory_index_L = numpy.array(lsm_indices[:int(Number_of_neurons_lsm*0.2)]) # The first (approx.) 20% as inhibitory
excitatory_index = numpy.array(lsm_indices[int(Number_of_neurons_lsm*0.2):])# The rest of the neurons as excitatory


# This is the dictionary that has all the connections parameters according to Maass 2002.
# It is necessary to create the 3D connections and the STP configuration matrices
# E=>1 (excitatory) and I=>0 (inhibitory)
# Ex.: (0,0) => II
# Dynamical Synapses Parameters (STP):
Connections_Parameters={
              (0,0):[ # II
                      0.1,       # CGupta=0.1        # Parameter used at the connection probability - from Maass2002 paper
                      0.32,      # UMarkram=0.32     # Use (U) - Parameter used at the Dynamic Synapse - from Maass2002 paper
                      0.144,     # DMarkram=0.144    # Time constant for Depression (tau_rec) - used at the Dynamic Synapse - from Maass2002 paper                    
                      0.06,      # FMarkram=0.06     # Time constant for Facilitation (tau_facil) - used at the Dynamic Synapse - from Maass2002 paper
                      2.8,       # AMaass=2.8        # (nA) In the Maass2002 paper the value is negative, but because I need a positive scale (random.normal parameter) and there is a negative sign in front of the abs function I changed this to positive
                      0.8        # Delay_trans = 0.8 # In Maass paper the transmission delay is 0.8 to II, IE and EI        
                  ],
              (0,1):[ # IE
                      0.4,    # CGupta=0.4
                      0.25,   # UMarkram=0.25
                      0.7,    # DMarkram=0.7
                      0.02,   # FMarkram=0.02
                      3.0,    # AMaass=3.0 #in the Maass2002 paper the value is negative, but because I need a positive scale (random.normal parameter) and there is a negative sign in front of the abs function I changed this to positive
                      0.8     # Delay_trans = 0.8 # in Maass paper the transmission delay is 0.8 to II, IE and EI
                  ],
              (1,0):[ # EI
                      0.2,    # CGupta=0.2
                      0.05,   # UMarkram=0.05
                      0.125,  # DMarkram=0.125
                      1.2,    # FMarkram=1.2
                      1.6,    # AMaass=1.6
                      0.8     # Delay_trans = 0.8 # in Maass paper the transmission delay is 0.8 to II, IE and EI
                  ],
              (1,1):[ # EE
                      0.3,    # CGupta=0.3 
                      0.5,    # UMarkram=0.5
                      1.1,    # DMarkram=1.1
                      0.05,   # FMarkram=0.05
                      1.2,    # AMaass=1.2 #scaling parameter or absolute synaptic efficacy or weight - from Maass2002 paper
                      1.5     # Delay_trans = 1.5 # in Maass paper the transmission delay is 1.5 to EE connection
                  ]
              }


# Utilizes the functions in the lsm_connections_probability.py
# =>output = {'exc':connections_list_exc,'inh':connections_list_inh, '3Dplot_a':positions_list_a, '3Dplot_b':positions_list_b}
# connections_list_exc= OR connections_list_inh=
  # ((i,j), # PRE and POS synaptic neuron indexes
  # pconnection, # probability of the connection
  # (W_n, U_ds, D_ds, F_ds), # parameters according to Maass2002
  # Delay_trans, # parameters according to Maass2002
  # connection_type)

# Generate the connections matrix inside the Liquid (Liquid->Liquid) - according to Maass2002
#
print "Liquid->Liquid connections..."

numpy.random.seed(my_seed[3]) # Forces the numpy to seed the random generator

print "Generating the Liquid->Liquid connections..."
output_L_L = lm.generate_connections(lsm_3dGrid_flat, inhibitory_index_L, Net_shape, 
                                  CParameters=Connections_Parameters, lbd=lbd_value) # lbd controls the connections

print "Liquid->Liquid connections...Done!"

#
# These are the cell (neuron) parameters according to Maass 2002
#
cell_params_lsm = {  'cm'        : 30*nF,    # Capacitance of the membrane 
                                           # =>>>> MAASS PAPER DOESN'T MENTION THIS PARAMETER DIRECTLY
                                           #       but the paper mentions a INPUT RESISTANCE OF 1MEGA Ohms and tau_m=RC=30ms, so cm=30nF
                   'i_offset'  : 0.0*nA,   # Offset current - random for each neuron from [14.975nA to 15.025nA] => Masss2002 - see code below
                   'tau_m'     : 30.0*ms,  # Membrane time constant => Maass2002
                   'tau_refrac_E': 3.0*ms, # Duration of refractory period - 3mS for EXCITATORY => Maass2002
                   'tau_refrac_I': 2.0*ms, # Duration of refractory period - 2mS for INHIBITORY => Maass2002
                   'tau_syn_E' : 3.0*ms,   # Decay time of excitatory synaptic current => Maass2002
                   'tau_syn_I' : 6.0*ms,   # Decay time of inhibitory synaptic current => Maass2002
                   'v_reset'   : 13.5*mV,  # Reset potential after a spike => Maass2002
                   'v_rest'    : 0.0*mV,   # Resting membrane potential => Maass2002
                   'v_thresh'  : 15.0*mV,  # Spike threshold => Maass2002
                   'i_noise'   : 0.2*nA    # Used in Maass2002: mean 0 and SD=0.2nA
                }

# IF_curr_exp - MODEL EXPLAINED
# Leaky integrate and fire model with fixed threshold and
# decaying-exponential post-synaptic current. 
# (Separate synaptic currents for excitatory and inhibitory synapses)
lsm_neuron_eqs='''
  dv/dt  = (ie + ii + i_offset + i_noise)/c_m + (v_rest-v)/tau_m : mV
  die/dt = -ie/tau_syn_E                : nA
  dii/dt = -ii/tau_syn_I                : nA
  tau_syn_E                             : ms
  tau_syn_I                             : ms
  tau_m                                 : ms
  c_m                                   : nF
  v_rest                                : mV
  i_offset                              : nA
  i_noise                               : nA
  '''
# lsm_neuron_eqs='''
#   dv/dt  = (ie + ii + i_offset + i_noise)/c_m + (v_rest-v)/tau_m : mV
#   ie                                    : nA
#   ii                                    : nA
#   tau_m                                 : ms
#   c_m                                   : nF
#   v_rest                                : mV
#   i_offset                              : nA
#   i_noise                               : nA
#   '''



########################################################################################################################
#
# LIQUID - Setup
#
print "LIQUID - Setup..."

# Creates a vector with the corresponding refractory period according to the type of neuron (inhibitory or excitatory)
# IT MUST BE A NUMPY ARRAY OR BRIAN GIVES CRAZY ERRORS!!!!!
refractory_vector = [ cell_params_lsm['tau_refrac_E'] ]*Number_of_neurons_lsm # fills the list with the value corresponding to excitatory neurons
for i in range(Number_of_neurons_lsm):
  if i in inhibitory_index_L:
      refractory_vector[i]=cell_params_lsm['tau_refrac_I'] # only if the neuron is inibitory, changes the refractory period value!

refractory_vector=numpy.array(refractory_vector) # Here it is converted to a NUMPY ARRAY


# This is the population (neurons) used exclusively to the Liquid (pop_lsm).
# All the neurons receive the same threshold and reset voltages.
pop_lsm = NeuronGroup(Number_of_neurons_lsm, model=lsm_neuron_eqs, 
                                             threshold=cell_params_lsm['v_thresh'], 
                                             reset=cell_params_lsm['v_reset'], 
                                             refractory=refractory_vector, 
                                             max_refractory=max(cell_params_lsm['tau_refrac_E'], 
                                                                cell_params_lsm['tau_refrac_I']))


# Here I'm mixing numpy.fill with the access of the state variable "c_m" in Brian (because Brian is using a numpy.array)
# Sets the value of the capacitance according to the cell_params_lsm (same value to all the neurons)
pop_lsm.c_m.fill(cell_params_lsm['cm'])


# Sets the value of the time constant RC (or membrane constant) according to the cell_params_lsm (same value to all the neurons)
pop_lsm.tau_m.fill(cell_params_lsm['tau_m'])

# Sets the i_offset according to Maass2002
# The i_offset current is random, but never changes during the simulation.
# this current should be drawn from a uniform distr [14.975,15.025]
# Joshi2005 does [13.5,14.5] ???? Maybe is to avoid spikes without inputs...
numpy.random.seed(my_seed[1]) # Forces the numpy to seed the random generator
pop_lsm.i_offset=numpy.random.uniform(14.975,15.025, Number_of_neurons_lsm)*nA



pop_lsm.tau_syn_E.fill(cell_params_lsm['tau_syn_E']) # (same value to all the neurons)
pop_lsm.tau_syn_I.fill(cell_params_lsm['tau_syn_I']) # (same value to all the neurons)


# All neurons receive the same value to the resting voltage.
pop_lsm.v_rest.fill(cell_params_lsm['v_rest']) # (same value to all the neurons)


# Sets the initial membrane voltage according to Maass2002. Doesn't change during the simulation.
# this current should be drawn from a uniform distr [13.5mV,15.0mV]
numpy.random.seed(my_seed[0]) # Forces the numpy to seed the random generator
pop_lsm.v=numpy.random.uniform(13.5,15.0, Number_of_neurons_lsm)*mV


#
# Loading or creating the Synapses objects used within the Liquid
print "Liquid->Liquid connections..."

syn_lsm_obj = ls.LsmConnections(pop_lsm, pop_lsm, output_L_L, nostp=STP_OFF)

# Generates the Liquid->Liquid - EXCITATORY synapses
syn_lsm_exc = syn_lsm_obj.create_synapses('exc')

# Generates the Liquid->Liquid - INHIBITORY synapses
syn_lsm_inh = syn_lsm_obj.create_synapses('inh')


print "Liquid->Liquid connections...Done!"

total_number_of_connections_liquid = len(syn_lsm_exc) + len(syn_lsm_inh)

print "Number of inhibitory synapses in the Liquid: " + str(len(syn_lsm_inh)) # DEBUG to verify if it is working
print "Number of excitatory synapses in the Liquid: " + str(len(syn_lsm_exc)) # DEBUG to verify if it is working


# To understand what is being returned:
# pop_lsm: it is necessary to connect the neuron network with the rest of the world
# [syn_lsm_obj, syn_lsm_exc, syn_lsm_inh]: to include these objects at the simulation (net=Net(...); net.run(total_sim_time*ms)); 
# It is a list because is easy to concatenate lists :D

print "LIQUID - Setup...Done!"

#
# End of the LIQUID - Setup
########################################################################################################################

numpy.random.seed(my_seed[4]) # Forces the numpy to seed the random generator

########################################################################################################################
#
# INPUT - Setup
#
print "INPUT - Setup..."

tspk = defaultclock.dt*50 # The neurons spike after 50 time steps!
number_of_spikes = 50

spiketimes = [(i,tspk) for i in range(number_of_spikes)] 
                # The spikes are going to be received during the simulation, 
                # so this is always an empty list when using the step_by_step_brian_sim!


# I'm using only one big input layer because Brian docs say it is better for the performance
SpikeInputs = SpikeGeneratorGroup(number_of_spikes, spiketimes)


#
#
# Here the synapses are created. The synapses created are ALWAYS excitatory because it is 
# connecting through 'ie' in the neuron model!

syn_world_Input = Synapses(SpikeInputs, pop_lsm,
                                     model='''w : 1''',
                                     pre='''ie+=w''')


for i in range(len(spiketimes)):
    syn_world_Input[i,i] = True

weights_input_liquid = numpy.array([AINP]*number_of_spikes)*nA

syn_world_Input.w = weights_input_liquid
syn_world_Input.delay=0*ms

print "INPUT - Setup...Done!"

#
# End of the INPUT - Setup (creation of the connections between the Poisson input and the Liquid!)
#
########################################################################################################################


# DON'T FORGET TO INSERT THE FUNCTION "generate_i_noise" INTO THE MONITORS_OBJECT LIST!!!!
# Generates the noisy current at each time step (as seen in Joshi2005)
@network_operation(clock=defaultclock)
def generate_i_noise():
    # These are the noise currents inside each liquid's neuron
    pop_lsm.i_noise=numpy.random.normal(loc=0, scale=cell_params_lsm['i_noise'],size=Number_of_neurons_lsm)*nA

populations_sim = [pop_lsm, SpikeInputs]

synapses_sim = [syn_lsm_exc, syn_lsm_inh, syn_world_Input]

Input_layer, Output_layer, pop_objects, syn_objects, monitors_objects = SpikeInputs, pop_lsm, populations_sim, synapses_sim, [generate_i_noise]

OutputMonitor=brian.SpikeMonitor(pop_lsm, record=True)
VMonitor=brian.StateMonitor(pop_lsm,'v', record=True)
IEMonitor=brian.StateMonitor(pop_lsm,'ie', record=True)

net = brian.Network(pop_objects + syn_objects + monitors_objects + [OutputMonitor,VMonitor,IEMonitor])

print "Setup time:", time.time()-initial_time
initial_time = time.time()

net.run(500*my_STEP*ms)
print "Simulation time:", time.time()-initial_time
```
