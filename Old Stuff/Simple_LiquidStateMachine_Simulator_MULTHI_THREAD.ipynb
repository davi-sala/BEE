```python
%%file multi_SNN_simulator.c

/*

---------------------------------------
Gaucho's Liquid State Machine Simulator
---------------------------------------
http://en.wikipedia.org/wiki/Gaucho
http://en.wikipedia.org/wiki/Liquid_state_machine
https://github.com/ricardodeazambuja


This is a VERY VERY (VERY) simple simulator. It implements the neuron model:
    dv/dt  = (ie + ii + i_offset + i_inj)/cm + (v_rest-v)/taum  [in V]
    die/dt = -ie/taue   [in nA]
    dii/dt = -ii/taui   [in nA]

    => ie is the state variable corresponding to the excitatory neurons current
    => ii is the state variable corresponding to the inhibitory neurons current
    => v is the state variable corresponding to the membrane voltage
    => cm is the membrane capacitance
    => v_rest is the membrane voltage resting potential 
    => taum is the membrane time constant
    => taue is the excitatory current time constant
    => taui is the inhibitory current time constant    
    => i_offset is a constant random value generated at the liquid's initialization
    => i_inj is a random value generated at each time step.

    The differential equation is solved by a simple numerical method (midpoint rule / rectangle rule?)
    One questions is: What is better? Smaller time step and simpler integration algorithm or bigger time step and
    more advanced (precise) integration????

And uses the ideas from:
Maass, Wolfgang, Thomas Natschlager, and Henry Markram. 2002. 
“Real-Time Computing without Stable States: A New Framework for Neural Computation Based on Perturbations.” 
Neural Computation 14 (11): 2531–60. doi:10.1162/089976602760407955.

However, the dynamic synapses (STP) and the time delays are NOT implemented in an attempt of making the system faster
and also because I'm not sure if they are necessary.

My initial idea is to interface using Python (ctypes). I developed the same algorithm using Python, so for a step-by-step
simulation, maybe it is better to use the Python version.


=============
| Comments: |
=============
- rand from the C standard library is not a great random generator, but I don't know how good the other ones are, so I will stick to rand for now.
(what I could read about rand is that the implementations vary between systems, so simply using another generator is not a clever solution...)
- I'm not using sparse matrices because I'm not planning to work with big liquids sparsely connected. Because of the small world network style
the liquid has (http://en.wikipedia.org/wiki/Small-world_network), usually all neurons are connected to at least one neighbour.
- Looking through this code I was wondering what if I made a simulation with a variable time step.
- It's necessary to check for things that are not "thread safe" in the code, once again.
- Start thinking about CUDA (blocks with multiple GPU cores that run the same code in different memory positions) and 
not only normal CPUs whilst developing the threaded version!!!!!
- I'm supposing the threads are allways JOINABLE instead of DETACHED by default!!!!!
- Maybe, using CUDA, I could make the time spent in a step to be deterministic. Now the system spends more or less time 
according to the amount of received spikes, etc.


************************************************************
************************************************************
- I NEED TO VERIFY WHICH MEMORY POSITION I NEED TO FREE!!!!!
************************************************************
************************************************************

- Solution to the random stuff in parallel (N threads):
0) create N variables to store each of the seeds used by the N threads; This variables should be an 1D array to make it easier
to access using the thread index as the multiplier for the pointer.
1) use each of the user seed to generate a new set of N random numbers;
2) create an structure with the state variables used with the gaussian distribution

- I want at least two operation mode:
1) Receives a memory address where the input spikes should come and another where the output spikes should go.
In this situation I'm going to generate by hand (forced teaching) the values that feedback from the readout to the liquid's input.
The two variables (input and output spikes) are known since the start, so it's possible to easily use a numpy array.
I will need to think HOW AM I GOING TO INSERT NOISE in these inputs. If I generate the inputs already with the noise from Python,
it's going to be necessary to do a lot of processing in Python. Inserting noise directly from C enables me to use the same
inputs for all the experiments.

2) Do almost the same as 1), but now I want to simulate the system instead of forced teaching. 
Therefore I will need to have a function to do the readout calculations online and feedback those values to the right liquid's inputs.



************************************************************************************************************************

Compile as a shared library:
gcc -shared -Wl, -install_name, file_name.so -o file_name.so -fPIC file_name.c -O3 -lpthreads -Wall

(Wall forces to give all the warnings(???))

In order to compile it in Linux it is necessary to link the "math.h" manually including "-lm", "-lrt" and also "-std=gnu99"

gcc my_simulator_pthreads.c -o my_simulator_pthreads.o -lpthread -O3 -lrt -lm -std=gnu99 -g

the "-g" flag is ONLY necessary to use if you want to debug with gdb

In gdb:
1)If the program receives command line arguments:
r arg1 arg2 arg3...
2)Creates a break line at line 444 (must be executed before run or r)
b 444
3)Steps into
step or n
4)Keep going
next or n
5)Print variables content
print or p


To run the profiler:
gcc my_simulator_pthreads.c -o my_simulator_pthreads.o -lpthread -O3 -lrt -lm -std=gnu99 -g -pg

then run once the executable. After that, run (on Linux, because I could not make it work on OSX...)
gprof ./my_simulator_pthreads.o > profile.txt
Done! Read the txt file :)

./

*/

#include <stdlib.h>
#include <stdio.h>
#include <math.h>
#include <string.h> // only for the memcpy inside the main()
#include <pthread.h>

#include <limits.h> // has the maximum values of types: e.g UCHAR_MAX


//
// Sets the right function for the timing debug stuff for GCC LINUX(__linux) and OSX (__APPLE_CC__)
// in Linux it is necessary to add gcc (linker) option: -lrt
//
#if defined (__linux__) || defined (linux)
#include <time.h> // only for the time comparisons inside the main()
double clock_my_gettime() {
    //  time_t tv_sec; //seconds 
    //  long tv_nsec;  //NANOseconds    
    struct timespec now;
    clock_gettime(CLOCK_MONOTONIC, &now);
    return ((double)now.tv_sec*1E9 + (double)now.tv_nsec); //nanoseconds
}
#endif

#if defined (__APPLE_CC__)
#include <sys/time.h>
double clock_my_gettime(){
    // time_t       tv_sec;    //seconds since Jan. 1, 1970 
    // suseconds_t  tv_usec;   //and MICROseconds    
    struct timeval now;
    int rv = gettimeofday(&now, NULL);
    if (rv) return rv; //rv=0 means OK!
    return ((double)now.tv_sec*1E9 + (double)now.tv_usec*1E3); //nanoseconds
}
#endif


//
// Functions prototypes
void *SpikingLiquid_generate_connections_thread(void *args);
void SpikingLiquid_generate_connections_memory();
void *SpikingLiquid_Reset_thread(void *args);
void SpikingLiquid_Reset();
void SpikingLiquid_init();
unsigned SpkLiq_time_seed();
double SpkLiq_uniform_deviate(int seed);
double SpkLiq_gaussrand_threads(unsigned int *seed, int *phase, double *V1, double *V2, double *S);
double SpkLiq_euclidean_dist(int i, int j, int xd, int yd);
void SpkLiq_process_exc_spikes(int *spikes, double *weights, int number_of_spikes);
void SpkLiq_process_inh_spikes(int *spikes, double *weights, int number_of_spikes);
void *SpikingLiquid_process_internal_spikes_threads(void *args);
void *SpikingLiquid_update_internal_thread(void *args);
void SpikingLiquid_update(int *spikes_exc, int *spikes_inh, double *weights_exc, double *weights_inh, int size_exc, int size_inh);


// This is used to define when realloc is going to ask for more memory
// inside the SpikingLiquid_generate_connections_thread function
#define MAX_CONNECTIONS 10

// Here we define the maximum number of threads
#define MAX_THREAD 50


// SpkLiq_liquid_parametres
// #1:  CGupta       # Parameter used at the connection probability
// #2:  UMarkram     # Use (U) - Parameter used at the Dynamic Synapse
// #3:  DMarkram     # Time constant for Depression (tau_rec) - used at the Dynamic Synapse
// #4:  FMarkram     # Time constant for Facilitation (tau_facil) - used at the Dynamic Synapse
// #5:  AMaass       # Connection gain (weight) (in nA)
// #6:  Delay_trans  # Transmission delay
//
// The Dynamic Synapses (STP - Short Term Plasticity) and the Transmission delay are not implemented... yet.
//
double const SpkLiq_liquid_parametres[2][2][6] = {{
                                              { 0.1  ,  0.32 ,  0.144,  0.06 ,  -2.8  ,  0.8  },
                                              { 0.4  ,  0.25 ,  0.7  ,  0.02 ,  -3.0  ,  0.8  }
                                          },
                                          {
                                              { 0.2  ,  0.05 ,  0.125,  1.2  ,  1.6  ,  0.8  },
                                              { 0.3  ,  0.5  ,  1.1  ,  0.05 ,  1.2  ,  1.5  }
                                          }};


//
// Global variables
// These variables are global because they are used by most of the functions and would be a pain to pass as parametres!
// The first version of this algorithm was developed in an object-oriented way (Python), so these global variables were inside the main class.
// Yes, I know about global variables being bad practice, but sometimes it is faster to use them directly instead of moving memory contents, creating pointers, etc.
//


// User's input variables
int SpkLiq_net_shape[3]; // the sides of the parallelepiped modeling the liquid's shape {x,y,z} (unit less, because it's a relative measurement)
double SpkLiq_lbd_value; // value that controls the probability of a connection inside the liquid (see Maass,Markram 2002)
double SpkLiq_step; // simulation time step size (in seconds)
double SpkLiq_taum; // neuron membrane time constant (in seconds)
double SpkLiq_cm; // neuron membrane capacitance (in Farads)
double SpkLiq_taue; // excitatory neuron decay time constant (in seconds)
double SpkLiq_taui; // inhibitory neuron decay time constant (in seconds)
double SpkLiq_refractory[2]; // refractory periods {inhibitory, excitatory} (in seconds)
unsigned int SpkLiq_user_seeds[5]; //random seeds supplied by the user
double SpkLiq_vresets; // stores the voltage value the membrane must assume after a reset (refractory or spike) (in Volts)
double SpkLiq_vthres; // stores the membrane voltage value used to verify if the neuron is going to spike (in Volts)
double SpkLiq_vrest; // stores the voltage value the membrane must assume after total discharge (in Volts)
double SpkLiq_inhibitory_percentage; // the percentage of inhibitory neurons in the liquid (unit less, %)
double SpkLiq_membrane_rand[2]; //range of the possible values to the initialization of the membrane voltage (in Volts)
double SpkLiq_current_rand[2]; //range of the possible values to the initialization of the constant noisy offset current (in Amperes)
double SpkLiq_noisy_current_rand; //max range of the possible values to the initialization of the variable noisy current (in Amperes)


// Random Seeds - MASTERS
unsigned int SpkLiq_rand_r_seeds[5]; //stores the random seeds variables used by the rand_r function



// Pointers are used here because the memory size varies according to the user's values
double *SpkLiq_neurons_membrane = NULL; //the membrane potentials for all neurons
double *SpkLiq_neurons_exc_curr = NULL; //excitatory currents levels for all neurons
double *SpkLiq_neurons_inh_curr = NULL; //inhibitory currents levels for all neurons
double *SpkLiq_refrac_timer = NULL; //timers used to control the refractory period
int *SpkLiq_test_vthres = NULL; //controls which neuron is above the voltage threshold and should spike
double *SpkLiq_noisy_offset_currents = NULL; //values of the neuron model constant noisy offset currents
double *SpkLiq_noisy_currents = NULL; //values of the neuron model variable noisy currents
double *SpkLiq_refrac_values = NULL; //stores the refractory periods for all neurons in the liquid 
                                     //because the refractory period is different for inhibitory and excitatory neurons

int *SpkLiq_inhibitory_indices = NULL; //Indices of the inhibitory neurons inside the liquid
int *SpkLiq_excitatory_indices = NULL; //Indices of the excitatory neurons inside the liquid
int *SpkLiq_liquid_indices_shuffled = NULL; //used to generate the random SpkLiq_inhibitory_indices/SpkLiq_excitatory_indices


int *SpkLiq_pre_i = NULL; //stores the indices of the inhibitory=>? pre-synaptic connections in the liquid
int *SpkLiq_pos_i = NULL; //stores the indices of the inhibitory=>? pos-synaptic connections in the liquid
double *SpkLiq_w_i = NULL; //stores the weights of the connections above

int *SpkLiq_pre_e = NULL; //stores the indices of the inhibitory=>? pre-synaptic connections in the liquid
int *SpkLiq_pos_e = NULL; //stores the indices of the inhibitory=>? pos-synaptic connections in the liquid
double *SpkLiq_w_e = NULL; //stores the weights of the connections above

int *SpkLiq_spiked_i = NULL; //indices of the inhibitory neurons that spiked in the last simulation SpkLiq_step
int *SpkLiq_spiked_e = NULL; //indices of the inhibitory neurons that spiked in the last simulation SpkLiq_step

int SpkLiq_number_of_neurons; //total number of neurons inside the liquid, automatically calculated based on the liquid's shape
double SpkLiq_current_time; //the current simulation time (in seconds)

int SpkLiq_number_of_exc_neurons; //total number of excitatory neurons in the liquid
int SpkLiq_number_of_inh_neurons; //total number of inhibitory neurons in the liquid

int SpkLiq_total_number_of_connections; //stores the total number of connections (synapses) generated inside the liquid after initialization
int SpkLiq_inh_connections; //stores how many inhibitory=>? connections were created
int SpkLiq_exc_connections; //stores how many excitatory=>? connections were created


//
// New variables used with the multithreading
//
unsigned int *SpkLiq_threads_seeds[5] = {NULL,NULL,NULL,NULL,NULL}; 
                                       // Each row match one of the Random Variables used in the simulator (Random-0...4)
                                       // and the columns are equivalent to the thread index

// The gaussian distribution needs some state variables, so it is necessary to one set to each thread
// Random-3 (noisy currents) and Random-4 (generation of connections)
// Index 0 => Random-3 (noisy currents)
// Index 1 => Random-4 (generation of connections)
typedef struct _gaussian_states {
  int phase[2]; // I need two positions because I'll have two random variables (read comment above...)
  double V1[2];
  double V2[2];
  double S[2];
} SpkLiq_gaussian_states;

SpkLiq_gaussian_states *SpkLiq_threads_gaussian = NULL; // This array is going to have the size of number of threads N

int SpkLiq_threads_N; // Total number of threads used

int *SpkLiq_number_of_neurons_slice = NULL;

// Structure passed to the thread used inside the function SpikingLiquid_Reset
// typedef struct _SpkLiq_thread_data_reset {
//   int tid;
//   long int start;
//   long int end;
// } SpkLiq_thread_data_reset;

pthread_t *SpkLiq_threads_ptrs = NULL; // Creates the pointers used with the pthread_create
                                       // These pointer need to allocate SpkLiq_threads_N memory positions

double *SpkLiq_prob_dist = NULL; //Stores the distance probability values calculated by the threads to speed up

int *SpkLiq_thread_id = NULL; // Tells the thread function its thread index (thread_id) making easier to control which memory
                              // positions each thread is going to write to.


// SpikingLiquid_generate_connections_thread
int **SpkLiq_pre_i_thread = NULL; //stores the intermediary array of indices (generated by the threads) of the inhibitory=>? pre-synaptic connections in the liquid
int **SpkLiq_pos_i_thread = NULL; //stores the intermediary array of  indices (generated by the threads) of the inhibitory=>? pos-synaptic connections in the liquid
double **SpkLiq_w_i_thread = NULL; //stores the weights of the connections above

int **SpkLiq_pre_e_thread = NULL; //stores the intermediary array of  indices (generated by the threads) of the excitatory=>? pre-synaptic connections in the liquid
int **SpkLiq_pos_e_thread = NULL; //stores the intermediary array of  indices (generated by the threads) of the excitatory=>? pos-synaptic connections in the liquid
double **SpkLiq_w_e_thread = NULL; //stores the weights of the connections above

int *SpkLiq_inh_connections_thread = NULL; //stores how many inhibitory connections each thread is creating
int *SpkLiq_exc_connections_thread = NULL; //stores how many excitatory connections each thread is creating

int *SpkLiq_number_of_inh_connections_thread = NULL; //stores how many inhibitory connections each thread will process
int *SpkLiq_number_of_exc_connections_thread = NULL; //stores how many excitatory connections each thread will process

// SpikingLiquid_process_internal_spikes_threads
int *SpkLiq_number_of_inh_neurons_thread = NULL; //stores the number of inhibitory neurons each thread is going to receive
int *SpkLiq_number_of_exc_neurons_thread = NULL; //stores the number of excitatory neurons each thread is going to receive

int **SpkLiq_receive_spike_i_thread = NULL; //stores the indices of which liquid's inhibitory neuron receives a spike (output from the threads)
double **SpkLiq_receive_spike_i_w_thread = NULL; //stores the weights of the above received spikes
int *SpkLiq_receive_spike_i_idx_thread = NULL; //stores how many items are inside each of the above arrays (each thread is going to generate an variable size)

int **SpkLiq_receive_spike_e_thread = NULL; //stores the indices of which liquid's excitatory neuron receives a spike (output from the threads)
double **SpkLiq_receive_spike_e_w_thread = NULL; //stores the weights of the above received spikes
int *SpkLiq_receive_spike_e_idx_thread = NULL; //stores how many items are inside each of the above arrays (each thread is going to generate an variable size)


//
// Here comes the functions!
//


void *SpikingLiquid_generate_connections_thread(void *args){

    /*
    Generates the liquid's probabilistic small world network structure.

    Stores the liquid's structure in the following arrays (each index "thread_id" represents one thread):
    (at the end the data must be brought together becasue which thread generates a set of separated arrays)
    SpkLiq_pre_i_thread[thread_id]: indices of the inhibitory PREsynaptic connections
    SpkLiq_pos_i_thread[thread_id]: indices of the inhibitory POSsynaptic connections
    SpkLiq_w_i_thread[thread_id]: weights of the inhibitory=>? connections

    SpkLiq_pre_e_thread[thread_id]: indices of the excitatory PREsynaptic connections
    SpkLiq_pos_e_thread[thread_id]: indices of the excitatory POSsynaptic connections
    SpkLiq_w_e_thread[thread_id]: weights of the excitatory=>? connections

    SpkLiq_inh_connections_thread[thread_id]: total number of INHIBITORY connections generated by the thread
    SpkLiq_exc_connections_thread[thread_id]: total number of EXCITATORY connections generated by the thread
    */

    int *SpkLiq_tid = (int *)args; //Receives the number of the thread from the function input arguments 
    
    int PRE; //Flags to indicate the type of connection: inhibitory=>0 and excitatory=>1
    int POS; //during the double for loop where the connections are generated.

    int n_malloc_i = 1; //Counter used with realloc to dynamically allocate more memory: inhibitory connections
    int n_malloc_e = 1; //Counter used with realloc to dynamically allocate more memory: excitatory connections

    int i=0,j=0; //Used in the for loops

    int loop_start, loop_end;

//NEW
    SpkLiq_inh_connections_thread[*SpkLiq_tid] = 0;
    SpkLiq_exc_connections_thread[*SpkLiq_tid] = 0;

    //
    //Now each thread is going to have its own arrays and acesses using the thread index "SpkLiq_tid"
    //
    //Allocate memory to save the inhibitory=>? connections
    SpkLiq_pre_i_thread[*SpkLiq_tid] = NULL;
    SpkLiq_pre_i_thread[*SpkLiq_tid] = calloc(MAX_CONNECTIONS,sizeof(int));
    SpkLiq_pos_i_thread[*SpkLiq_tid] = NULL;
    SpkLiq_pos_i_thread[*SpkLiq_tid] = calloc(MAX_CONNECTIONS,sizeof(int));
    SpkLiq_w_i_thread[*SpkLiq_tid] = NULL;
    SpkLiq_w_i_thread[*SpkLiq_tid] = calloc(MAX_CONNECTIONS,sizeof(double));

    //Allocate memory to save the excitatory=>? connections
    SpkLiq_pre_e_thread[*SpkLiq_tid] = NULL;
    SpkLiq_pre_e_thread[*SpkLiq_tid] = calloc(MAX_CONNECTIONS,sizeof(int));
    SpkLiq_pos_e_thread[*SpkLiq_tid] = NULL;
    SpkLiq_pos_e_thread[*SpkLiq_tid] = calloc(MAX_CONNECTIONS,sizeof(int));
    SpkLiq_w_e_thread[*SpkLiq_tid] = NULL;
    SpkLiq_w_e_thread[*SpkLiq_tid] = calloc(MAX_CONNECTIONS,sizeof(double));


    if (SpkLiq_pre_i_thread[*SpkLiq_tid]==NULL || SpkLiq_pos_i_thread[*SpkLiq_tid]==NULL || SpkLiq_w_i_thread[*SpkLiq_tid]==NULL || SpkLiq_pre_e_thread[*SpkLiq_tid]==NULL || SpkLiq_pos_e_thread[*SpkLiq_tid]==NULL || SpkLiq_w_e_thread[*SpkLiq_tid]==NULL){
        fprintf(stderr,"connections memory malloc ERROR!");
        exit(EXIT_FAILURE); //I need to check which error I should signal here...
    }

// Each thread receives a slice of the total number of neurons.
// but if the result of the division of total number of neurons by number of threads
// has a remainder, then the last thread will have a bigger set of values to include the remainder.

    // Sets up the loop variables according to the thread number
    if ((*SpkLiq_tid)!=(SpkLiq_threads_N-1)){
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)];
      loop_end = (SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)])*((*SpkLiq_tid)+1);
    }else{ // It means the last thread that could have more work to do because of the remainder
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)-1]; //Uses the anterior value (the rounded one) and multiplies
      loop_end = (SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)-1])*(*SpkLiq_tid)+(SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)]);
    }


    // Goes through all the neurons indices and verify if they are inhibitory (excitatory test is implicit)
    // Instead of using the ordered list, I'm going to use the shuffled one because this way I don't need to check 
    // which neuron is excitatory or inhibitory in a difficult (cpu costly) way.
    // for (i=(*SpkLiq_tid)*SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)]; i<SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)]*((*SpkLiq_tid)+1); i++){ //These are the PRE indices    
    for (i=loop_start; i<loop_end; i++){ //These are the PRE indices
        PRE = 1; //Because it's only tested against inhibitory indices

        // Verifies if the neuron index 'i' belongs to SpkLiq_inhibitory_indices
        if(i<SpkLiq_number_of_inh_neurons){
            PRE = 0; //Indicates the PRE-synaptic neuron is inhibitory
        }


        for (j=0;j<SpkLiq_number_of_neurons; j++){ //These are the POS indices
            POS = 1;//Because it's only tested against inhibitory indices
            
            // Verifies if the neurons indices 'i' and 'j' belongs to SpkLiq_inhibitory_indices
            if(j<SpkLiq_number_of_inh_neurons){
                POS = 0; //Indicates the POS-synaptic neuron is inhibitory
            }
            

            //Here the variables about PRE/POS are set

            // This test is the most expensive calculation in this loop
            // According to my tests, the nested for loops are about 170 faster than only this calculation.
            // Also, to allocate memory to store the results of ALL possible combinations is super fast.
            // So I should do this calculation in parallel, storing the results in memory.
            // After that I couldl do one thread to generate the vectors SpkLiq_pre_i_thread, ...
            // or even think about a more optimized way because, after all, the update function is VERY important too.


            // Verifies if a connection (synapse) will be created or not depending on the calculated probability.
            // # RANDOM-3
            // # Internal connections of the liquid
            if ( SpkLiq_uniform_deviate(rand_r(&SpkLiq_threads_seeds[3][*SpkLiq_tid])) <= (SpkLiq_liquid_parametres[PRE][POS][0]*(exp(-pow(SpkLiq_euclidean_dist(SpkLiq_liquid_indices_shuffled[i],SpkLiq_liquid_indices_shuffled[j],SpkLiq_net_shape[0],SpkLiq_net_shape[1])/SpkLiq_lbd_value,2)))) ){            
                //It means we have a synapse here!

                if (PRE==0){//Inhibitory=>? connection
                    SpkLiq_inh_connections_thread[*SpkLiq_tid]++;
                    
                    //Verifies if the array (used to store the Inhibitory=>? connections) needs more memory
                    if (SpkLiq_inh_connections_thread[*SpkLiq_tid]>(MAX_CONNECTIONS*n_malloc_i)){
                        n_malloc_i++;
                        SpkLiq_pre_i_thread[*SpkLiq_tid] = realloc(SpkLiq_pre_i_thread[*SpkLiq_tid],n_malloc_i*MAX_CONNECTIONS*sizeof(int));
                        SpkLiq_pos_i_thread[*SpkLiq_tid] = realloc(SpkLiq_pos_i_thread[*SpkLiq_tid],n_malloc_i*MAX_CONNECTIONS*sizeof(int));
                        SpkLiq_w_i_thread[*SpkLiq_tid] = realloc(SpkLiq_w_i_thread[*SpkLiq_tid],n_malloc_i*MAX_CONNECTIONS*sizeof(double));
                        if (SpkLiq_pre_i_thread[*SpkLiq_tid]==NULL || SpkLiq_pos_i_thread[*SpkLiq_tid]==NULL || SpkLiq_w_i_thread[*SpkLiq_tid]==NULL){
                            fprintf(stderr,"SpikingLiquid_generate_connections INH - realloc ERROR!");
                            exit(EXIT_FAILURE);
                        }
                    }
                         
                    SpkLiq_pre_i_thread[*SpkLiq_tid][SpkLiq_inh_connections_thread[*SpkLiq_tid]-1]=SpkLiq_liquid_indices_shuffled[i];
                    SpkLiq_pos_i_thread[*SpkLiq_tid][SpkLiq_inh_connections_thread[*SpkLiq_tid]-1]=SpkLiq_liquid_indices_shuffled[j];
                    SpkLiq_w_i_thread[*SpkLiq_tid][SpkLiq_inh_connections_thread[*SpkLiq_tid]-1]=1E-9*SpkLiq_liquid_parametres[PRE][POS][4]*fabs(SpkLiq_gaussrand_threads(&SpkLiq_threads_seeds[3][*SpkLiq_tid], &SpkLiq_threads_gaussian[*SpkLiq_tid].phase[0], &SpkLiq_threads_gaussian[*SpkLiq_tid].V1[0], &SpkLiq_threads_gaussian[*SpkLiq_tid].V2[0], &SpkLiq_threads_gaussian[*SpkLiq_tid].S[0])*0.5+1); //AMaass

                }else{//Excitatory=>? connections
                    SpkLiq_exc_connections_thread[*SpkLiq_tid]++;   
                    
                    //Verifies if the array (used to store the Excitatory=>? connections) needs more memory
                    if (SpkLiq_exc_connections_thread[*SpkLiq_tid]>(MAX_CONNECTIONS*n_malloc_e)){
                        n_malloc_e++;
                        SpkLiq_pre_e_thread[*SpkLiq_tid] = realloc(SpkLiq_pre_e_thread[*SpkLiq_tid],n_malloc_e*MAX_CONNECTIONS*sizeof(int));
                        SpkLiq_pos_e_thread[*SpkLiq_tid] = realloc(SpkLiq_pos_e_thread[*SpkLiq_tid],n_malloc_e*MAX_CONNECTIONS*sizeof(int));
                        SpkLiq_w_e_thread[*SpkLiq_tid] = realloc(SpkLiq_w_e_thread[*SpkLiq_tid],n_malloc_e*MAX_CONNECTIONS*sizeof(double));
                        if (SpkLiq_pre_e_thread[*SpkLiq_tid]==NULL || SpkLiq_pos_e_thread[*SpkLiq_tid]==NULL || SpkLiq_w_e_thread[*SpkLiq_tid]==NULL){
                            fprintf(stderr,"SpikingLiquid_generate_connections EXC - realloc ERROR!");
                            exit(EXIT_FAILURE);
                        }
                    }
                
                    SpkLiq_pre_e_thread[*SpkLiq_tid][SpkLiq_exc_connections_thread[*SpkLiq_tid]-1]=SpkLiq_liquid_indices_shuffled[i];
                    SpkLiq_pos_e_thread[*SpkLiq_tid][SpkLiq_exc_connections_thread[*SpkLiq_tid]-1]=SpkLiq_liquid_indices_shuffled[j];
                    SpkLiq_w_e_thread[*SpkLiq_tid][SpkLiq_exc_connections_thread[*SpkLiq_tid]-1]=1E-9*SpkLiq_liquid_parametres[PRE][POS][4]*fabs(SpkLiq_gaussrand_threads(&SpkLiq_threads_seeds[3][*SpkLiq_tid], &SpkLiq_threads_gaussian[*SpkLiq_tid].phase[0], &SpkLiq_threads_gaussian[*SpkLiq_tid].V1[0], &SpkLiq_threads_gaussian[*SpkLiq_tid].V2[0], &SpkLiq_threads_gaussian[*SpkLiq_tid].S[0])*0.5+1); //AMaass
                }
            }

        }
    }
    return NULL;
}


void SpikingLiquid_generate_connections_memory(){

    /*
    Copies the data generated by the SpikingLiquid_generate_connections_thread into unique arrays.
    Moving to a continuous segment of memory is going to make the simulation's update faster and avoid extra code 
    necessary to deal with a diffuse memory model.

    In a system with less memory, maybe it is going to be necessary to break the memory copy into small chunks to make
    it possible to free memory just after it had been copied (maybe using some realloc calls too).

    Because this memory copy is very fast, I decided to keep it in a serial code instead of using threads.
    */

    int i=0,j=0; //Used in the for loops

    // Sums up the totals generated by the SpikingLiquid_generate_connections_thread
    // This is necessary because we need to allocate the memory for the variables.
    SpkLiq_inh_connections = 0;
    SpkLiq_exc_connections = 0;
    SpkLiq_total_number_of_connections = 0;
    for(i=0; i<SpkLiq_threads_N; i++){
            SpkLiq_inh_connections+=SpkLiq_inh_connections_thread[i];
            SpkLiq_exc_connections+=SpkLiq_exc_connections_thread[i];
    }
    SpkLiq_total_number_of_connections=SpkLiq_inh_connections+SpkLiq_exc_connections; // This variable is only useful for debug or user interfacing

    // Allocates the memory according to the totals generated above
    SpkLiq_pre_i = malloc(sizeof(SpkLiq_pre_i)*SpkLiq_inh_connections);
    SpkLiq_pos_i = malloc(sizeof(SpkLiq_pos_i)*SpkLiq_inh_connections);
    SpkLiq_w_i = malloc(sizeof(SpkLiq_w_i)*SpkLiq_inh_connections);

    SpkLiq_pre_e = malloc(sizeof(SpkLiq_pre_e)*SpkLiq_exc_connections);
    SpkLiq_pos_e = malloc(sizeof(SpkLiq_pos_e)*SpkLiq_exc_connections);
    SpkLiq_w_e = malloc(sizeof(SpkLiq_w_e)*SpkLiq_exc_connections);

    if (SpkLiq_pre_i==NULL || SpkLiq_pos_i==NULL || SpkLiq_w_i==NULL || SpkLiq_pre_e==NULL || SpkLiq_pos_e==NULL || SpkLiq_w_e==NULL){
        fprintf(stderr,"SpikingLiquid_generate_connections_memory -  malloc ERROR!");
        exit(EXIT_FAILURE); //I need to check which error I should signal here...
    }


    // Effectively moves the data to the continuous positions allocated above
    int index_conter_i=0;
    int index_conter_e=0;
    for(i=0; i<SpkLiq_threads_N; i++){ // iterates through the threads
        for(j=0; j<SpkLiq_inh_connections_thread[i]; j++){ // iterates through the arrays generated by each thread
            SpkLiq_pre_i[index_conter_i] = SpkLiq_pre_i_thread[i][j];
            SpkLiq_pos_i[index_conter_i] = SpkLiq_pos_i_thread[i][j];
            SpkLiq_w_i[index_conter_i] = SpkLiq_w_i_thread[i][j];
            index_conter_i++;
        }
        // Frees the memory used by these variables!
        free(SpkLiq_pre_i_thread[i]);
        free(SpkLiq_pos_i_thread[i]);
        free(SpkLiq_w_i_thread[i]);

        for(j=0; j<SpkLiq_exc_connections_thread[i]; j++){
            SpkLiq_pre_e[index_conter_e] = SpkLiq_pre_e_thread[i][j];
            SpkLiq_pos_e[index_conter_e] = SpkLiq_pos_e_thread[i][j];
            SpkLiq_w_e[index_conter_e] = SpkLiq_w_e_thread[i][j];
            index_conter_e++;
        }
        // Frees the memory used by these variables!        
        free(SpkLiq_pre_e_thread[i]);
        free(SpkLiq_pos_e_thread[i]);
        free(SpkLiq_w_e_thread[i]);
    }
        free(SpkLiq_pre_i_thread);
        free(SpkLiq_pos_i_thread);
        free(SpkLiq_w_i_thread);

        free(SpkLiq_pre_e_thread);
        free(SpkLiq_pos_e_thread);
        free(SpkLiq_w_e_thread);

        free(SpkLiq_inh_connections_thread);
        free(SpkLiq_exc_connections_thread);
}



void *SpikingLiquid_Reset_thread(void *args){
    /*
    Resets the neurons internal variables according to the thread id.
    - Each thread must iterate through a slice of the SpkLiq_number_of_neurons and
    uses its own seed value/memory position.

    I've created this as a threaded function because it is necessary to iterate through all the liquid's neurons.

    This functions is called inside the SpikingLiquid_Reset.
    */

    int i=0; //Used in the for loops

    int loop_start, loop_end; //

    int *SpkLiq_tid = (int *)args; //Receives the number of the thread (thread_id) from the function input arguments


    // Sets up the loop variables according to the thread number
    if ((*SpkLiq_tid)!=(SpkLiq_threads_N-1)){
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)];
      loop_end = (SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)])*((*SpkLiq_tid)+1);
    }else{ // It means the last thread that could have more work to do because of the remainder
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)-1]; //Uses the anterior value (the rounded one) and multiplies
      loop_end = (SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)-1])*(*SpkLiq_tid)+(SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)]);
    }

    for(i=loop_start; i<loop_end; i++){
        // # RANDOM-0
        // # Membrane initial potentials
        // # Initializes the membrane potentials for all neurons randomly according to (Maass, Natschlager, and Markram 2002)
        // &SpkLiq_threads_seeds[0][(*SpkLiq_tid)]
        SpkLiq_neurons_membrane[i] = SpkLiq_membrane_rand[0] + SpkLiq_uniform_deviate(rand_r(&SpkLiq_threads_seeds[0][(*SpkLiq_tid)])) * (SpkLiq_membrane_rand[1] - SpkLiq_membrane_rand[0]); //in mV 

        SpkLiq_neurons_exc_curr[i] = 0.0; //Initializes the excitatory currents levels for all neurons
        
        SpkLiq_neurons_inh_curr[i] = 0.0; //Initializes the inhibitory currents levels for all neurons
        
        SpkLiq_refrac_timer[i] = 0.0; //Initializes the refractory timers
        
        SpkLiq_test_vthres[i]  = 0; // Initializes all positions with False

        // # RANDOM-1
        // # Noisy offset currents
        // # These are the offset noisy currents according to (Maass, Natschlager, and Markram 2002)
        // &SpkLiq_threads_seeds[1][(*SpkLiq_tid)]
        SpkLiq_noisy_offset_currents[i] = SpkLiq_current_rand[0] + SpkLiq_uniform_deviate(rand_r(&SpkLiq_threads_seeds[1][(*SpkLiq_tid)])) * (SpkLiq_current_rand[1] - SpkLiq_current_rand[0]); //in nA
    }

    return NULL;
}


void SpikingLiquid_Reset(){
    //
    // Resets the initial values of the simulation, but keeps the internal structure of the liquid
    // The use of this function makes possible to restart a simulation without generating again the liquid.
    //
    // I NEED TO CHECK THE STATEMENT BELOW!!!!!
    // The affected variables are:
    // - unsigned int SpkLiq_rand_r_seeds[5]
    // - double SpkLiq_neurons_membrane[SpkLiq_number_of_neurons]
    // - double SpkLiq_neurons_exc_curr[SpkLiq_number_of_neurons]
    // - double SpkLiq_neurons_inh_curr[SpkLiq_number_of_neurons]
    // - double SpkLiq_refrac_timer[SpkLiq_number_of_neurons]
    // - int SpkLiq_test_vthres[SpkLiq_number_of_neurons]
    // - double SpkLiq_noisy_offset_currents[SpkLiq_number_of_neurons]
    // - double SpkLiq_noisy_currents[SpkLiq_number_of_neurons]
    
    int i=0,j=0; //Used in the for loops

    printf("Reseting the simulation...\n");

    // Resets the current time to zero!
    SpkLiq_current_time = 0;

    // Resets the random state variables to their initial values
    for(i=0; i<5; i++){
        SpkLiq_rand_r_seeds[i] = SpkLiq_user_seeds[i];
        // RANDOM-0: Membrane initial potentials
        // RANDOM-1: Noisy offset currents
        // RANDOM-2: Selection of the inhibitory and excitatory neurons
        // RANDOM-3: Internal connections of the liquid
        // RANDOM-4: Noisy corrents

        // Here are generated random seeds for each thread using the user supplied seed as the seed ?!?!? :D
        for(j=0; j<SpkLiq_threads_N; j++){
            SpkLiq_threads_seeds[i][j]=(unsigned int)rand_r(&SpkLiq_rand_r_seeds[i]);
            // each row is a random variable (0...4)
            // the columns are the thread indices (0...SpkLiq_threads_N)

            SpkLiq_thread_id[j]=j; // Initializes the thread ids
        }
        
    }

    for(i=0; i<SpkLiq_threads_N; i++){
        // Initializes the state variables used within the gaussian distribution generation
        // It's necessary only fill the phase with 0 to reset it.      
        SpkLiq_threads_gaussian[i].phase[0]=0;
        SpkLiq_threads_gaussian[i].phase[1]=0;

        // Resets the counters used inside the liquid's connections generation
        SpkLiq_inh_connections_thread[i]=0;
        SpkLiq_inh_connections_thread[i]=0;
    }



//
// MULTITHREADS CODE
//

    // Creates the threads to initialize the neuron's variables
    for (i = 0; i < SpkLiq_threads_N; ++i) {
        if (pthread_create(&SpkLiq_threads_ptrs[i], NULL, SpikingLiquid_Reset_thread, &SpkLiq_thread_id[i])) {
          fprintf(stderr, "error: SpikingLiquid_Reset_thread\n");
          exit(EXIT_FAILURE);
        }
    }
    // Block (join) all threads - the program is stuck here until all threads return
    for (i = 0; i < SpkLiq_threads_N; ++i) {
        pthread_join(SpkLiq_threads_ptrs[i], NULL);
    }

}
        
void SpikingLiquid_init(){
    /*

    Initializes the variables, resets everything (SpikingLiquid_Reset) 
    and creates the liquid (SpikingLiquid_generate_connections_thread).

    This functions needs to be called only at the beginning of the simulations. 
    After the first run, it's possible to only reset (SpikingLiquid_Reset) if
    it's not necessary to generate a new (different) liquid's structure.

    */

    int i=0; //Used in the for loops

    printf("Generating the liquid internal structure...\n");

    // Calculates the total number of neurons based on the liquids shape (x*y*z)
    SpkLiq_number_of_neurons = SpkLiq_net_shape[0]*SpkLiq_net_shape[1]*SpkLiq_net_shape[2];

    // This is going to be used to control the memory positions each thread can access.
    int SpkLiq_number_of_neurons_in_slice_init = SpkLiq_number_of_neurons/SpkLiq_threads_N;

    SpkLiq_number_of_neurons_slice = (int *)malloc(sizeof(int)*SpkLiq_number_of_neurons_in_slice_init);
    // I'm not even testing this because if the system doesn't have memory here, it is going to blow up with the next mallocs...

    // Spreads the liquid's indices among the threads
    for(i=0; i<SpkLiq_threads_N; i++){
        SpkLiq_number_of_neurons_slice[i] = SpkLiq_number_of_neurons_in_slice_init;
    }        

    // Adjusts the amount of "work" each thread should get if the division is not an integer
    if (SpkLiq_number_of_neurons % SpkLiq_threads_N != 0){ // Verifies if there's a remainder
        // In the case of a remainder, makes the last thread "heavier" adding the remainder
        SpkLiq_number_of_neurons_slice[SpkLiq_threads_N-1] = SpkLiq_number_of_neurons_in_slice_init + (SpkLiq_number_of_neurons % SpkLiq_threads_N);
    }




    // Calculates the number of inhibitory and excitatory neurons
    SpkLiq_number_of_inh_neurons = (int) (SpkLiq_number_of_neurons*(SpkLiq_inhibitory_percentage/100.0)); // The int casting is for truncating into an integer
    SpkLiq_number_of_exc_neurons = SpkLiq_number_of_neurons-SpkLiq_number_of_inh_neurons;


    //
    // Memory allocations
    //

    // Here I need to allocate memory for all the dynamic variables acessed by SpikingLiquid_Reset.
    SpkLiq_neurons_membrane = (double *) malloc(sizeof(double)*SpkLiq_number_of_neurons);
    SpkLiq_neurons_exc_curr = (double *) malloc(sizeof(double)*SpkLiq_number_of_neurons);
    SpkLiq_neurons_inh_curr = (double *) malloc(sizeof(double)*SpkLiq_number_of_neurons);
    SpkLiq_refrac_timer = (double *) malloc(sizeof(double)*SpkLiq_number_of_neurons);
    SpkLiq_test_vthres = (int *) malloc(sizeof(int)*SpkLiq_number_of_neurons);
    SpkLiq_noisy_offset_currents = (double *) malloc(sizeof(double)*SpkLiq_number_of_neurons);
    SpkLiq_noisy_currents = (double *) malloc(sizeof(double)*SpkLiq_number_of_neurons);

    // And also the variables used in this function, but that are not going to be reset by SpikingLiquid_Reset
    SpkLiq_inhibitory_indices = (int *) malloc(sizeof(int)*SpkLiq_number_of_inh_neurons);
    SpkLiq_excitatory_indices = (int *) malloc(sizeof(int)*SpkLiq_number_of_exc_neurons);

    SpkLiq_liquid_indices_shuffled = (int *)malloc(sizeof(int)*SpkLiq_number_of_neurons);

    SpkLiq_refrac_values = (double *) malloc(sizeof(double)*SpkLiq_number_of_neurons);

    // Used inside the SpikingLiquid_generate_connections_thread
    SpkLiq_pre_i_thread = malloc(sizeof(int*)*SpkLiq_threads_N);
    SpkLiq_pos_i_thread = malloc(sizeof(int*)*SpkLiq_threads_N);
    SpkLiq_w_i_thread = malloc(sizeof(double*)*SpkLiq_threads_N);
    
    SpkLiq_pre_e_thread = malloc(sizeof(int*)*SpkLiq_threads_N);
    SpkLiq_pos_e_thread = malloc(sizeof(int*)*SpkLiq_threads_N);
    SpkLiq_w_e_thread = malloc(sizeof(double*)*SpkLiq_threads_N);


    // Actually they are reset to zero inside the SpikingLiquid_Reset
    // They are used as counters during the generation of the liquid's connections
    SpkLiq_inh_connections_thread = malloc(sizeof(int)*SpkLiq_threads_N);
    SpkLiq_exc_connections_thread = malloc(sizeof(int)*SpkLiq_threads_N);

    // These are used during the update of the liquid to keep the number of connections each thread will process
    SpkLiq_number_of_inh_connections_thread = malloc(sizeof(int)*SpkLiq_threads_N);
    SpkLiq_number_of_exc_connections_thread = malloc(sizeof(int)*SpkLiq_threads_N);    

    // Tests to verify if the malloc was successful with all the variables
    if (SpkLiq_neurons_membrane==NULL || SpkLiq_neurons_exc_curr==NULL || SpkLiq_neurons_inh_curr==NULL || 
        SpkLiq_refrac_timer==NULL || SpkLiq_test_vthres==NULL ||
        SpkLiq_noisy_offset_currents==NULL || SpkLiq_noisy_currents==NULL || SpkLiq_inhibitory_indices==NULL || SpkLiq_excitatory_indices==NULL || 
        SpkLiq_liquid_indices_shuffled==NULL || SpkLiq_refrac_values==NULL ||
        SpkLiq_pre_i_thread==NULL || SpkLiq_pos_i_thread==NULL || SpkLiq_w_i_thread==NULL || 
        SpkLiq_pre_e_thread==NULL || SpkLiq_pos_e_thread==NULL || SpkLiq_w_e_thread==NULL ||
        SpkLiq_inh_connections_thread==NULL || SpkLiq_exc_connections_thread==NULL){

        fprintf(stderr,"SpikingLiquid_init - Malloc ERROR!");
        exit(EXIT_FAILURE); //I need to check which error I should signal here...
    }

    //
    // These memory allocations should be moved together with the ones above to save some ifs...
    //

    // Allocation of the multiple threads random seeds
    // Each row represents one of the 5 random variables (Random-0...4)
    // Then the columns are the threads indices.
    for (i=0; i<5; i++){
        SpkLiq_threads_seeds[i] = NULL;
        SpkLiq_threads_seeds[i] = (unsigned int *)malloc(sizeof(unsigned int)*SpkLiq_threads_N);
        if (SpkLiq_threads_seeds[i]==NULL){
            fprintf(stderr,"SpkLiq_threads_seeds - columns- malloc ERROR!");
            exit(EXIT_FAILURE);        
        }
    }

    
    // Here I allocate the state variables used with the gaussian distribution (Random-3 and 4)
    SpkLiq_threads_gaussian = (SpkLiq_gaussian_states *)malloc(sizeof(SpkLiq_gaussian_states)*SpkLiq_threads_N);
    if (SpkLiq_threads_gaussian==NULL){
        fprintf(stderr,"SpkLiq_threads_gaussian - malloc ERROR!");
        exit(EXIT_FAILURE);
    }

    // Allocates the memory positions to be used with the threads call
    SpkLiq_threads_ptrs = (pthread_t *)malloc(sizeof(pthread_t)*SpkLiq_threads_N);
    if (SpkLiq_threads_ptrs==NULL){
        fprintf(stderr,"SpkLiq_threads_ptrs - malloc ERROR!");
        exit(EXIT_FAILURE);
    }

    SpkLiq_thread_id = (int *)malloc(sizeof(int)*SpkLiq_threads_N);
    if (SpkLiq_thread_id==NULL){
        fprintf(stderr,"SpkLiq_thread_id - malloc ERROR!");
        exit(EXIT_FAILURE);
    }


    printf("Calling SpikingLiquid_Reset from the SpikingLiquid_init...\n");
    // Initializes the internal variables to start the simulation

    // Initializes (resets) the internal variables to start the simulation
    SpikingLiquid_Reset();


    //
    // Starts the shuffle of the liquid's neuron indices
    // This shuffle is necessary to select randomly the X% of neurons to be inhibitory.
    //
    // I NEED TO THINK MORE ABOUT THE BEST WAY TO SHUFFLE IN A MULTITHREAD WAY BECAUSE
    // RIGHT NOW I'M USING A SERIAL ALGORITHM THAT IS A BIT HARD TO MAKE PARALLEL...
    //
    for(i=0; i<SpkLiq_number_of_neurons; i++){
        SpkLiq_liquid_indices_shuffled[i] = i; //Fills with integers from 0 to (SpkLiq_number_of_neurons-1) to be shuffled
    }

    //# RANDOM-2
    //# Selection of the inhibitory and excitatory neurons
    // Knuth-Fisher-Yates shuffle algorithm - from http://blog.codinghorror.com/the-danger-of-naivete/
    // Creates a shuffled array with all the liquid's indices
    for(i=SpkLiq_number_of_neurons - 1; i > 0; i--){
        // Generates a random integer from 0 to i+1
        int n = (int) (SpkLiq_uniform_deviate(rand_r(&SpkLiq_rand_r_seeds[2])) * (i+1));
        // I can use the &SpkLiq_rand_r_seeds[2] because no other part of the code is going to use it.

        int temp;

        // Swaps the current position 'i' with the random position 'n'
        temp = SpkLiq_liquid_indices_shuffled[i], 
        SpkLiq_liquid_indices_shuffled[i] = SpkLiq_liquid_indices_shuffled[n];
        SpkLiq_liquid_indices_shuffled[n] = temp;
    }


    for(i=0; i<SpkLiq_number_of_inh_neurons; i++){
        //Populates the SpkLiq_inhibitory_indices with the first N=SpkLiq_number_of_inh_neurons neurons from the shuffled liquid
        SpkLiq_inhibitory_indices[i] = SpkLiq_liquid_indices_shuffled[i];

        //Populates the refractory_vetor with inhibitory values
        SpkLiq_refrac_values[SpkLiq_inhibitory_indices[i]] = SpkLiq_refractory[1]; // Sets the refractory periods according to the type of neuron
    }


    for(i=SpkLiq_number_of_inh_neurons; i<SpkLiq_number_of_neurons; i++){
        //Populates the SpkLiq_excitatory_indices with the neurons that are not inhibitory
        SpkLiq_excitatory_indices[(i-SpkLiq_number_of_inh_neurons)] = SpkLiq_liquid_indices_shuffled[i];

        //Populates the refractory_vetor with excitatory values
        SpkLiq_refrac_values[SpkLiq_excitatory_indices[(i-SpkLiq_number_of_inh_neurons)]] = SpkLiq_refractory[0]; // Sets the refractory periods according to the type of neuron
    }


// DEBUG TIME
    double time_begin, time_end;
    // Gets the initial time in nanoseconds
    time_begin=clock_my_gettime();


    printf("Generating liquid's connections...\n");

//
// MULTITHREADS CODE
//

    // Creates the threads to calculate the probabilities based on the distance between neurons
    // and generate the connections between the inhibitory and excitatoty neurons
    for (i = 0; i < SpkLiq_threads_N; ++i) {
        if (pthread_create(&SpkLiq_threads_ptrs[i], NULL, SpikingLiquid_generate_connections_thread, &SpkLiq_thread_id[i])) {
          fprintf(stderr, "error: SpikingLiquid_generate_connections_thread\n");
          exit(EXIT_FAILURE);
        }
    }
    // Block (join) all threads - the program is stuck here until all threads return
    for (i = 0; i < SpkLiq_threads_N; ++i) {
        pthread_join(SpkLiq_threads_ptrs[i], NULL);
    }


// DEBUG TIME
    // Gets the initial time in nanoseconds
    time_end=clock_my_gettime();
    printf("SpikingLiquid_generate_connections time:%f\n",(time_end-time_begin)/1E6); // in milliseconds



// DEBUG TIME
    // Gets the initial time in nanoseconds
    time_begin=clock_my_gettime();


    printf("Copying connections...\n");

    // Groups the connections generated by each thread in a single array (for inhibitory and excitatory separately)
    SpikingLiquid_generate_connections_memory(); // Copy the memory content is WAY FASTER than the generation of connections!


// DEBUG TIME
    // Gets the initial time in nanoseconds
    time_end=clock_my_gettime();
    printf("SpikingLiquid_generate_connections_memory time:%f\n",(time_end-time_begin)/1E6); // in milliseconds


    // 
    // Initializes the arrays that will be used during the update with the threads
    //

    SpkLiq_number_of_inh_neurons_thread = (int *)malloc(sizeof(int)*SpkLiq_threads_N);
    SpkLiq_number_of_exc_neurons_thread = (int *)malloc(sizeof(int)*SpkLiq_threads_N);

    if (SpkLiq_number_of_inh_neurons_thread==NULL || SpkLiq_number_of_exc_neurons_thread==NULL){
        fprintf(stderr, "SpkLiq_number_of_(inh/exc)_neurons_thread - malloc error!\n");
        exit(EXIT_FAILURE);   
    }



    int SpkLiq_inh_connections_slice = SpkLiq_inh_connections/SpkLiq_threads_N;
    int SpkLiq_exc_connections_slice = SpkLiq_exc_connections/SpkLiq_threads_N;

    int SpkLiq_number_of_inh_neurons_slice = SpkLiq_number_of_inh_neurons/SpkLiq_threads_N;
    int SpkLiq_number_of_exc_neurons_slice = SpkLiq_number_of_exc_neurons/SpkLiq_threads_N;

    // Spreads the liquid's SpkLiq_inh_connections among the threads
    for(i=0; i<SpkLiq_threads_N; i++){
        // Here I define how many connections each thread is going to deal with
        // ATTENTION: I'M REUSING THESE TWO GLOBAL VARIABLES BELOW!!!!
        SpkLiq_number_of_inh_connections_thread[i] = SpkLiq_inh_connections_slice;
        SpkLiq_number_of_exc_connections_thread[i] = SpkLiq_exc_connections_slice;

        // And here how many neurons (inhibitory / excitatory) each thread is going to receive
        SpkLiq_number_of_inh_neurons_thread[i] = SpkLiq_number_of_inh_neurons_slice;
        SpkLiq_number_of_exc_neurons_thread[i] = SpkLiq_number_of_exc_neurons_slice;
    }        

    
    if (SpkLiq_inh_connections % SpkLiq_threads_N != 0){ // Verifies if there's a remainder
        // In the case of a remainder, makes the last thread heavier adding the remainder
        SpkLiq_number_of_inh_connections_thread[SpkLiq_threads_N-1] = SpkLiq_inh_connections_slice + (SpkLiq_inh_connections % SpkLiq_threads_N);
    }

    if (SpkLiq_exc_connections % SpkLiq_threads_N != 0){ // Verifies if there's a remainder
        // In the case of a remainder, makes the last thread heavier adding the remainder
        SpkLiq_number_of_exc_connections_thread[SpkLiq_threads_N-1] = SpkLiq_exc_connections_slice + (SpkLiq_exc_connections % SpkLiq_threads_N);
    }

    if (SpkLiq_number_of_inh_neurons % SpkLiq_threads_N != 0){ // Verifies if there's a remainder
        // In the case of a remainder, makes the last thread heavier adding the remainder
        SpkLiq_number_of_inh_neurons_thread[SpkLiq_threads_N-1] = SpkLiq_number_of_inh_neurons_slice + (SpkLiq_number_of_inh_neurons % SpkLiq_threads_N);
    }

    if (SpkLiq_number_of_exc_neurons % SpkLiq_threads_N != 0){ // Verifies if there's a remainder
        // In the case of a remainder, makes the last thread heavier adding the remainder
        SpkLiq_number_of_exc_neurons_thread[SpkLiq_threads_N-1] = SpkLiq_number_of_exc_neurons_slice + (SpkLiq_number_of_exc_neurons % SpkLiq_threads_N);
    }

    
    // These arrays are going to receive the variables addresses from SpikingLiquid_process_internal_spikes_threads
    // They work like matrices (array of arrays) being the thread_id the first index.
    SpkLiq_receive_spike_i_thread = (int **)malloc(sizeof(int*)*SpkLiq_threads_N);
    SpkLiq_receive_spike_i_w_thread = (double **)malloc(sizeof(double*)*SpkLiq_threads_N);
    SpkLiq_receive_spike_i_idx_thread = (int *)malloc(sizeof(int)*SpkLiq_threads_N);

    SpkLiq_receive_spike_e_thread = (int **)malloc(sizeof(int*)*SpkLiq_threads_N);
    SpkLiq_receive_spike_e_w_thread = (double **)malloc(sizeof(double*)*SpkLiq_threads_N);
    SpkLiq_receive_spike_e_idx_thread = (int *)malloc(sizeof(int)*SpkLiq_threads_N);

    if (SpkLiq_receive_spike_i_thread==NULL || SpkLiq_receive_spike_i_w_thread==NULL || SpkLiq_receive_spike_i_idx_thread==NULL ||
        SpkLiq_receive_spike_e_thread==NULL || SpkLiq_receive_spike_e_w_thread==NULL || SpkLiq_receive_spike_e_idx_thread==NULL){
        fprintf(stderr, "SpkLiq_receive_spike_ * _thread - malloc error!\n");
        exit(EXIT_FAILURE);   
    }


    // These are the arrays each thread are going to use (from the matrices above).
    // Because I'm defining the maximum number of liquid's connections each thread is going to deal with, I can allocate
    // the memory here. According to the liquid's activation level, each thread needs to indicate how much space it used
    // at every update step.
    for(i=0; i<SpkLiq_threads_N; i++){
        SpkLiq_receive_spike_i_thread[i] = NULL;
        SpkLiq_receive_spike_i_w_thread[i] = NULL;

        SpkLiq_receive_spike_i_thread[i] = (int *)malloc(sizeof(int)*SpkLiq_number_of_inh_connections_thread[i]);
        SpkLiq_receive_spike_i_w_thread[i] = (double *)malloc(sizeof(double)*SpkLiq_number_of_inh_connections_thread[i]);

        if (SpkLiq_receive_spike_i_thread[i]==NULL || SpkLiq_receive_spike_i_w_thread[i]==NULL){
            fprintf(stderr, "SpkLiq_receive_spike_i*_thread[i] - malloc error!\n");
            exit(EXIT_FAILURE);
        }

        SpkLiq_receive_spike_e_thread[i] = NULL;
        SpkLiq_receive_spike_e_w_thread[i] = NULL;

        SpkLiq_receive_spike_e_thread[i] = (int *)malloc(sizeof(int)*SpkLiq_number_of_exc_connections_thread[i]);
        SpkLiq_receive_spike_e_w_thread[i] = (double *)malloc(sizeof(double)*SpkLiq_number_of_exc_connections_thread[i]);

        if (SpkLiq_receive_spike_e_thread[i]==NULL || SpkLiq_receive_spike_e_w_thread[i]==NULL){
            fprintf(stderr, "SpkLiq_receive_spike_e*_thread[i] - malloc error!\n");
            exit(EXIT_FAILURE);
        }

    }


    printf("Total number of neurons: %d\n", SpkLiq_number_of_neurons);
    printf("Total number of INH neurons: %d\n", SpkLiq_number_of_inh_neurons);
    printf("Total number of EXC neurons: %d\n", SpkLiq_number_of_exc_neurons);
    printf("Number of connections INH=>: %d\n", SpkLiq_inh_connections);
    printf("Number of connections EXC=>: %d\n", SpkLiq_exc_connections);
    printf("Total number of connections: %d\n", SpkLiq_total_number_of_connections);
      
}



//time_seed and uniform_deviate functions came from:
//http://eternallyconfuzzled.com/arts/jsw_art_rand.aspx
unsigned SpkLiq_time_seed(){
     time_t now = time ( 0 );
     unsigned char *p = (unsigned char *)&now;
     unsigned seed = 0;
     size_t i;

    for (i = 0; i < sizeof now; i++ )
        seed = seed * ( UCHAR_MAX + 2U ) + p[i];

    return seed;
}


double SpkLiq_uniform_deviate(int seed){
    /*
    Tries to minimize the problem of a not 
    very uniform distribution generated using rand()
    Examples:
    int r = SpkLiq_uniform_deviate( rand() ) * N; // => (0,N]
    int r = M + SpkLiq_uniform_deviate( rand() ) * ( N - M ); // => (M,N]
    */
    return seed * ( 1.0 / ( RAND_MAX + 1.0 ) );
} 



double SpkLiq_gaussrand_threads(unsigned int *seed, int *phase, double *V1, double *V2, double *S){
    /*
    Multithread SAFE version of the gaussian generator
    In this version, instead of using static variables inside the function, the variables are supplied as parametres

    Gaussian generator algorithm from: 
            http://c-faq.com/lib/gaussian.html
            Marsaglia and Bray, "A Convenient Method for Generating Normal Variables"
            http://en.wikipedia.org/wiki/Marsaglia_polar_method
            Abramowitz and Stegun, Handbook of Mathematical Functions
    Expect value: 0
    Standard deviation: 1

    The functions (re)initializes when *phase==0. Therefore the value *seed is going to be the initial seed.
    */

    double X;


    if(*phase == 0) {
        do {
            double U1 = SpkLiq_uniform_deviate(rand_r(seed));
            double U2 = SpkLiq_uniform_deviate(rand_r(seed));

            *V1 = 2 * U1 - 1;
            *V2 = 2 * U2 - 1;
            *S = *V1 * (*V1) + (*V2) * (*V2);
            } while(*S >= 1 || *S == 0);

        X = *V1 * sqrt(-2 * log(*S) / (*S));
    } else
        X = *V2 * sqrt(-2 * log(*S) / (*S));

    *phase = 1 - *phase; // This line makes the phase oscilate jumping between 0 and 1.

    return X;
}



//
// I NEED TO CHECK IF IT'S BEST TO CAST TO DOUBLE OR USE A DOUBLE VARIABLE!!!
//

double SpkLiq_euclidean_dist(int i, int j, int xd, int yd){
    /*
    Calculates the euclidean distance between two neurons inside the liquid

    Because the neurons are stored in a 1D array, it is necessary a bit of math to convert
    from that array index to the X,Y and Z positions.

    I'm assuming that in the 3D structure the values i and j are equivalent 
    to the incrementing of x (until complete the line) then y and (after complete xy plane) z values.
    xd, yd are the x and y dimensions of the liquid (shape)

    ***********************************************************************************************
    This function costs me a lot of multiplications and also the sqrt and pow calls.
    One possible improvement would be simplify this calculations, even with some errors, because
    the results are used with a probabilistic comparison after all.
    ***********************************************************************************************

    */

    // I did some simplifications here. Instead of doing a proper calculation of the distance, I'm
    // trying to do an approximated one.

    int xy = xd*yd; // How many neurons I have in a complete XY plane (think as a liquid's slice)

    // Verify how many XY planes are filled, because this is equivalent to the Z values
    int zi = (i/xy); //the integer cast is used here to round down the values. 
    int zj = (j/xy); //the integer cast is used here to round down the values. 


    int yi = (i-zi*xy)/xd; 
    int yj = (j-zj*xy)/xd;

    int xi = (i-yi*zi*xy); //the integer cast is used here to round down the values. 
    int xj = (j-yj*zj*xy); //the integer cast is used here to round down the values. 

    // It's necessary to centralize the axis, otherwise we are going to have a bias
    xi = -xi*0.5;   // Centralize in the x=0 axis
    xj = -xj*0.5;   // Centralize in the x=0 axis 

    zi = -zi*0.5;   // Centralize in the z=0 axis
    zj = -zj*0.5;   // Centralize in the z=0 axis    

    return sqrt(pow((xi-xj),2)+pow((yi-yj),2)+pow((zi-zj),2));
}




void SpkLiq_process_exc_spikes(int *spikes, double *weights, int number_of_spikes){
    /* 
    Processes the received spikes at the current time
    spikes: list with the indexes of the neurons who spiked.
    weights: the weights to the neurons who spiked

    THIS IS NOT THREAD SAFE BECAUSE A NEURON CAN RECEIVE MORE THAN ONE SPIKE AT EACH STEP
    */

    for(int i=0; i<number_of_spikes; i++){
        SpkLiq_neurons_exc_curr[spikes[i]] +=  weights[i]; // Adds to the ones that received spikes
    }
}


void SpkLiq_process_inh_spikes(int *spikes, double *weights, int number_of_spikes){
    /*
    Processes the received spikes at the current time
    spikes: list with the indexes of the neurons who spiked.
    weights: the weights to the neurons who spiked        
    
    THIS IS NOT THREAD SAFE BECAUSE A NEURON CAN RECEIVE MORE THAN ONE SPIKE AT EACH STEP    
    */

    for(int i=0; i<number_of_spikes; i++){
        SpkLiq_neurons_inh_curr[spikes[i]] +=  weights[i]; // Adds to the ones that received spikes
    }
}


void *SpikingLiquid_process_internal_spikes_threads(void *args){
    /*
    Tests if the liquid's neurons spiked
    This function is called every simulation SpkLiq_step, so it's important to optimize for speed!
    */

    int i=0,j=0; //Used in the for loops

    int loop_start, loop_end;

    int *SpkLiq_tid = (int *)args; //Receives the number of the thread (thread_id) from the function input arguments


    // Checks which internal INHIBITORY neurons spiked
    // This is accomplished checking against the SpkLiq_test_vthres array (length: SpkLiq_number_of_neurons)
    // SpkLiq_test_vthres indicates with 1 which neurons inside the liquid spiked in the last simulation SpkLiq_step
    int SpkLiq_spiked_i[SpkLiq_number_of_inh_neurons_thread[*SpkLiq_tid]]; // Stores which inhibitory neuron spiked or not
                                         // It's NOT necessary initialize the array because the SpkLiq_spiked_i_idx
                                         // controls the maximum valid index.

    // Sets up the loop variables according to the thread number
    if ((*SpkLiq_tid)!=(SpkLiq_threads_N-1)){
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_inh_neurons_thread[(*SpkLiq_tid)];
      loop_end = (SpkLiq_number_of_inh_neurons_thread[(*SpkLiq_tid)])*((*SpkLiq_tid)+1);
    }else{ // It means the last thread that could have more work to do because of the remainder
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_inh_neurons_thread[(*SpkLiq_tid)-1]; //Uses the anterior value (the rounded one) and multiplies
      loop_end = (SpkLiq_number_of_inh_neurons_thread[(*SpkLiq_tid)-1])*(*SpkLiq_tid)+(SpkLiq_number_of_inh_neurons_thread[(*SpkLiq_tid)]);
    }

    int SpkLiq_spiked_i_idx = 0; // Counts how many inhibitory neurons spiked
    // for(i=(*SpkLiq_tid)*SpkLiq_number_of_inh_neurons_thread[*SpkLiq_tid]; i<(*SpkLiq_tid+1)*SpkLiq_number_of_inh_neurons_thread[*SpkLiq_tid]; i++){
    for(i=loop_start; i<loop_end; i++){      
        //SpkLiq_test_vthres has the size of SpkLiq_number_of_neurons and has the value one
        //if the neuron spiked in the last SpkLiq_step
        if(SpkLiq_test_vthres[SpkLiq_inhibitory_indices[i]]){ // The next two lines are excuted only for neurons that spiked
            SpkLiq_spiked_i[SpkLiq_spiked_i_idx] = SpkLiq_inhibitory_indices[i]; // Saves the index of the neuron who spiked
            SpkLiq_spiked_i_idx++;
        }
    } // At the end of this loop, SpkLiq_spiked_i has which inhibitory neuron spiked, and the total is SpkLiq_spiked_i_idx
    

    // Sets up the loop variables according to the thread number
    if ((*SpkLiq_tid)!=(SpkLiq_threads_N-1)){
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_inh_connections_thread[(*SpkLiq_tid)];
      loop_end = (SpkLiq_number_of_inh_connections_thread[(*SpkLiq_tid)])*((*SpkLiq_tid)+1);
    }else{ // It means the last thread that could have more work to do because of the remainder
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_inh_connections_thread[(*SpkLiq_tid)-1]; //Uses the anterior value (the rounded one) and multiplies
      loop_end = (SpkLiq_number_of_inh_connections_thread[(*SpkLiq_tid)-1])*(*SpkLiq_tid)+(SpkLiq_number_of_inh_connections_thread[(*SpkLiq_tid)]);
    }

    // Now I need to check which neurons are going to receive these INHIBITORY spikes
    // Liquid's inhibitory neurons are connected to the neurons indicated in the array SpkLiq_pos_i
    // The arrays SpkLiq_pre_i, SpkLiq_pos_i and SpkLiq_w_i always have the same lenght: SpkLiq_number_of_inh_connections_thread[*SpkLiq_tid]
    SpkLiq_receive_spike_i_idx_thread[*SpkLiq_tid] = 0;
    // for(i=(*SpkLiq_tid)*SpkLiq_number_of_inh_connections_thread[*SpkLiq_tid]; i<(*SpkLiq_tid+1)*SpkLiq_number_of_inh_connections_thread[*SpkLiq_tid]; i++) { //iterates through the SpkLiq_pre_i array
    for(i=loop_start; i<loop_end; i++) { //iterates through the SpkLiq_pre_i array      
      for (j=0;j<SpkLiq_spiked_i_idx;j++){ //iterates through the SpkLiq_spiked_i array (can be smaller than SpkLiq_number_of_inh_neurons!)
//This loop increases/decreases according to the number of spikes!!! 
        if (SpkLiq_pre_i[i]==SpkLiq_spiked_i[j]){ // Checks for neurons who spiked
            SpkLiq_receive_spike_i_thread[*SpkLiq_tid][SpkLiq_receive_spike_i_idx_thread[*SpkLiq_tid]] = SpkLiq_pos_i[i];
            SpkLiq_receive_spike_i_w_thread[*SpkLiq_tid][SpkLiq_receive_spike_i_idx_thread[*SpkLiq_tid]] = SpkLiq_w_i[i];
            SpkLiq_receive_spike_i_idx_thread[*SpkLiq_tid]++;
        }
      }
    } // At the end of this loop, SpkLiq_receive_spike_i has which neuron should receive a spike from INHIBITORY neurons, 
      // and the total is SpkLiq_receive_spike_e_idx_thread[*SpkLiq_tid].
      //
      // ATTENTION: one neuron can receive more than one spike at each simulation SpkLiq_step.
      //

    // Sets up the loop variables according to the thread number
    if ((*SpkLiq_tid)!=(SpkLiq_threads_N-1)){
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_exc_neurons_thread[(*SpkLiq_tid)];
      loop_end = (SpkLiq_number_of_exc_neurons_thread[(*SpkLiq_tid)])*((*SpkLiq_tid)+1);
    }else{ // It means the last thread that could have more work to do because of the remainder
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_exc_neurons_thread[(*SpkLiq_tid)-1]; //Uses the anterior value (the rounded one) and multiplies
      loop_end = (SpkLiq_number_of_exc_neurons_thread[(*SpkLiq_tid)-1])*(*SpkLiq_tid)+(SpkLiq_number_of_exc_neurons_thread[(*SpkLiq_tid)]);
    }

    // Checks which internal EXCITATORY neurons spiked
    // This is accomplished checking against the SpkLiq_test_vthres array (length: SpkLiq_number_of_neurons)
    // SpkLiq_test_vthres indicates with 1 which neurons inside the liquid spiked in the last simulation SpkLiq_step
    int SpkLiq_spiked_e[SpkLiq_number_of_exc_neurons_thread[*SpkLiq_tid]]; // Stores which excitatory neuron spiked or not    
                                         // It's NOT necessary initialize the array because the SpkLiq_spiked_e_idx
                                         // controls the maximum valid index.    
    int SpkLiq_spiked_e_idx = 0; // Counts how many excitatory neurons spiked
    // for(i=(*SpkLiq_tid)*SpkLiq_number_of_exc_neurons_thread[*SpkLiq_tid]; i<(*SpkLiq_tid+1)*SpkLiq_number_of_exc_neurons_thread[*SpkLiq_tid]; i++){
    for(i=loop_start; i<loop_end; i++){      
        //SpkLiq_test_vthres has the size of SpkLiq_number_of_neurons and has the value one
        //if the neuron spiked in the last SpkLiq_step
        if(SpkLiq_test_vthres[SpkLiq_excitatory_indices[i]]){
            SpkLiq_spiked_e[SpkLiq_spiked_e_idx] = SpkLiq_excitatory_indices[i];
            SpkLiq_spiked_e_idx++;
        }
    } // At the end of this loop, SpkLiq_spiked_i has which excitatory neurons spiked, and the total is SpkLiq_spiked_e_idx

    
    // Sets up the loop variables according to the thread number
    if ((*SpkLiq_tid)!=(SpkLiq_threads_N-1)){
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_exc_connections_thread[(*SpkLiq_tid)];
      loop_end = (SpkLiq_number_of_exc_connections_thread[(*SpkLiq_tid)])*((*SpkLiq_tid)+1);
    }else{ // It means the last thread that could have more work to do because of the remainder
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_exc_connections_thread[(*SpkLiq_tid)-1]; //Uses the anterior value (the rounded one) and multiplies
      loop_end = (SpkLiq_number_of_exc_connections_thread[(*SpkLiq_tid)-1])*(*SpkLiq_tid)+(SpkLiq_number_of_exc_connections_thread[(*SpkLiq_tid)]);
    }    

    // Now I need to check which neurons are going to receive these EXCITATORY spikes
    // Liquid's excitatory neurons are connected to the neurons indicated in the array SpkLiq_pos_e
    // The arrays SpkLiq_pre_e, SpkLiq_pos_e and SpkLiq_w_e always have the same lenght: SpkLiq_number_of_exc_connections_thread[*SpkLiq_tid]
    SpkLiq_receive_spike_e_idx_thread[*SpkLiq_tid] = 0;
    // for(i=(*SpkLiq_tid)*SpkLiq_number_of_exc_connections_thread[*SpkLiq_tid]; i<(*SpkLiq_tid+1)*SpkLiq_number_of_exc_connections_thread[*SpkLiq_tid]; i++) { //iterates through the SpkLiq_pre_e array
    for(i=loop_start; i<loop_end; i++) { //iterates through the SpkLiq_pre_e array
      for (j=0;j<SpkLiq_spiked_e_idx;j++){ //iterates through the SpkLiq_spiked_e array (can be smaller than SpkLiq_number_of_exc_neurons!)
        if (SpkLiq_pre_e[i]==SpkLiq_spiked_e[j]){ // Checks for neurons who spiked
            SpkLiq_receive_spike_e_thread[*SpkLiq_tid][SpkLiq_receive_spike_e_idx_thread[*SpkLiq_tid]] = SpkLiq_pos_e[i];
            SpkLiq_receive_spike_e_w_thread[*SpkLiq_tid][SpkLiq_receive_spike_e_idx_thread[*SpkLiq_tid]] = SpkLiq_w_e[i];
            SpkLiq_receive_spike_e_idx_thread[*SpkLiq_tid]++;
        }
      }
    } // At the end of this loop, SpkLiq_receive_spike_e has which neurons should receive a spike from EXCITATORY neurons, 
      // and the total is *SpkLiq_receive_spike_e_idx_thread[*SpkLiq_tid].
      // ATTENTION: one neuron can receive more than one spike at each simulation SpkLiq_step.

    return NULL;
}


void *SpikingLiquid_update_internal_thread(void *args){
  /*
    Updates the values of the membrane voltage after all the currents had been processed.
  */
    int i=0; //Used in the for loops
    int *SpkLiq_tid = (int *)args; //Receives the number of the thread (thread_id) from the function input arguments

    int loop_start, loop_end;

    // Sets up the loop variables according to the thread number
    if ((*SpkLiq_tid)!=(SpkLiq_threads_N-1)){
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)];
      loop_end = (SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)])*((*SpkLiq_tid)+1);
    }else{ // It means the last thread that could have more work to do because of the remainder
      loop_start = (*SpkLiq_tid)*SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)-1]; //Uses the anterior value (the rounded one) and multiplies
      loop_end = (SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)-1])*(*SpkLiq_tid)+(SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)]);
    }    

    // for(i=(*SpkLiq_tid)*SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)]; i<SpkLiq_number_of_neurons_slice[(*SpkLiq_tid)]*((*SpkLiq_tid)+1); i++){
    for(i=loop_start; i<loop_end; i++){

        // Generates the new values to the noisy currents
        // # RANDOM-4
        // # Noisy corrents
        SpkLiq_noisy_currents[i] = SpkLiq_gaussrand_threads(&SpkLiq_threads_seeds[0][*SpkLiq_tid], &SpkLiq_threads_gaussian[*SpkLiq_tid].phase[1], &SpkLiq_threads_gaussian[*SpkLiq_tid].V1[1], &SpkLiq_threads_gaussian[*SpkLiq_tid].V2[1], &SpkLiq_threads_gaussian[*SpkLiq_tid].S[1])*SpkLiq_noisy_current_rand; // Variable noisy currents

        // # After the currents, it is necessary to update the membrane voltage
        // # Integrating the function:
        // dv/dt  = (ie + ii + i_offset + i_inj)/c_m + (v_rest-v)/tau_m
        if(SpkLiq_refrac_timer[i]>0){
            SpkLiq_neurons_membrane[i] = SpkLiq_vresets; // Resets the membrane of the refractory ones
        }else{ //If the membrane did not reset, then it is necessary to integrate
            SpkLiq_neurons_membrane[i] += ((SpkLiq_neurons_exc_curr[i] + SpkLiq_neurons_inh_curr[i] + SpkLiq_noisy_offset_currents[i] + SpkLiq_noisy_currents[i])/SpkLiq_cm + (SpkLiq_vrest-SpkLiq_neurons_membrane[i])/SpkLiq_taum)*SpkLiq_step;
        }

        // After the integration, check if the neuron spiked by comparing its the membrane voltage with the threshold
        if(SpkLiq_neurons_membrane[i]>SpkLiq_vthres){
            SpkLiq_test_vthres[i] = 1; //# Verifies who should spike and creates a boolean vector with this information
            SpkLiq_neurons_membrane[i] = SpkLiq_vresets; //# Resets the membrane of the spiked ones
            SpkLiq_refrac_timer[i] = SpkLiq_refrac_values[i]; //# Resets the refractory timer to the ones who spiked            
        }else{
//NEW
            SpkLiq_test_vthres[i] = 0; //# Verifies who should spike and creates a boolean vector with this information
        }

//
//NEW - I don't know if this is necessary, but if the simulation goes on during a long period of time this could overflow.
//        
        if (SpkLiq_refrac_timer[i]<0)
                  SpkLiq_refrac_timer[i]=0;

        // HERE THERE IS A POSSIBLE BUG:
        // if the neuron NEVER spikes the refrac timer can count down until blow up the variable...
        SpkLiq_refrac_timer[i] -= SpkLiq_step; //# Subtracts the refractory timer to enable the refractory calculation
    }
    return NULL;
}


void SpikingLiquid_update(int *spikes_exc, int *spikes_inh, double *weights_exc, double *weights_inh, int size_exc, int size_inh){
    /*
    Updates the simulation, running one time SpkLiq_step, and processes the received excitatory and inhibitory spikes.
    
    SpikingLiquid_update(spikes_exc, spikes_inh, weights_exc, weights_inh):
    spikes_exc: list/array with the index of the neurons receiving excitatory spikes
    spikes_inh: list/array with the index of the neurons receiving inhibitory spikes
    weights_exc: list/array with the weights in the same order as spikes_exc
    weights_inh: list/array with the weights in the same order as spikes_inh        
    
    The first update is time=0s

    I need to verify if these functions are thread safe:
        SpikingLiquid_process_internal_spikes
        SpkLiq_process_inh_spikes
        SpkLiq_process_exc_spikes

    In order to optimize this function, I think it needs a static variable to control
    the threads initialization, so the overhead occurs only during the first call.

    */

    int i=0; //Used in the for loops

// THIS IS FAST ENOUGH
    // Solves the diff equation for the currents ( die(t)/dt=-ie(t)/taue and dii(t)/dt=-ii(t)/taui )
    // The membrane depends on these two currents, so they must be solved before the membrane.
    // These updates also must occur BEFORE the processing of the spikes, otherwise the spikes are going 
    // to be reduced just like when one time SpkLiq_step had passed.
    for(i=0; i<SpkLiq_number_of_neurons; i++){
        SpkLiq_neurons_inh_curr[i] += (-SpkLiq_neurons_inh_curr[i]/SpkLiq_taui)*SpkLiq_step; // Updates the INHIBITORY current values
                                                                                             // integrating the function dii/dt = -ii/SpkLiq_taui
        SpkLiq_neurons_exc_curr[i] += (-SpkLiq_neurons_exc_curr[i]/SpkLiq_taue)*SpkLiq_step; // Updates the EXCITATORY current values
                                                                                             // integrating the function die/dt = -ie/taue
    }




//
// MULTITHREADS CODE
//

    // Creates the threads to processes the internal spikes
    for (i = 0; i < SpkLiq_threads_N; ++i) {
        if (pthread_create(&SpkLiq_threads_ptrs[i], NULL, SpikingLiquid_process_internal_spikes_threads, &SpkLiq_thread_id[i])) {
          fprintf(stderr, "error: SpikingLiquid_process_internal_spikes_threads\n");
          exit(EXIT_FAILURE);
        }
    }
    // Block (join) all threads - the program is stuck here until all threads return
    for (i = 0; i < SpkLiq_threads_N; ++i) {
        pthread_join(SpkLiq_threads_ptrs[i], NULL);
    }



// SERIAL
    // Processes the information generated by the threads
    // This cannot be done in parallel because some neurons can receive more than one spike and 
    // then the threads could try to update the same synapse current at the same time!
    for(i=0; i<SpkLiq_threads_N; i++){
        SpkLiq_process_inh_spikes(SpkLiq_receive_spike_i_thread[i], SpkLiq_receive_spike_i_w_thread[i], SpkLiq_receive_spike_i_idx_thread[i]);
        SpkLiq_process_exc_spikes(SpkLiq_receive_spike_e_thread[i], SpkLiq_receive_spike_e_w_thread[i], SpkLiq_receive_spike_e_idx_thread[i]);
    }


// SERIAL
    // Processes the EXTERNAL spikes (inhibitory and excitatory)
    if(size_inh!=0)
        SpkLiq_process_inh_spikes(spikes_inh, weights_inh, size_inh);

    if(size_exc!=0)
        SpkLiq_process_exc_spikes(spikes_exc, weights_exc, size_exc);


//
// MULTITHREADS CODE
//

    // Creates the threads to processes the membrane, noisy currents, etc
    for(i = 0; i < SpkLiq_threads_N; ++i) {
        if (pthread_create(&SpkLiq_threads_ptrs[i], NULL, SpikingLiquid_update_internal_thread, &SpkLiq_thread_id[i])) {
          fprintf(stderr, "error: SpikingLiquid_update_internal_thread\n");
          exit(EXIT_FAILURE);
        }
    }
    // Block (join) all threads - the program is stuck here until all threads return
    for(i = 0; i < SpkLiq_threads_N; ++i) {
        pthread_join(SpkLiq_threads_ptrs[i], NULL);
    }


    //# If I update the time here makes it possible to generate things at time zero!
    SpkLiq_current_time += SpkLiq_step; //# Advances the simulation time one SpkLiq_step
}


// ##########################################################################################
// ##########################################################################################
// ##########################################################################################
// ##########################################################################################
// ##########################################################################################


int user_setup(
                int *_SpkLiq_net_shape,
                double _SpkLiq_lbd_value,
                double _SpkLiq_step,
                double _SpkLiq_taum,
                double _SpkLiq_cm,
                double _SpkLiq_taue,
                double _SpkLiq_taui,
                double *_SpkLiq_membrane_rand,
                double *_SpkLiq_current_rand,
                double _SpkLiq_noisy_current_rand,
                double _SpkLiq_vresets,
                double _SpkLiq_vthres,
                double _SpkLiq_vrest,
                double *_SpkLiq_refractory,
                double _SpkLiq_inhibitory_percentage,
                unsigned int *my_seeds,
                int _SpkLiq_threads_N
                ){
    //
    // USER PARAMETRES
    //

    // Network user parametres
    SpkLiq_net_shape[0] = _SpkLiq_net_shape[0];//15;
    SpkLiq_net_shape[1] = _SpkLiq_net_shape[1];//30;
    SpkLiq_net_shape[2] = _SpkLiq_net_shape[2];//3;

    SpkLiq_lbd_value = _SpkLiq_lbd_value;//1.2;
    SpkLiq_step = _SpkLiq_step;//0.2E-3;
    SpkLiq_taum = _SpkLiq_taum;//30E-3;
    SpkLiq_cm = _SpkLiq_cm;//30E-9;
    SpkLiq_taue = _SpkLiq_taue;//3E-3;
    SpkLiq_taui = _SpkLiq_taui;//6E-3;

    SpkLiq_membrane_rand[0] = _SpkLiq_membrane_rand[0];//13.5E-3;
    SpkLiq_membrane_rand[1] = _SpkLiq_membrane_rand[1];//15E-3; //range of the possible values to the initialization of the membrane voltage
    

    SpkLiq_current_rand[0] = _SpkLiq_current_rand[0];//14.975E-9;
    SpkLiq_current_rand[1] = _SpkLiq_current_rand[1];//15.025E-9; //range of the possible values to the initialization of the constant noisy offset current

    SpkLiq_noisy_current_rand = _SpkLiq_noisy_current_rand;//0.2E-9; //max range of the possible values to the initialization of the variable noisy current    

    SpkLiq_vresets = _SpkLiq_vresets;//13.5E-3;
    SpkLiq_vthres = _SpkLiq_vthres;//15E-3;
    SpkLiq_vrest = _SpkLiq_vrest;//0.0;

    SpkLiq_refractory[0] = _SpkLiq_refractory[0];//3E-3;
    SpkLiq_refractory[1] = _SpkLiq_refractory[1];//2E-3;
    SpkLiq_inhibitory_percentage = _SpkLiq_inhibitory_percentage;//20; //it means the liquid is going to have approximately 20% of inhibitory neurons


    // RANDOM-0: Membrane initial potentials
    // RANDOM-1: Noisy offset currents
    // RANDOM-2: Selection of the inhibitory and excitatory neurons
    // RANDOM-3: Internal connections of the liquid
    // RANDOM-4: Noisy corrents
    // User random seeds
    for(int i=0;i<5;i++)
        SpkLiq_user_seeds[i] = my_seeds[i];

    SpkLiq_threads_N = _SpkLiq_threads_N;

    return EXIT_SUCCESS;
}


// Exposes some of the variables
void reads_neurons(int *output){
    memcpy(output,SpkLiq_test_vthres,sizeof(int)*SpkLiq_number_of_neurons); // saves the spikes output from the liquid
}

void reads_membranes(double *output){
    memcpy(output,SpkLiq_neurons_membrane,sizeof(double)*SpkLiq_number_of_neurons); // saves the membranes voltages
}

void reads_exc_synapses(double *output){
    memcpy(output,SpkLiq_neurons_exc_curr,sizeof(double)*SpkLiq_number_of_neurons); // saves the excitatory currents
}

void reads_inh_synapses(double *output){
    memcpy(output,SpkLiq_neurons_inh_curr,sizeof(double)*SpkLiq_number_of_neurons); // saves the inhibitory currents
}

void reads_pre_i_connections(int *output){
    memcpy(output,SpkLiq_pre_i,sizeof(int)*SpkLiq_inh_connections); // saves the inhibitory currents
}

void reads_pre_i_weights(double *output){
    memcpy(output,SpkLiq_w_i,sizeof(double)*SpkLiq_inh_connections); // saves the inhibitory currents
}

void reads_pre_e_connections(int *output){
    memcpy(output,SpkLiq_pre_e,sizeof(int)*SpkLiq_exc_connections); // saves the inhibitory currents
}

void reads_pre_e_weights(double *output){
    memcpy(output,SpkLiq_w_e,sizeof(double)*SpkLiq_exc_connections); // saves the inhibitory currents
}

void reads_inhibitory_indices(int *output){
    memcpy(output,SpkLiq_inhibitory_indices,sizeof(int)*SpkLiq_number_of_inh_neurons); // saves the inhibitory currents
}

void reads_excitatory_indices(int *output){
    memcpy(output,SpkLiq_excitatory_indices,sizeof(int)*SpkLiq_number_of_exc_neurons); // saves the inhibitory currents
}

void reads_noisy_offset_currents(double *output){
    memcpy(output,SpkLiq_noisy_offset_currents,sizeof(double)*SpkLiq_number_of_neurons); // saves the inhibitory currents
}

void reads_noisy_currents(double *output){
    memcpy(output,SpkLiq_noisy_currents,sizeof(double)*SpkLiq_number_of_neurons); // saves the inhibitory currents
}


void stats_liquid(int *output){
    output[0]=SpkLiq_number_of_inh_neurons;
    output[1]=SpkLiq_number_of_exc_neurons;
    output[2]=SpkLiq_inh_connections;
    output[3]=SpkLiq_exc_connections;
}


void free_all(){

    // Pointers are used here because the memory size varies according to the user values
    free(SpkLiq_neurons_membrane); //the membrane potentials for all neurons
    free(SpkLiq_neurons_exc_curr); //excitatory currents levels for all neurons
    free(SpkLiq_neurons_inh_curr); //inhibitory currents levels for all neurons
    free(SpkLiq_refrac_timer); //timers used to control the refractory period
    free(SpkLiq_test_vthres); //controls which neuron is above the voltage threshold and should spike
    free(SpkLiq_noisy_offset_currents); //values of the neuron model constant noisy offset currents
    free(SpkLiq_noisy_currents); //values of the neuron model variable noisy currents
    free(SpkLiq_refrac_values); //stores the refractory periods for all neurons in the liquid 
                                         //because the refractory period is different for inhibitory and excitatory neurons

    free(SpkLiq_inhibitory_indices); //Indices of the inhibitory neurons inside the liquid
    free(SpkLiq_excitatory_indices); //Indices of the excitatory neurons inside the liquid
    free(SpkLiq_liquid_indices_shuffled); //used to generate the random SpkLiq_inhibitory_indices/SpkLiq_excitatory_indices


    free(SpkLiq_pre_i); //stores the indices of the inhibitory=>? pre-synaptic connections in the liquid
    free(SpkLiq_pos_i); //stores the indices of the inhibitory=>? pos-synaptic connections in the liquid
    free(SpkLiq_w_i); //stores the weights of the connections above

    free(SpkLiq_pre_e); //stores the indices of the inhibitory=>? pre-synaptic connections in the liquid
    free(SpkLiq_pos_e); //stores the indices of the inhibitory=>? pos-synaptic connections in the liquid
    free(SpkLiq_w_e); //stores the weights of the connections above

}

// Here I'm testing the simulator. In the future I want to call this simulator from Python passing all the parametres
// and also the data from where the input spikes are going to be read.
int main(int argc, char *argv[]){

    double time_begin, time_end;
    
    int number_of_iterations = 500;

    if (argc != 2)
    {
        fprintf(stderr,"Usage: %s N\n  where N is no. of thread\n", argv[0]);
        exit(EXIT_FAILURE);
    }

    SpkLiq_threads_N = atoi(argv[1]); //user input total number of threads

    if ((SpkLiq_threads_N < 1) || (SpkLiq_threads_N > MAX_THREAD))
    {
        fprintf(stderr,"The no of thread should between 1 and %d.\n", MAX_THREAD);
        exit(EXIT_FAILURE);
    }

    printf("Initializing the input variables...\n");

    // Network user parametres
    SpkLiq_net_shape[0] = 15;
    SpkLiq_net_shape[1] = 3;
    SpkLiq_net_shape[2] = 3;

    SpkLiq_lbd_value = 1.2;
    SpkLiq_step = 0.2E-3;
    SpkLiq_taum = 30E-3;
    SpkLiq_cm = 30E-9;
    SpkLiq_taue = 3E-3;
    SpkLiq_taui = 6E-3;

    SpkLiq_membrane_rand[0] = 13.5E-3;
    SpkLiq_membrane_rand[1] = 15E-3; //range of the possible values to the initialization of the membrane voltage

    SpkLiq_current_rand[0] = 14.975E-9;
    SpkLiq_current_rand[1] = 15.025E-9; //range of the possible values to the initialization of the constant noisy offset current

    SpkLiq_noisy_current_rand = 0.2E-9; //max range of the possible values to the initialization of the variable noisy current    

    SpkLiq_vresets = 13.5E-3;
    SpkLiq_vthres = 15E-3;
    SpkLiq_vrest = 0.0;

    SpkLiq_refractory[0] = 3E-3;
    SpkLiq_refractory[1] = 2E-3;

    SpkLiq_inhibitory_percentage = 20; //it means the liquid is going to have approximately 20% of inhibitory neurons

    // RANDOM-0: Membrane initial potentials
    // RANDOM-1: Noisy offset currents
    // RANDOM-2: Selection of the inhibitory and excitatory neurons
    // RANDOM-3: Internal connections of the liquid
    // RANDOM-4: Noisy corrents


    // User random seeds
    unsigned int my_seeds[5] = {1234,1234,1234,1234,1234};
    for(int i=0;i<5;i++)
        SpkLiq_user_seeds[i] = my_seeds[i];



    for(int k=0; k<1; k++){
        printf("Calling SpikingLiquid_init...\n");

        time_begin=clock_my_gettime();

        SpikingLiquid_init();
        
        time_end=clock_my_gettime();
        printf("Total init time (ms):%f\n", (time_end-time_begin)/1E6);    


        // Creates the input spikes
        int input_spikes_exc[50];
        double input_spikes_exc_w[50];
        for(int i=0; i<50; i++){
          input_spikes_exc[i]=i;
          input_spikes_exc_w[i]=1000E-9;
        }


        // Array to save the output spikes
        // int *output_spikes_mem = malloc(number_of_iterations*SpkLiq_number_of_neurons*sizeof(int));
        int **output_spikes_mem = malloc(number_of_iterations*sizeof(int*));

        for (int i=0; i<number_of_iterations; i++){
            output_spikes_mem[i] = malloc(SpkLiq_number_of_neurons*sizeof(int));
            
            if(output_spikes_mem[i]==NULL){
                printf("Error!!!\n");
                exit(1);
            }

        }

        printf("Allocated memory output(%dpositions):%luKbytes\n",number_of_iterations*SpkLiq_number_of_neurons ,(unsigned long int) ((number_of_iterations*SpkLiq_number_of_neurons*sizeof(int))/1000.0));

        // Gets the initial time in nanoseconds
        time_begin=clock_my_gettime();

        for(int i=0; i<50; i++){
            SpikingLiquid_update(NULL, NULL, NULL, NULL, 0, 0);
            reads_neurons(output_spikes_mem[i]);
            // reads_neurons((output_spikes_mem+i*SpkLiq_number_of_neurons*sizeof(int)));
            // memcpy((output_spikes_mem+i*sizeof(int*)),SpkLiq_test_vthres,sizeof(int)*SpkLiq_number_of_neurons); // saves the spikes output from the liquid
        } // i=0 to 49

        //Sends 50 spikes (0 to 49) @ i=50
        SpikingLiquid_update(input_spikes_exc, NULL, input_spikes_exc_w, NULL, 50, 0);

        reads_neurons(output_spikes_mem[50]);
        // reads_neurons((output_spikes_mem+50*SpkLiq_number_of_neurons*sizeof(int)));
        // memcpy((output_spikes_mem+50*sizeof(int*)),SpkLiq_test_vthres,sizeof(int)*SpkLiq_number_of_neurons); // saves the spikes output from the liquid


        for(int i=51; i<500; i++){
            SpikingLiquid_update(NULL, NULL, NULL, NULL, 0, 0);
            reads_neurons(output_spikes_mem[i]);
            // reads_neurons((output_spikes_mem+i*SpkLiq_number_of_neurons*sizeof(int)));
            // memcpy((output_spikes_mem+i*sizeof(int*)),SpkLiq_test_vthres,sizeof(int)*SpkLiq_number_of_neurons); // saves the spikes output from the liquid
        } // i=51 to 499

        // Gets the final time in nanoseconds
        time_end=clock_my_gettime();
        printf("Total update time (ms):%f\n", (time_end-time_begin)/1E6);
        printf("Average update time (ms):%f\n", ((time_end-time_begin)/1E6)/number_of_iterations);

        int c=0;
        for(int i=0; i<number_of_iterations; i++){
          printf("(%4.1f):",1E3*i*SpkLiq_step);
          for(int j=0;j<SpkLiq_number_of_neurons;j++){
              printf("%d",output_spikes_mem[i][j]);
              if(output_spikes_mem[i][j]){
                c++;
              }
              // printf("%d",*(output_spikes_mem+(i+j)*sizeof(int)));
              // if(*(output_spikes_mem+i*j*sizeof(int))){
              //   c++;
              // }
          }
          printf("\n");
        }

        printf("Total Spikes:%d\n",c);

        free_all();

    }
    return 0;


}
// // Here I'm testing the simulator. In the future I want to call this simulator from Python passing all the parametres
// // and also the data from where the input spikes are going to be read.
// int main(int argc, char *argv[]){

//     double time_begin, time_end;

//     int number_of_iterations = 500;
    
//     printf("main - Initializing the input variables...\n");

//     //
//     // USER PARAMETRES
//     //

//     // Network user parametres
//     SpkLiq_net_shape[0] = 15;
//     SpkLiq_net_shape[1] = 30;
//     SpkLiq_net_shape[2] = 3;

//     SpkLiq_lbd_value = 1.2;
//     SpkLiq_step = 0.2E-3;
//     SpkLiq_taum = 30E-3;
//     SpkLiq_cm = 30E-9;
//     SpkLiq_taue = 3E-3;
//     SpkLiq_taui = 6E-3;

//     SpkLiq_membrane_rand[0] = 13.5E-3;
//     SpkLiq_membrane_rand[1] = 15E-3; //range of the possible values to the initialization of the membrane voltage

//     SpkLiq_current_rand[0] = 14.975E-9;
//     SpkLiq_current_rand[0] = 15.025E-9; //range of the possible values to the initialization of the constant noisy offset current

//     SpkLiq_noisy_current_rand = 0.2E-9; //max range of the possible values to the initialization of the variable noisy current    

//     SpkLiq_vresets = 13.5E-3;
//     SpkLiq_vthres = 15E-3;
//     SpkLiq_vrest = 0.0;

//     SpkLiq_refractory[0] = 3E-3;
//     SpkLiq_refractory[1] = 2E-3;

//     SpkLiq_inhibitory_percentage = 20; //it means the liquid is going to have approximately 20% of inhibitory neurons


//     // RANDOM-0: Membrane initial potentials
//     // RANDOM-1: Noisy offset currents
//     // RANDOM-2: Selection of the inhibitory and excitatory neurons
//     // RANDOM-3: Internal connections of the liquid
//     // RANDOM-4: Noisy corrents
//     // User random seeds
//     unsigned int my_seeds[5] = {1234,1234,1234,1234,1234};
//     for(int i=0;i<5;i++)
//         SpkLiq_user_seeds[i] = my_seeds[i];


//     if (argc != 2)
//     {
//         fprintf(stderr,"Usage: %s N\n  where N is no. of thread\n", argv[0]);
//         exit(EXIT_FAILURE);
//     }

//     SpkLiq_threads_N = atoi(argv[1]); //user input total number of threads

//     if ((SpkLiq_threads_N < 1) || (SpkLiq_threads_N > MAX_THREAD))
//     {
//         fprintf(stderr,"The no of thread should between 1 and %d.\n", MAX_THREAD);
//         exit(EXIT_FAILURE);
//     }

//     for(int k=0; k<2; k++){
//       //
//       // Beginning of the simulation
//       printf("Calling SpikingLiquid_init...\n");

//       // Gets the initial time in nanoseconds
//       time_begin=clock_my_gettime();

//       SpikingLiquid_init();
      
//       time_end=clock_my_gettime();
//       printf("Total init time (ms):%f\n", (time_end-time_begin)/1E6);    

      
//       // Generates the input spikes (exc)
//       int input_spikes_exc[50];
//       double input_spikes_exc_w[50];
//       for(int i=0; i<50; i++){
//         input_spikes_exc[i]=i;
//         input_spikes_exc_w[i]=100E-9;
//       }


//       // Allocates the memory for the output
//       int *output_spikes_mem = NULL;
//       output_spikes_mem = (int*) malloc(number_of_iterations*sizeof(int*)+number_of_iterations*SpkLiq_number_of_neurons*sizeof(int));
//       if (output_spikes_mem==NULL){
//           fprintf(stderr,"output_spikes_mem - malloc ERROR!");
//           exit(EXIT_FAILURE);        
//       }

//       printf("Allocated memory output:%luMbytes\n", (unsigned long int) ((number_of_iterations*sizeof(int*)+number_of_iterations*SpkLiq_number_of_neurons*sizeof(int))/1000000.0));


//       // Gets the initial time in nanoseconds
//       time_begin=clock_my_gettime();

//       for(int i=0; i<number_of_iterations; i++){
//           if(i==50)
//             SpikingLiquid_update(input_spikes_exc, NULL, input_spikes_exc_w, NULL, 50, 0);
//           else
//             SpikingLiquid_update(NULL, NULL, NULL, NULL, 0, 0);

//           memcpy((output_spikes_mem+i*SpkLiq_number_of_neurons),SpkLiq_test_vthres,sizeof(int)*SpkLiq_number_of_neurons); // saves the spikes output from the liquid
//       }

//       // Gets the final time in nanoseconds
//       time_end=clock_my_gettime();
//       printf("Total update time (ms):%f\n", (time_end-time_begin)/1E6);
//       printf("Average update time (ms):%f\n", ((time_end-time_begin)/1E6)/number_of_iterations);


//       // Analyses the output
//       int c=0;
//       for(int i=0; i<number_of_iterations; i++){
//         for(int j=0;j<SpkLiq_number_of_neurons;j++){
//             if(*(output_spikes_mem+i*sizeof(int*)+j*sizeof(int)))
//               // printf("%d",*(output_spikes_mem+i*sizeof(int*)+j*sizeof(int)));
//               c++;  
//         }
//       }
//       printf("Average Spikes:%f\n",(double)c/number_of_iterations);
//       printf("Total Spikes:%d\n",c);
//     }

//     return 0;

// //
// // MY NEXT STEP IS REPRODUCE THE SAME SIMULATION I DID IN BRIAN AND IN MY PYTHON SIMULATOR
// //


// }
```

```python
# Update the simulation (step) and returns the current time
def updates_sim(input_spikes_exc, input_spikes_inh, input_spikes_exc_w, input_spikes_inh_w, size_exc, size_inh):
    input_spikes_exc = numpy.array(input_spikes_exc,dtype=numpy.int32)
    input_spikes_inh = numpy.array(input_spikes_inh,dtype=numpy.int32)
    input_spikes_exc_w = numpy.array(input_spikes_exc_w,dtype=numpy.double)
    input_spikes_inh_w = numpy.array(input_spikes_inh_w,dtype=numpy.double)
    size_exc = ctypes.c_int32(size_exc)
    size_inh = ctypes.c_int32(size_inh)
    return liquid_update(input_spikes_exc,input_spikes_inh,input_spikes_exc_w,input_spikes_inh_w,size_exc,size_inh)
```

```python
voltages = []
def compare_to_brian():
    global voltages
    voltages = []
    # Initialize the arrays with the simulated values
    # and saves the values for t=0s
    t = []
    spikes_idx = numpy.arange(NofN)
    spikes = []
    initialize_sim()
    tstp = 0.2E-3
    ti=0
    spks = output_sim(NofN)
    t.append(ti)
    spikes.append(spikes_idx[spks])
    voltages.append(output_voltages(NofN).mean())

    for i in range(50):
        # Runs one step and save the result
        updates_sim([], [], [], [], 0, 0)

        # Appends to the output
        spks = output_sim(NofN)
        ti+=tstp
        t.append(ti)
        spikes.append(spikes_idx[spks])
        voltages.append(output_voltages(NofN).mean())
        
    AINP=100

    # Injects some spikes into the liquid!
    exc_inputs = range(50)
    inh_inputs = []
    exc_weights = [AINP*1E-9]*50
    inh_weights = []
    updates_sim(exc_inputs, inh_inputs, exc_weights, inh_weights,50,0)

    # Appends to the output
    spks = output_sim(NofN)
    ti+=tstp
    t.append(ti)
    spikes.append(spikes_idx[spks])
    voltages.append(output_voltages(NofN).mean())
    
    for i in range(5450):
        # Runs one step and save the result
        updates_sim([], [], [], [], 0, 0)
        # Appends to the output
        spks = output_sim(NofN)
        ti+=tstp
        t.append(ti)
        spikes.append(spikes_idx[spks])
        voltages.append(output_voltages(NofN).mean())
    return t,spikes
```
